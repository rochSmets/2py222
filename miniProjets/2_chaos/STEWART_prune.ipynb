```python
#importation des modules
import numpy as np
import matplotlib.pyplot as plt

#initialisation des variables 
x0 = 0.1
r = 2.5
N = 50

#definition de la fonction F
def F(x) :
    return r*(1-x)*x

#liste des éléments 
liste_element = [x0]

#creation de boucle
for k in range(N):
    x1 = F(x0)
    liste_element.append(x1)
    #recurrence :
    x0 = x1

# Tracé des éléments de la suite
plt.plot(range(N+1), liste_element, 'g')
plt.show()

#on constate que la suite à une limite finie en 0,6. 
```

```python
# initialisation des variables 
x0 = 0.001
r = 2.5
N = 1000

#definition de la fonction F
def F(x) :
    return r*(1-x)*x

#liste des éléments 
liste_element = np.zeros(1000, dtype = 'float64')


#creation de boucle
for r in np.linspace(2.5,4.0,1000) :
    for k in range(N):
            x1 = F(x0)
            liste_element[k] = x1
            #recurrence :
            x0 = x1           
    liste_r = liste_element[900:]
    plt.plot(np.ones(100)*r, liste_r, ',' )
plt.title('Diagramme de bifurcation')
plt.xlabel('Valeurs de r')
plt.ylabel('les 100 dernier termes de la suite') 
plt.show()  
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Paramètres initiaux pour le calcul de l'exposant de Lyapunov
#valeurs de r entre 2.5 et 4.0
r_linspace = np.linspace(2.5, 4.0, 1000)  
x0 = 0.00001
#Nombre total d'itérations pour stabiliser
N = 1000 

#liste pour stocker les valeurs de l'exposant de Lyapunov
lyapunov_exposants = []

#Calcul de l'exposant de Lyapunov pour chaque valeur de r
for r in r_linspace:
    x = x0
    #Initialisation de la somme pour le calcul de l'exposant
    lyapunov_somme = 0  
    
    #Itéreration pour calculer l'exposant de Lyapunov après la stabilisation
    for i in range(N):
        x = r*x*(1-x)
        lyapunov_somme += np.log(abs(r-2*r*x))
    
    lyapunov_exposant = lyapunov_somme / N # on fait la moyenne
    lyapunov_exposants.append(lyapunov_exposant) #ajout d'un therme a la liste

#Convertion de la liste en un tableau numpy
lyapunov_exposants = np.array(lyapunov_exposants)

#Initialisation et paramétrisation du grapgique
plt.figure(figsize=(12, 7))
plt.plot(r_linspace, lyapunov_exposants, ',k',)
plt.axhline(0, color='blue', lw=0.5,)
plt.title('Exposant de Lyapunov en fonction de r')
plt.xlabel('Valeur de r')
plt.ylabel('Exposant de Lyapunov')
plt.grid(True)
plt.show()
```
