```python
import numpy as np
import matplotlib.pyplot as plt
```

## Comportement de la suite ##

```python
r = 2.5
x0 = 0.00001

def f(x,r) :
    return r * (1-x)*x
def X_n(r,x0,N=100) :
    x_n = np.ones(100)
    x_n[0] = x0
    for i in range(1,N):
        x_n[i] = f(x_n[i-1],r)
    return x_n

n = np.linspace(0,99,100)
x_n1 = X_n(r,x0)
x_n2 = X_n(3.0,x0)
x_n3 = X_n(3.5,x0)
x_n4 = X_n(4.0,x0)

plt.plot(n,x_n1)
plt.show()
plt.plot(n,x_n2)
plt.show()
plt.plot(n,x_n3)
plt.show()
plt.plot(n,x_n4)
plt.show()
```

On remarque que la suite varie de manire diffrente selon la valeur de r.

## Diagramme de bifurcation ##

```python
def last_U_n(r,x0 = 0.00001,n = np.linspace(0,999,1000)) :
    x_n = np.zeros(100)
    x_i = x0
    for i in range(0,len(n)-100) :
        x_i = f(x_i,r)
    x_n[0] = f(x_i,r)
    for i in range(1,100) :
        x_n[i] = f(x_n[i-1],r)
    plt.scatter(r*np.ones(100),x_n,color='k',marker='.',s=1)
    return x_n

nbr_r = 500
rs = np.linspace(2.5,4,nbr_r)
for r in rs :
    last_U_n(r)
plt.show()
for r in rs :
    last_U_n(r,x0 = 0.001)
plt.show()
```

La limite de l'quation logistique n'xiste que pour certaine valeur de r. Losque l'quation logistique tend pas vers une limite, le nombre de valeur diffrente qu'elle va prendre dpend de r. Le diagramme de bifurcation ne dpend pas de x0.

## Exposant de Lyapunov ##

```python
def f_d(x,r):
    return r*(1-2*x)

def U_n(r,n_inf,x0 = 0.00001):
    u_n = np.zeros(n_inf)
    u_n[0] = x0
    for i in range(1,n_inf) :
        u_n[i] = f(u_n[i-1],r)
    return u_n

def lambda_r(r,n_inf,x0 = 0.00001):
    x_n = U_n(r,n_inf,x0)
    logs_f = np.log(abs(f_d(x_n,r)))
    return np.sum(logs_f)/n_inf

nbr_r = 2000
rs = np.linspace(2.5,4,nbr_r)
n_inf = 500
lambdas = np.zeros(len(rs))

for i,r in enumerate(rs):
    lambdas[i] = lambda_r(r,n_inf)

lambda_neg = lambdas<=0
lambda_pos = lambdas>0
plt.scatter(rs[lambda_neg],lambdas[lambda_neg],color='k',marker='.',s=1)
plt.scatter(rs[lambda_pos],lambdas[lambda_pos],color='r',marker='.',s=1)
plt.plot(rs,np.zeros(len(rs)),"b")
plt.show()
```

On observe que les bifurcation se produiseent lorsque lambda > 0. La stabilit de l'quation logistique dpend bien de r.
