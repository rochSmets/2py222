# Écrire du texte dans un calepin jupyter

Il est possible d'écrire du texte mis en forme dans un calepin jupyter. Pour cela, il suffit d'écrire du texte  dans une cellule, sélectionner pour la cellule un type `Markdown` (Menu Cellule > Type de cellule > Markdown), puis d'exécuter la cellule comme d'habitude avec `Ctrl + Entrée`. Vous devriez alors obtenir votre texte plus joliment formaté. Pour modifier la cellule, il faut lui double-cliquer dessus (ou la sélectionner et appuyer sur `Entrée`), puis à nouveau l'exécuter après les modifications.


## Écrire des formules mathématiques

Pour pouvoir écrire des formules mathématiques dans un calepin, il va falloir les écrire avec une syntaxe inspirée de LaTeX qui est un standard dans l'édition scientifique. Comme toujours, n'hésitez pas à regarder comment sont écrites les formules mathématiques dans vos calepins de cours pour vous en inspirer. 

Pour écrire une formule mathématique, il faut l'entourer de `$…$` pour l'insérer au milieu d'une ligne, ou de `$$…$$` pour mettre la formule à part. Observez la différence :
* `$a = 1$` : la formule est insérée au milieu du texte $a = 1$, et l'on peut continuer sa phrase. 
* `$$a = 1$$` : la formule est au mise en évidence $$a = 1$$

Les formules simples s'écrivent naturellement : `b = a + 1` donne $b = a + 1$. 
Mais la grande force de LaTeX est l'existence de commandes pour écrire un très grand nombre de symboles mathématiques. 

Une commande LaTeX s'écrit en général sous la forme `\nom_de_commande`, et `\nom_de_commande{arg}` si elle prend un argument. 

Voici une liste des principales commandes, qui peuvent ensuite se combiner entre elles pour donner des formules complexes. 

### Fonctions de base

Nom | Exemple | Résultat
--- | --- | ---
Exposant | `a^b` | $a^b$
Exposant multi-caractères | `a^{2b}` | $a^{2b}$
Indice | `a_b` | $a_b$
Indice multi-caractères | `a_{2b}` | $a_{2b}$
Fraction | `\frac{a}{b}` | $\frac{a}{b}$
Racine carrée | `\sqrt{a}` | $\sqrt{a}$

### Fonctions standards

Par défaut, le texte sera interprété par LaTeX
comme des variables et affiché en italique.
Si l'on veut écrire des noms de fonction, la norme est de les écrire en lettres romanes.
Pour les fonctions courantes, il suffit d'utiliser `\nom_de_la_fonction`. Il est aussi possible de directement demander une police romane avec `\mathrm{}` (`$\mathrm{sin}$` : $\mathrm{sin}$)

Nom | Exemple | Résultat
--- | --- | ---
Fonctions trigonométriques | `\cos(a + b) = \cos a \cos b - \sin a \sin b` | $\cos(a + b) = \cos a \cos b - \sin a \sin b$
Fonctions exponentielles | `a^b = \exp(b \ln a)` | $a^b = \exp(b \ln a)$

### Lettres grecques 

Il suffit d'écrire `\nom_de_la_lettre` avec une première lettre en capitale ou non selon si l'on veut la lettre capitalisée ou non

Nom | Exemple | Résultat
--- | --- | ---
alpha | `\alpha` | $\alpha$ 
pi | `\pi` | $\pi$ 
Pi | `\Pi` | $\Pi$
lambda | `\lambda` | $\lambda$
Lambda | `\Lambda` | $\Lambda$
epsilon | `\epsilon` | $\epsilon$
varepsilon | `\varepsilon` | $\varepsilon$
phi | `\phi` | $\phi$
varphi | `\varphi` | $\varphi$

Etc. Notez au passage la différence entre `\epsilon` et `\varepsilon` : on préférera toujours `\varepsilon` qui est bien plus lisible. Et pour le choix entre `phi` ou `varphi`, ça dépend du contexte.

### Intégrales et autres opérations

Nom | Exemple | Résultat
--- | --- | ---
Intégrale | `\int` | $\int$
Intégrale définie | `\int_a^b f(x) \,\mathrm{d}x`| $\int_a^b  f(x) \,\mathrm{d}x$
Intégrale double | `\iint` | $\iint$
Somme | `\sum_{i=0}^n u_i` | $\sum_{i=0}^n u_i$
Limite | `\lim_{x \to 0} f(x)` | $\lim_{x \to 0} f(x)$

Il est aussi possible d'écrire des parenthèses dont la taille dépend du contenu de la parenthèse, avec la syntaxe `\left( … \right)`. Ça marche aussi avec `\{` et `\}` (notez l'antislash d'échappement), `[` et `]`, et pour les chevrons il faut utiliser `\langle` et `\rangle`. Observez la différence avec des parenthèses « standards » :
$$(\frac{a}{b}), \left(\frac{a}{b}\right)$$
$$\{\frac{a}{b}\}, \left\{\frac{a}{b}\right\}$$
$$[\frac{a}{b}], \left[\frac{a}{b}\right]$$
$$\langle\frac{a}{b}\rangle, \left\langle\frac{a}{b}\right\rangle$$

Nom | Exemple | Résultat

Vecteur unitaire | `\hat{a}` | $\hat{a}$
Dérivées temporelles | `\dot{a}, \ddot{a}` | $\dot{a}, \ddot{a}$

--- | --- | ---

Si vous voulez un symbole non recensé dans cette table, il existe très probablement, il faut seulement trouver son nom via votre moteur de recherche favori. 

### Matrices

On peut aussi facilement faire des matrices, comme ceci :
```
$$
A\in\mathcal{M}_{MN}(\mathbb{C}), 
A = 
\begin{pmatrix}
   a_{11} & \cdots & a_{1N} \\
   \vdots & \ddots & \vdots \\
   a_{M1} & \cdots & a_{MN}
\end{pmatrix}
$$
```

$$
A\in\mathcal{M}_{MN}(\mathbb{C}), 
A = 
\begin{pmatrix}
   a_{11} & \cdots & a_{1N} \\
   \vdots & \ddots & \vdots \\
   a_{M1} & \cdots & a_{MN}
\end{pmatrix}
$$

### Gestion des espaces

Par défaut, LaTex gére tout seul à peu près correctement les espaces. Mais vous pouvez en rajouter  manuellement avec les commandes suivantes, classées par taille croissante d'espace  :

Nom | Exemple | Résultat
--- | --- | ---
Espacement négatif  | `a \! b` | $a \! b$
Pas d'espacement | `a b` | $a b$
Espace fine | `a\,b` | $a \, b$
Espace moyenne | `a\;b` | $a \; b$
Espace normale | `a\ b` | $a \ b$
Espace cadratin | `a \quad b` | $a \quad b$
Espace double cadratin | `a \qquad b` | $a \qquad b$
