## Particule dans un cyclotron

1. Les instants $t_n$ où le proton sort des D (soit inférieur, soit supérieur, voir la figure) sont donnés, **sans unité**, par
$$t_n=n+\tau \sqrt{n}.$$
On a $\tau=0.5$. 

```python
# Import des modules utiles
import numpy as np

#Creation du tablleau de 51 entier pour n
n= np.array(range(51))

#Tableau des instants de sorties
tn=np.ones(51)

#recurrence, les 50 instants de sortie
T= 0.5
tn[0] = 0
for i in range(0,51) :
    tn[i] = i + T*np.sqrt(i)
```

2. On vérifie maintenant que $t_1=1.5$, $t_4=5$, $t_9=10.5$, $t_{16}=18$, et $t_{25}=27.5$ avec des opérateurs booléens.

```python
# Vérification avec opérateurs booléens
print(tn[1] == 1.5)
print(tn[4] == 5)
print(tn[9] == 10.5)
print(tn[16] == 18)
print(tn[25] == 27.5)
```

3. On calcule les valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$.
On utilise la formule donnée pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$


```python
#choix des valeurs pour le temps
y=np.arange(tn[2]-1,tn[2],1000)

#iteration pour avoir vx et vy
for i in y:
    Vxi=0
    Vyi=0
    Vxi+=np.sqrt(1+1)*np.sin(np.pi*(i-(i+1)))
    Vyi+=-1*np.sqrt(1+1)*np.cos(np.pi*(i-(i+1)))                 

#choix des valeurs pour le temps
u=np.arange(tn[4]-1,tn[4],1000)

#iteration pour avoir vx et vy
for i in u :
    Vxu=0
    Vyu=0
    Vxu +=np.sqrt(1+1)*np.sin(np.pi*(i-(i+1)))
    Vyu +=-1*np.sqrt(1+1)*np.cos(np.pi*(i-(i+1)))                 
    
```

4. On représente un graphique pour $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $t_2-1$ a $t_2$ et de même pour $t$ variant de $t_4-1$ à $t_4$.

```python
import matplotlib.pyplot as plt
```

```python
#Graphiaue
plt.plot(y, Vxi)
plt.title("Vx en fonction de t ")
plt.xlabel("t")
plt.ylabel("vx")
plt.show()

plt.plot(y, Vyi)
plt.title("Vy en fonction de t ")
plt.xlabel("t")
plt.ylabel("vy")
plt.show()
```
