## Partie 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Définition du tableau t 
t=[]
tau=0.5
for n in range (51):
    t.append(n+tau*np.sqrt(n))
```

```python
print(t[1]==1.5)
print(t[4]==5)
print(t[9]==10.5)
print(t[16]==18)
print(t[25]==27.5)
```

```python
temps_2=np.linspace(t[2]-1,t[2],50 )# définition des tableaux de temps dans les intervalles demandés
temps_4=np.linspace(t[4]-1,t[4],50)
def vx(temps,n):
    """Fonction vitesse selon x"""
    return (-1)**n*np.sqrt(n+1)*np.sin(np.pi*(temps-t[n+1]))
def vy(temps,n):
    """Fonction vitesse selon y"""
    return -(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(temps-t[n+1]))

vx_2=vx(temps_2,1)
vy_2=vy(temps_2,1)

vx_4=vx(temps_4,3)
vy_4=vy(temps_4,3)
```

```python
# On trace les graphes
plt.plot(temps_2,vx_2,label="vx")
plt.plot(temps_2,vy_2,label="vy")
plt.xlabel ("vitesses sans unités")
plt.ylabel("temps sans unité")
plt.title("Graphe de vx et vy en fonction de t choisi entre t2-1 et t2")
plt.legend()
plt.show()
```

```python
plt.plot(temps_4,vx_4,label="vx")
plt.plot(temps_4,vy_4,label="vy")
plt.xlabel ("vitesses sans unités")
plt.ylabel("temps sans unité")
plt.title("Graphe de vx et vy en fonction de t choisi entre t4-1 et t4")
plt.legend()
plt.show()
```

```python
On extrait les maximums
for n in range (len(vx_2)):
    if vx_2[n]==np.max(vx_2):
        print("le maximum de vx2 est :",np.max(vx_2))
        print("le temps correspondant est:",temps_2[n])# On extrait le temps associé à la valeur maximale
    if vy_2[n]==np.max(vy_2):
        print("le maximum de vy2 est :",np.max(vy_2))
        print("le temps correspondant est:",temps_2[n])
```

```python
for n in range (len(vx_4)):
    if vx_4[n]==np.max(vx_4):
        print("le maximum de vx4 est :",np.max(vx_4))
        print("le temps correspondant est:",temps_4[n])
    if vy_4[n]==np.max(vy_4):
        print("le maximum de vy4 est :",np.max(vy_4))
        print("le temps correspondant est:",temps_4[n])
```

On observe que la vitesse du proton augmente entre les intervalles $[t_2-1;t_2]$ et $[t_4-1;t_4]$. Or le cyclotron accèlère le proton grace à cette trajectoire, notre observation est donc bien cohérente avec le problème étudié. 

## Partie 2:

```python
def vxmin(temps,n):
    """ fonction vitesse selon x pour t dans l'intervalle inferieur"""
    return np.zeros(len(temps))
def vymin(temps,n):
     """ fonction vitesse selon y pour t dans l'intervalle inferieur"""
    return (-1)**n*(np.sqrt(n)+(temps-t[n])/tau)
def vxsup(temps,n):
     """ fonction vitesse selon x pour t dans l'intervalle superieur"""
    return (-1)**n*np.sqrt(n+1)*np.sin(np.pi*(temps-t[n+1]))
def vysup(temps,n):
     """ fonction vitesse selon x pour t dans l'intervalle superieur"""
    return -(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(temps-t[n+1]))

```

```python
for n in range (6):
    # Pour chaque intervalle on trace vmin et vsup
    tmin=np.linspace(t[n],t[n+1]-1,20)
    plt.plot(tmin,vxmin(tmin,n),color="blue",)
    plt.plot(tmin,vymin(tmin,n),color="r",)
    tmax=np.linspace(t[n+1]-1,t[n+1],20)
    plt.plot(tmax,vxsup(tmax,n),color="blue")
    plt.plot(tmax,vysup(tmax,n),color="r")
plt.title("Vitesses en x et y dans un intervalle de temps 0, t6")    
plt.xlabel("temps sans unité")
plt.ylabel("vitesses (vx en bleu et vy en rouge) sans unité")
```

On remarque que vy et vx sont des fonctions continues du temps, en effet cela semble approprié car vy et vx représentent la vitesse du proton selon ses coordonnées respectives y et x. Or la vitesse ne peut être discontinue. 

```python
VX=[]
VY=[]
T=[]

for n in range (6):
    tmin=np.linspace(t[n],t[n+1]-1,20)
    tmax=np.linspace(t[n+1]-1,t[n+1],20)
    VX.append(vxmin(tmin,n))
    VX.append(vxsup(tmax,n))
    VY.append(vymin(tmin,n))
    VY.append(vymin(tmin,n))
    T.append(tmin)
    T.append(tmax)
V=[]
for i in range (len(T)): # calcul du module de vitesse 
    V.append(np.sqrt(VX[i]**2+VY[i]**2))

plt.plot(T,V,"b")
plt.title("Norme de la vitesse")
plt.xlabel("temps sans unité")
plt.ylabel("vitesse")
plt.show()
```

On observe bien que la vitesse augmente à chaque tour. Le cyclotron permet donc d'accélerer une particule en la maintenant sur une trajectoire circulaire plutot que linéaire comme les accélérateurs de pârticules classiques. 

```python
VX=[]
VY=[]
T=[]

for n in range (50):
    tmin=np.linspace(t[n],t[n+1]-1,20)
    tmax=np.linspace(t[n+1]-1,t[n+1],20)
    VX.append(vxmin(tmin,n))
    VX.append(vxsup(tmax,n))
    VY.append(vymin(tmin,n))
    VY.append(vymin(tmin,n))
    T.append(tmin-n)
    T.append(tmax-n)
V=[]
for i in range (len(T)): 
    V.append(np.sqrt(VX[i]**2+VY[i]**2))

plt.plot(T,V,"b")
plt.title("Norme de la vitesse")
plt.xlabel("temps total passé par le proton entre les D à l'instant  tn")
plt.ylabel("vitesse")
plt.show()
```

On observe que le module de la vitesse augmente linéairement par rapport au temps total passé par le proton entre les D à l'instant 𝑡𝑛.
