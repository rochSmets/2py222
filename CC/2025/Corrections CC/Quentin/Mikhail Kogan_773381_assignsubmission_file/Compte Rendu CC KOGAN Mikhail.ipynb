```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Définition des constantes et autres paramètres utiles
Tau = 0.5
epsilon = (-1)**i

# Création d'un tableau avec les valeurs de t_n
tn = []                          
for i in range(0, 50):
    tn.append(i + Tau* np.sqrt(i))

# Vérification que les valeurs de temps correspondent à celles attendues
print("t1 = 1.5 : ",tn[1] == 1.5)
print("t4 = 5 : ",tn[4] == 5)
print("t9 = 10.5 : ",tn[9] == 10.5)
print("t16 = 18 : ",tn[16] == 18)
print("t25 = 27.5 : ",tn[25] == 27.5)

### Ici on fait les calculs pour  t_2-1 < t < t_2
# Création d'un nouveau tableau subdivisant en 50 valeurs le temps entre les instants t_2-1 et t_2
t = np.linspace(tn[2]-1, tn[2])
n = 2

#Calcul de la vitesse suivant x entre ces moments
vx = epsilon*np.sqrt(n+1)*np.sin(t - tn[n])
#Calcul de la vitesse suivant y entre ces moments
vy = -epsilon*np.sqrt(n+1)*np.cos(np.pi*(t - tn[n]))

# Valeurs maximales
a = vx[1]
b= vy[49]

# Tracé du graphe de ces deux vitesses en fonction du temps
plt.figure(1)
plt.title("Evolution temporelle de la vitesse entre les moments $t_2 -1$ et $t_2$")
plt.plot(t,vx, "-r", label = "$v_x$")
plt.plot(t,vy,"-g", label = "$v_y$")
plt.xlabel("temps (s)")
plt.ylabel("vitesse ($m.s^{-1}$)")
plt.legend()
plt.show()

### Ici on fait les calculs pour  t_4-1 < t < t_4
# Création d'un nouveau tableau subdivisant en 50 valeurs le temps entre les instants t_4-1 et t_4
t = np.linspace(tn[4]-1, tn[4])
n = 4

#Calcul de la vitesse suivant x entre ces moments
vx = epsilon*np.sqrt(n+1)*np.sin(t - tn[n])

#Calcul de la vitesse suivant y entre ces moments
vy = -epsilon*np.sqrt(n+1)*np.cos(np.pi*(t - tn[n]))

# Valeurs maximales
c = vx[1]
d = vy[49]

# Tracé du graphe de ces deux vitesses en fonction du temps
plt.figure(2)
plt.title("Evolution temporelle de la vitesse entre les moments $t_4 -1$ et $t_4$")
plt.plot(t,vx, "-r", label = "$v_x$")
plt.plot(t,vy,"-g", label = "$v_y$")
plt.xlabel("temps (s)")
plt.ylabel("vitesse ($ m.s^{-1}$)")
plt.legend()
plt.show()

# Extraction des valeurs maximales:

print(" Pour t_2-1 < t < t_2 : ")
print("valeur maximale de vx :", a)
print("valeur maximale de vy : ", b)
print(" Pour t_4-1 < t < t_4 : ")
print("valeur maximale de vx :", c)
print("valeur maximale de vy : ", d)
```

D'abord nous observons que les graphes ont tous les deux les mèmes formes, ce qui est normal vu que la particule parcourt des portions du synchrotron ayant une forme similaire.  Nous constatons néanmoins que  les vitesses pour $t_2-1 < t < t_2$ sont inférieures que pour $t_4-1 < t < t_4$. La particule accélère donc.
