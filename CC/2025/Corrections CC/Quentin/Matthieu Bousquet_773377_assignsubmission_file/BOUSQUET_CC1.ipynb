# Particule dans un cyclotron

### 1) Rappel des formules données: 

Les instants $t_n$ où le proton sort des D (soit inférieur, soit supérieur, voir la figure) sont donnés, **sans unité**, par
$$t_n=n+\tau \sqrt{n}.$$
On choisit $\tau=0.5$. 
1) Pour $t$ variant de $t_n$ à $t_{n+1}-1$ c’est-à-dire quand il est entre les D :
$$
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$$
2) Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

## Partie 1:

```python
import matplotlib.pyplot as plt
import numpy as np

n = 50

tableau_D = np.zeros(n)

for i in range(n):
    tableau_D[i] = i + 0.5*np.sqrt(i)
print("t1 =",tableau_D[1])
print("t4 =",tableau_D[4])
print("t16 =",tableau_D[16])
print("t25 =",tableau_D[25])
```

```python
t_2 = np.linspace(tableau_D[2]-1,tableau_D[2],10) # Creation d'un intervalle de t pour t = t2
#Creation de tableaux 
vx_t2 = np.zeros(len(t_2))
vy_t2 = np.zeros(len(t_2))

# Boucle qui permet de calculer la vitesse Vx et Vy
for n in range(len(t_2)-1):
    vx_t2[n] = ((-1)**n)*np.sqrt(n+1)*np.sin(np.pi*(t_2[n]-t_2[n+1]))
    vy_t2[n] = -((-1)**n)*np.sqrt(n + 1)*np.cos(np.pi*(t_2[n] - t_2[n+1]))


# Representaion graphique de vx_t2 et vy_t2
plt.plot(vx_t2,t_2, label= "vitesse x")
plt.plot(vy_t2,t_2, label = "vitesse y")
plt.title("Vitesse vx et vy pour $t$ variant de $t_2-1$ à $t_2$ ")
plt.xlabel(" $t$ variant de $t_2-1$ à $t_2$")
plt.ylabel("Vitesse vx et vy")
plt.grid()
plt.legend()
plt.show()
```

En affichant le tableau complet pour vx_t2 et vy_t2 on repere facilement le maxima correspondant:

```python
print("maxima de la vitesse selon x pour t2:",abs(vx_t2[-2]))
print("maxima de la vitesses selon y pour t2:",abs(vy_t2[-2]))
```

```python
t_4 = np.linspace(tableau_D[4]-1,tableau_D[4],10) # Creation d'un intervalle de t pour t = t4
#Creation de tableaux 
vx_t4 = np.zeros(len(t_4))
vy_t4 = np.zeros(len(t_4))

# Boucle qui permet de calculer la vitesse Vx et Vy
for n in range(len(t_4)-1):
    vx_t4[n] = ((-1)**n)*np.sqrt(n+1)*np.sin(np.pi*(t_4[n]-t_4[n+1]))
    vy_t4[n] = -((-1)**n)*np.sqrt(n + 1)*np.cos(np.pi*(t_4[n] - t_4[n+1]))
    
# Representaion graphique de vx_t4 et vy_t4
plt.plot(vx_t4,t_4, label= "vitesse x")
plt.plot(vy_t4,t_4, label = "vitesse y")
plt.title("Vitesse vx et vy pour $t$ variant de $t_4-1$ à $t_4$ ")
plt.xlabel(" $t$ variant de $t_4-1$ à $t_4$")
plt.ylabel("Vitesse vx et vy")
plt.grid()
plt.legend()
plt.show() 

print("tableau des vitesses selon x:",vx_t4)
print("tableau des vitesses selon y:",vy_t4)
```

En faisant de même pour trouver les maximas, on trouve:

```python
print("maxima de la vitesse selon x pour t4:",abs(vx_t2[-2]))
print("maxima de la vitesses selon y pour t4:",abs(vy_t2[-2]))
```

On remarque que pour les 2 instants t étudié, on a retrouve le même maxima pour la vitesse selon x et la vitesse selon y. On peut dire que vx et vy sont des fonctions continues.
