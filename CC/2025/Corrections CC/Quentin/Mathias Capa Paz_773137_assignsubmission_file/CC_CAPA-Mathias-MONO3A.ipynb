## Partie 1

1. On définit une fonction pour calculer les termes de la suite, puis on les calcule les 50 premiers termes:

```python
import numpy as np
import matplotlib.pyplot as plt


# On définit la suite
def suite(n):
    return n + 0.5*np.sqrt(n)

# On crée le tableau des temps (des n)
temps = [suite(k) for k in range(0,51)]
```

2. On verifie que les valeurs correpondent bien à celles attendues.

```python
# Des instructions boléeenes pour vérifier les valeurs trouvées avec celles de l'exercice

if temps[1] == 1.5:
    print("La valeur 1 coincide")
if temps[4] == 5.0:
    print("La valeur coincide")
if temps[9] == 10.5:
    print("La valeur coincide")
if temps[16] == 18.0:
    print("La valeur coincide")
if temps[25] == 27.5:
    print("La valeur coincide")
```

3. On calcule les valeurs de $v_x$ et $v_y$ :

```python
temps_n2 = np.linspace(temps[2]-1,temps[2],100)  # Tableau des temps régulièrement spacés pour entre t_2 - 1 et t_2
temps_n4 = np.linspace(temps[4]-1,temps[4],100)  # Tableau des temps régulièrement spacés pour entre t_2 - 1 et t_2

# On définit les focntions pour tracer les vitesses en fonction du temps

def v_dans_D (tableau,n):
    vx = []
    vy = []
    for i in range (len(tableau)):
        v_x =  (-1)**n * ( np.sqrt(n+1) * np.sin(np.pi*(tableau[i]-temps[n+1])) )
        v_y = -(-1)**n * ( np.sqrt(n+1) * np.cos(np.pi*(tableau[i]-temps[n+1])) )
        vx.append(v_x)
        vy.append(v_y)
    return vx, vy

def v_entre_D (tableau,n):
    vx = []
    vy = []
    for i in range (len(tableau)):
        v_x =  0 
        v_y = -(-1)**n * ( np.sqrt(n) + (tableau[i]-temps[n])/0.5 )
        vx.append(v_x)
        vy.append(v_y)
    return vx, vy


vitesses_n2 = v_dans_D (temps_n2,2)
vx_n2 = vitesses_n2[0]
vy_n2 = vitesses_n2[1]

vitesses_n4 = v_dans_D (temps_n4,4)
vx_n4 = vitesses_n4[0]
vy_n4 = vitesses_n4[1]
```

4. On trace les graphiques demandés :

```python
plt.plot(temps_n2,vx_n2, label = "Vitesse selon x")
plt.plot(temps_n2,vy_n2, label = "Vitesse selon y")
plt.title("Vitesses de la particule pour pour t variant de $t_2 - 1$ à $t_2$")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(temps_n4,vx_n4, label = "Vitesse selon x")
plt.plot(temps_n4,vy_n4, label = "Vitesse selon y")
plt.title("Vitesses de la particule pour pour t variant de $t_4 - 1$ à $t_4$")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.legend()
plt.grid()
plt.show()
```

5. On détermine les maximum des vitesses

```python
# On définit une fonction pour trouver le temps correspondant au maximum de vitesse

def temps_max(tableau_temps, tableau_vitesse, maximum):
    for i in range (len(tableau_vitesse)):
        if tableau_vitesse[i] == maximum:
            return tableau_temps[i]


# On trouve le max avec la fonction np.max de numpy, puis on utilise notre fonction temps_max pour trouver le temps correspondant

max_vx_n2 = np.max(vx_n2)
tmax_vx_n2 = temps_max(temps_n2, vx_n2, max_vx_n2)
max_vy_n2 = np.max(vy_n2)
tmax_vy_n2 = temps_max(temps_n2, vy_n2, max_vy_n2)


max_vx_n4 = np.max(vx_n4)
tmax_vx_n4 = temps_max(temps_n4, vx_n4, max_vx_n4)
max_vy_n4 = np.max(vy_n4)
tmax_vy_n4 = temps_max(temps_n4, vy_n4, max_vy_n4)


print(f"La valeur maximale vx pour n = 2  est {max_vx_n2:.3} atteinte pour t = {tmax_vx_n2:.3f}")
print(f"La valeur maximale vx pour n = 2  est : {max_vy_n2:.3} atteinte pour t = {tmax_vy_n2:.3f}")
print("\n")
print(f"La valeur maximale vx pour n = 4  est : {max_vx_n4:.3} atteinte pour t = {tmax_vx_n4:.3f}")
print(f"La valeur maximale vy pour n = 4  est : {max_vy_n4:.3} atteinte pour t = {tmax_vy_n4:.3f}")
```

On constate que le maximum de la vitesse en x et y augmente en fonction du temps.
