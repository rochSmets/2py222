## Contrôle : Particule dans un cyclotron

```python
#Importation des modules utiles
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt

tau = 0.5
def t(n):
    """ donne la valeur de l'instant t pour un n donné """
    return n + tau * sqrt(n)

#On crée un tableau qui contient les instants pour n de 0 à 50
instants = np.array([t(n) for n in range(0, 50)])

#On vérifie que les instants connus ont les valeurs attendues
print(instants[1] == 1.5)
print(instants[4] == 5)
print(instants[9] == 10.5)
print(instants[16] == 18)
print(instants[25] == 27.5)
```

La réponse est "True", cela signifie que nos propositions sont vraies. Sinon, l'opérateur booléen affiché aurait été "False".

```python
#On crée des fonctions pour la vitesse dans les D

def mtn1(n):
    return np.linspace(t(n) - 1, t(n), 100)

def v_x1(n):
    return (-1)**(n - 1) * sqrt(n) * np.sin(np.pi * (mtn1(n) - t(n)))

def v_y1(n):
    return - (-1)**(n - 1) * sqrt(n) * np.cos(np.pi * (mtn1(n) - t(n)))
```

```python
def vitesses(n):
    plt.plot(mtn1(n), v_x1(n) / mtn1(n), v_y1(n))
    plt.legend(["v_x", "v_y"])
    plt.xlabel("Temps (sans unité)")
    plt.ylabel("Vitesses")

vitesses(2)
plt.title("Vitesses horizontale et verticale pour n + 1 = 2")
```

```python
vitesses(4)
plt.title("Vitesses horizontale et verticale pour n + 1 = 4")
```

```python
#Maxima des vitesses pour n + 1 = 2
print("Maximum de vx : ", max(v_x1(2)))
print("Maximum de vy : ", max(v_y1(2)))
```
