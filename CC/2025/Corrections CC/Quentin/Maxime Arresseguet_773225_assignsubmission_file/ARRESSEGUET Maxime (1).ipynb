```python
# Importation du module Numpy
import numpy as np
import matplotlib.pyplot as plt


tau = 0.5 # Importation de la contante tau
t = np.zeros((50,1))  #Creation d'un tableau vierge
for i in range (1,50):    #Calcul des 50 valeurs
    t[i] = i + tau * np.sqrt(i)
print (len(t))     #Verification
```

2.

```python
print(t[1] == 1.5 and t[4] == 5 and t[9] == 10.5 and t[16] == 18 and t[25] == 27.5 )  #Operation booléennes pour verification
```

3.

```python
t1  = np.linspace(t[2]-1,t[2],30)     #Création d'un tableau échatillonant les valeurs entre 
t2  = np.linspace(t[4]-1,t[4],30)

def Vx(d,n):     #Creation de la fonction calculant la vitesse selon X
    V = (-1)**n*np.sqrt(n+1)*np.sin(np.pi*(d-d[n+1]))
    return V

def Vy(d,n):   #Creation de la fonction calculant la vitesse selon Y
    V = -(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(d-d[n+1]))
    return V

#print ("Valeurs de Vx entre T2-1 et T2: ",Vx(t,2))
#print ("Valeurs de Vy entre T2-1 et T2: ",Vy(t,2))

#print ("Valeurs de Vx entre T4-1 et T4: ",Vx(t,4))
#print ("Valeurs de Vy entre T4-1 et T4: ",Vy(t,4))
```

4.

```python
fig, modèle = plt.subplots(figsize=(10,6))       #création de la figure
modèle.plot(t1,Vx(t1,2),label = "Vx")
modèle.plot(t1,Vy(t1,2),label = "Vy")#création du plot du modèle
modèle.legend()                                  #Affichage de la légende en haut à gauche
plt.xlabel("temps(s)")                       #Définition et affichage du nomes des axes
plt.ylabel("vitesse(m/s)")
modèle.set_title("Valeurs de Vx et Vy entre T2-1 et T2")  #Définition et affichage du titre
plt.show()                                       #Afin de ne pas afficher d'informations supplémentaires au dessus du graphique
```

```python
fig, modèle = plt.subplots(figsize=(10,6))       #création de la figure
modèle.plot(t2,Vx(t2,4),label = "Vx")
modèle.plot(t2,Vy(t2,4),label = "Vy")#création du plot du modèle
modèle.legend()                                  #Affichage de la légende en haut à gauche
plt.xlabel("temps(s)")                       #Définition et affichage du nomes des axes
plt.ylabel("vitesse(m/s)")
modèle.set_title("Valeurs de Vx et Vy entre T4-1 et T4")  #Définition et affichage du titre
plt.show() 
```

5.

```python
def max(x):
    max = x[0]
    for i in range (len(x)):
        if x[i] > max:
            max = x[i]
    return max

print("Maximum de Vx entre T2-1 et T2",max(Vx(t1,2)))
print("Maximum de Vy entre T2-1 et T2",max(Vy(t1,2)))

print("Maximum de Vx entre T4-1 et T4",max(Vx(t2,4)))
print("Maximum de Vy entre T4-1 et T4",max(Vy(t2,4)))
```

Nous observons qu'entre T2 et T4 la vitesse générale à augmentée, cela est positif dans le contexte d'un accélérateur de particule.
Aussi il semble que la vitesse selon x atteint un valeur plus élevé que celle selon y.
