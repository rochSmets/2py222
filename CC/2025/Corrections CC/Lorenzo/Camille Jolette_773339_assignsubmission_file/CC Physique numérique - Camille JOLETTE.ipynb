### Partie 1

```python
#import des modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Paramètres initial
tau = 0.5

def t(n):
    """Définition de la suite tn"""
    return n + tau*np.sqrt(n)

#Tableau contennant les valeurs de n
n = np.arange(0,51,1)

#Tableau contennant les valeurs de temps
tn = np.zeros(n.shape)

#Boucle calculant les valeurs de temps pour chaque n
for i in n:
    tn[i] = t(i)

    #On vérifie que les valeurs obtenues sont correctes
if tn[1]== float(1.5) and tn[4] == float(5) and tn[9] == float(10.5) and tn[16] == float(18) and tn[25] == float(27.5):
    print("Le tableau obtenu a l'air d'être correct")
else :
    print("Le tableau obtenu a l'air d'être incorrect")
```

Les valeurs obtenues sont bien celles qui étaient attendues.

```python
#Tableau de temps t, pour t compris entre t2-1 et t2 (n=1!)
t2 = np.arange(tn[2]-1,tn[2],0.05)

def vy_dans(t,n):
    return (-(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(t-tn[n+1])))

def vx_dans(t,n):
    return ((-1)**n*np.sqrt(n+1)*np.sin(np.pi*(t-t[n+1])))

#Calcul des valeurs de vy, vx entre t2-1 et t2
vy_dans2 = vy_dans(t2,1)
vx_dans2 = vx_dans(t2,1)

#Tableau de temps t, pour t compris entre t4-1 et t4:
t4 = np.arange(tn[4]-1,tn[4],0.05)

#Calcul des valeurs de vy, vx entre t4-1 et t4 (n=3)
vy_dans4 = vy_dans(t4,3)
vx_dans4 = vx_dans(t4,3)

#Tracé des graphiques
plt.title("Représentation de vx, vy en fonction du temps pour t variant de t2-1 à t2")
plt.xlabel("Temps, pour t allant de t2-1 à t2")
plt.ylabel("Vitesse")
plt.plot(t2, vy_dans2, label = 'vy')
plt.plot(t2, vx_dans2, label = 'vx')
plt.legend()
plt.show()
```

```python
plt.title("Représentation de vx, vy en fonction du temps pour t variant de t4-1 à t4")
plt.xlabel("Temps, pour t allant de t4-1 à t4")
plt.ylabel("Vitesse")
plt.plot(t4, vy_dans4, label = 'vy')
plt.plot(t4, vx_dans4, label = 'vx')
plt.legend()
plt.show()
```

```python
#Détermination des maximas et minimas pour t2-1,t2:
print("Minimum de vy entre t2-1, t2:", np.min(vy_dans2))
print("Maximum de vy entre t2-1, t2:", np.max(vy_dans2))
print("Maximum de vx entre t2-1, t2:", np.max(vx_dans2))

#Détermination des maximas et minimas pour t4-1,t4:
print("Minimum de vy entre t4-1, t4:", np.min(vy_dans4))
print("Maximum de vy entre t4-1, t4:", np.max(vy_dans4))
print("Maximum de vx entre t4-1, t4:", np.max(vx_dans4))

#Récupère l'indice du minimum et du maximum de vy
mask_vy2_min = np.argmin(vy_dans2)
mask_vy2_max = np.argmax(vy_dans2) 
mask_vy4_min = np.argmin(vy_dans4)
mask_vy4_max = np.argmax(vy_dans4)
print("Le temps correspondant au minimum de vy entre t2-1,t2 est:", t2[mask_vy2_min])
print("Le temps correspondant au maximum de vy entre t2-1,t2 est:", t2[mask_vy2_max])
print("Le temps correspondant au minimum de vy entre t4-1,t4 est:", t4[mask_vy4_min])
print("Le temps correspondant au maximum de vy entre t4-1,t4 est:", t4[mask_vy4_max])

#Récupère l'indice du minimum de vx
mask_vx2 = np.argmax(vx_dans2) 
mask_vx4 = np.argmax(vx_dans4) 
print("Le temps correspondant au minimum de vx entre t2-1,t2 est:", t2[mask_vx2])
print("Le temps correspondant au minimum de vx entre t4-1,t4 est:", t4[mask_vx4])
```

Les valeurs maximales et minimales de vy sont dans les deux cas opposées. On remarque que la vitesse maximale selon x augmente avec n pour n=1, n=3 (ce cas ci). La valeur minimale de vy, elle, a diminué, tandis que la valeur maximale, elle a augmenté.

### Partie II

```python
def vy_entre(n,t):
    """Fonction calculant la vitesse selon y pour un n donné entre les D"""
    return (-1)**n*(np.sqrt(n) + (t-tn[n])/tau)

def vx_entre(n,t):
    """Fonction calculant la vitesse selon x pour un n donné entre les D"""
    return np.zeros(t.shape)

def t_entre(n):
    """ Fonction calculant le temps entre tn et tn+1-1"""
    return np.linspace(tn[n],tn[n+1]-1,51)

def t_dans(n):
    """ Fonction calculant le temps entre tn+1-1 et tn+1"""
    return np.linspace(tn[n+1]-1,tn[n+1],51)
    
def vy_dans(n,t):
    """ Fonction calculant la vitesse selon y dans les D"""
    return (-(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(t-tn[n+1])))

def vx_dans(n,t):
    """ Fonction calculant la vitesse selon x dans les D"""
    return ((-1)**n*np.sqrt(n+1)*np.sin(np.pi*(t-tn[n+1])))  

for i in range(0,6):
    #On trace la vitesse entre tn+1-1 et tn+1, puis tn et tn+1 pour chaque n, pour vx et vy
    plt.plot(t_dans(i), vy_dans(i,t_dans(i)), color ='b')
    plt.plot(t_entre(i), vy_entre(i,t_entre(i)), color ='b')
    plt.plot(t_dans(i), vx_dans(i,t_dans(i)), color ='r')
    plt.plot(t_entre(i), vx_entre(i,t_entre(i)), color ='r')
    
plt.title("Vitesse du proton dans le cyclotron en fonction du temps")
plt.xlabel("Temps")
plt.ylabel("Vitesse selon y (en bleu), et x (en rouge)")
plt.show()
```

On remmarque qu'on obtient bien des fonctions continues, mais pas lisses

```python
for i in range(0,6):
    #On trace le module au carré entre tn+1-1 et tn+1, puis entre tn et tn+1-1
    plt.plot(t_dans(i), np.sqrt(vy_dans(i,t_dans(i))**2 + vx_dans(i,t_dans(i))**2), color ='b')
    plt.plot(t_entre(i), np.sqrt(vy_entre(i,t_entre(i))**2 + vx_entre(i,t_entre(i))), color ='b')
    
plt.title("Module de la vitesse du proton dans le cyclotron en fonction du temps")
plt.xlabel("Temps")
plt.ylabel("Module de la vitesse")
plt.show()
```

On a une augmentation du module de la vitesse au cours du temps par palier, c'est à dire à chaque fois qu'on passe entre les D. Un cyclotron sert donc à accélérer les protons

```python
for i in range(0,50):
    #On trace le module au carré entre tn+1-1 et tn+1, puis entre tn et tn+1-1
    plt.plot(tn-i, np.sqrt(vy_dans(i,t_dans(i))**2 + vx_dans(i,t_dans(i))**2), color ='b')
    plt.plot(tn-i, np.sqrt(vy_entre(i,t_entre(i))**2 + vx_entre(i,t_entre(i))), color ='b')
    
plt.title("Module de la vitesse en fonction de $t_n - n$")
plt.xlabel("$t_n - n$")
plt.ylabel("Module de la vitesse")
plt.show()
```

A mesure que le temps total passé par le proton entre les D augmente (qui diminue au cours du temps), le module de la vitesse diminue (c'est également visible au niveau des graphiques précédents).
