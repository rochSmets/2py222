```python
# Definition des fonctions vitesses 
def v_x(t, tableau_temps, n): # l'instant t, le tableau des temps et n la position du temps utlisé dans le tableau 
    ''' Fonction calculant la composante selon x de la vitesse '''
    return ((-1)**(n)) *np.sqrt(n+1) * np.sin(np.pi*(t-tableau_temps[n+1])) 

def v_y(t, tableau_temps, n): # l'instant t, le tableau des temps et n la position du temps utlisé dans le tableau 
    ''' Fonction calculant la composante selon y de la vitesse '''
    return (-(-1)**(n)) *np.sqrt(n+1) * np.cos(np.pi*(t-tableau_temps[n+1])) 

```
