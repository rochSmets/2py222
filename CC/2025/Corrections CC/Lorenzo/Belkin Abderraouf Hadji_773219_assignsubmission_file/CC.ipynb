## Partie I:

```python
import numpy as np 
import matplotlib.pyplot as plt

# On fixe les constantes du problème
tau = 0.5

### Question 1
# On définit la fonction t qui calcule les temps
def t_n(x):
    return x + tau * np.sqrt(x)

# On définit la liste des n de 0 à 50
n = np.linspace(0,50, 51)

# On crée le tableau des t
t = np.zeros(51, dtype="float64")
for i in range(len(n)):
    t[i] = t_n(n[i])
print("tableau t:", t)
```

```python
### Question 2
#On vérifie les valeurs de t
print(t[1] == 1.5)
print(t[4] == 5)
print(t[9] == 10.5)
print(t[16] == 18)
print(t[25] == 27.5)
```

```python
###Question 3

epsi = 1
rep = 20

t2 = t[2]
t_2 = np.linspace(t2-1 , t2, rep)

t4 = t[4]
t_4 = np.linspace(t4-1, t4, rep)

def vx(x: float,y: float ,z: float):
    return (epsi*np.sqrt(x+1)*np.sin(np.pi*(y-z)))

def vy(x,y,z):
    return -epsi*np.sqrt(x+1)*np.cos(np.pi*(y-z))

#Calcul des valeurs de vx et vy pour t2
vx2 = np.zeros(len(t_2), dtype="float64")
vy2 = np.zeros(len(t_2), dtype="float64")

for i in range(len(t_2)):
    vx2[i] = vx(n[i], t_2[i], t2)
print("Liste de vx pour t allant de t_2-1 à t_2:", vx2)

for i in range(len(t_2)):
    vy2[i] = vy(n[i], t_2[i], t2)
print("Liste de vy pour t allant de t_2-1 à t_2:", vy2)



#On fait la même chose pour t_4
vx4 = np.zeros(len(t_4), dtype="float64")
vy4 = np.zeros(len(t_4), dtype="float64")

for i in range(len(t_4)):
    vx4[i] = vx(n[i], t_4[i], t4)
print("Liste de vx pour t allant de t_4-1 à t_4:", vx4)

for i in range(len(t_4)):
    vy4[i] = vy(n[i], t_4[i], t4)
print("Liste de vy pour t allant de t_4-1 à t_4:", vy4)
```

```python
### Question 4
plt.subplot(2,1,1)
plt.plot(t_2, vx2, 'r', label="$v_x$")
plt.xlabel("temps allant de $t_2 -1$ à $t_2$")
plt.ylabel("vitesses selon x et y")
plt.plot(t_2, vy2, color='b', label="$v_y$")
plt.title("Vitesse $v_x$ et $v_y$ en fonction de t allant de $t_2 - 1$ à $t_2$")
plt.legend()

plt.subplot(2,1,2)
plt.plot(t_4, vx4, 'g', label="$v_x$")
plt.xlabel("temps allant de $t_4 -1$ à $t_4$")
plt.ylabel("vitesses selon x et y")
plt.plot(t_4, vy4, color='c', label="$v_y$")
plt.title("Vitesse $v_x$ et $v_y$ en fonction de t allant de $t_4 - 1$ à $t_4$")
plt.legend()

plt.show()
```
