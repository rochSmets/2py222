```python
import numpy as np
import matplotlib.pyplot as plt

def temps(n, tau = 0.5):
    t_n = n + tau*np.sqrt(n)
    return t_n

t_list = []
for i in range(50):
    t_valeurs = temps(i,tau = 0.5)
    t_list.append(t_valeurs)

t_array = np.array(t_list)
print(t_array)
```

Question 3  
Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$
Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

```python
intervalle_t2 = np.linspace(t_array[2]-1, t_array[2],100)
intervalle_t4 = np.linspace(t_array[4]-1, t_array[4],100)

vx_t2 = -np.sqrt(1+1)*np.sin(np.pi*(intervalle_t2 -t_array[2]))
vy_t2 = -np.sqrt(1+1)*np.cos(np.pi*(intervalle_t2 -t_array[2]))

vx_t4 = -np.sqrt(3+1)*np.sin(np.pi*(intervalle_t4 -t_array[4]))
vy_t4 = -np.sqrt(3+1)*np.cos(np.pi*(intervalle_t4 -t_array[4]))
```

Question 4

```python
plt.figure(figsize = (10,10))

plt.subplot(2, 2, 1)
plt.plot(intervalle_t2, vx_t2, 'b')
plt.title('Vitesse vx pour  t variant de  t2−1 à  t2 ')
plt.xlabel("temps [s]")
plt.ylabel("vitesse selon x (m/s)")
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(intervalle_t2, vy_t2, 'r')
plt.title('Vitesse vy pour  t variant de  t2−1 à  t2 ')
plt.xlabel("temps [s]")
plt.ylabel("vitesse selon y (m/s)")
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(intervalle_t4, vx_t4, 'b')
plt.title('Vitesse vx pour  t variant de  t4−1 à t4 ')
plt.xlabel("temps [s]")
plt.ylabel("vitesse selon x (m/s)")
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(intervalle_t4, vy_t4, 'r')
plt.title('Vitesse vy pour  t variant de  t4−1 à  t4 ')
plt.xlabel("temps [s]")
plt.ylabel("vitesse selon y (m/s)")
plt.grid()

plt.tight_layout()  
plt.show()
```
