## Partie I

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Paramètre donné
tau = 0.5
n_max = 50

# Tableau des instants t_n
t_values = []

# Calcul des instants t_n pour n = 0 à 50
for n in range(n_max + 1):
    t_n = n + tau*math.sqrt( n)
    t_values.append(t_n)

# Affichage du tableau des instants
for n, t in enumerate(t_values):
    print(f"t_{n} = {t:.4f}")
```

```python
#Fonction pour calculer t_n
def calculate_t(n, tau=0.5):
    return n + tau*math.sqrt(n)

# Liste des indices à vérifier et des valeurs attendues
indices = [1, 4, 9, 16, 25]
valeurs_attendues = [1.5, 5, 10.5, 18, 27.5]

# Vérification des résultats
for n, valeur_attendue in zip(indices, valeurs_attendues):
    t_n = calculate_t(n)
    test = math.isclose(t_n, valeur_attendue, rel_tol=1e-4)  # tolérance de 1e-4 pour la comparaison
    result = "OK" if test else "faux"
    print(f"Vérification pour t_{n}: Calculé = {t_n:.4f}, Attendu = {valeur_attendue}, Résultat: {result}")
```

Apres avoir calculé les valeurs des vitesses pour les intervalles de temps demandés, nous allons les représenter graphiquement:

```python
plt.plot(t, v_x1, label="Vx", color="red")
plt.plot(t, v_y1, label="Vy", color="blue")
plt.xlabel("Temps en s")
plt.ylabel("Vitesse en m/s")
plt.title ("Vx et Vy en fonction du temps pour t allant de t_2-1 a t_2")
```
