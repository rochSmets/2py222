```python
# On importe les bibliothèques dont on va se servir
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1)

tau=0.5

def t(n): # la fonction nous donne un tableau des temps pour un nombre de valeurs données
    tn=[]
    for i in range(n+1):
        tn.append(i+tau*np.sqrt(i))
    return tn

tn50=t(50) # tabelau de 50 temps
```

```python
# 2)
# on vérifie que chaque valeur demandée correspond aux valeurs de notre tableau
print(tn50[1]==1.5) 
print(tn50[4]==5)
print(tn50[9]==10.5)
print(tn50[16]==18)
print(tn50[25]==27.5)
```

```python
# 3)

temps2=np.linspace(tn50[2]-1,tn50[2],100) # tableau des temps entre t2-1 et t2
temps4=np.linspace(tn50[4]-1,tn50[4],100) # tableau des temps entre t4-1 et t4

#on définie les fonction des vitesses en fonction d'une liste de temps et de n

def Vx2(t,n): 
    Tvx2=[]
    for i in range(len(t)):
       Tvx2.append(((-1)**n)*np.sqrt(n+1)*np.sin(np.pi*(t[i]-t[len(t)-1])))
    return Tvx2
    
    
def Vy2(t,n):
    Tvy2=[]
    for i in range(len(t)):
        Tvy2.append((-(-1)**n)*np.sqrt(n+1)*np.cos(np.pi*(t[i]-t[len(t)-1])))
    return Tvy2


tvx2=Vx2(temps2,1)
tvy2=Vy2(temps2,1)
tvx4=Vx2(temps4,3)
tvy4=Vy2(temps4,3)
```

```python
# 4)

# on trace les courbes pour les temps entre t2-1 et t2
plt.plot(temps2,tvx2,label='Vx')
plt.xlabel('temps')
plt.legend()
plt.plot(temps2,tvy2,label='Vy')
plt.xlabel('temps')
plt.ylabel('Viteses')
plt.legend()
plt.title('Vx et Vy en fonction du temps entre t(2-1) et t2')
```

```python
# on trace les courbes pour les temps entre t4-1 et t4
plt.plot(temps4,tvx4,label='Vx')
plt.xlabel('temps')
plt.legend()
plt.plot(temps4,tvy4,label='Vy')
plt.xlabel('temps')
plt.ylabel('Viteses')
plt.legend()
plt.title('Vx et Vy en fonction du temps entre t(4-1) et t4')
```

```python
# 5)

#Valeurs max des vitesses pour l'intervalle de temps entre t(2-1) et t2'
print(max(tvx2))
print(max(tvy2))

#Valeurs max des vitesses pour l'intervalle de temps entre t(4-1) et t4'
print(max(tvx4)) 
print(max(tvy4))
```
