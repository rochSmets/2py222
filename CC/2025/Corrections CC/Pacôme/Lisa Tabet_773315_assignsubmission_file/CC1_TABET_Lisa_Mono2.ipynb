# <font color="red"> 1. Introduction: </font> 

Dans ce contrôle on s'interesse au mouvement d'une particule (un proton) dans un cyclotron.  On injecte un proton à **$t_0=0$**  au niveau du D infénieur sans vitesse initiale.

La tension appliquée entre les D crée un **Champ électrique** dans le même sens et direction que la vitesse du proton. 

Les instants $t_n$ où le proton sort des D sont donnés par: 

### $$t_n=n+\tau \sqrt{n}.$$

On choisit **$\tau=0.5$** . Le temps passé par le proton dans un D est toujours de $1$. 


# <font color="orange"> Partie I:  </font>  

#### <font color="magenta">  1)- Créer un tableau des instants de sortie des boucles:   </font>  


```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres donnés dans l'énoncé: 
tau = 0.5
t0 = 0
n_max = 50
epsilon = lambda n: (-1)**n  # epsilon_n = (-1)^n


# Création du tableau des instants de sortie des boucles "D"
tn = np.array([n + np.sqrt(tau * n) for n in range(51)])
```

#### <font color="magenta">  2)-Vérification des valeurs de t_n à différents instants:   </font>  

#### <font color="magenta">  3)- Calcul des vitesses vx et vy en fonction du temps:  </font>  

```python
import matplotlib.pyplot as plt

# Fonction pour calculer vx et vy (comme donné dans l'énoncé) 
def calculate_vx_vy(tn, t, tau, n):
    epsilon_n = (-1) ** n
    if tn <= t < tn + 1 - 1:
        vx = 0
        vy = epsilon_n * (np.sqrt(n) + (t - tn) / tau)
    else:
        vx = epsilon_n * np.sqrt(n + 1) * np.sin(np.pi * (t - (tn + 1)))
        vy = -epsilon_n * np.sqrt(n + 1) * np.cos(np.pi * (t - (tn + 1)))
    return vx, vy

# Calculer les valeurs pour t variant de t2-1 à t2
t2 = tn[2]
t4 = tn[4]

t2_values = np.linspace(t2 - 1, t2, 1000)
vx_t2 = []
vy_t2 = []

for t_val in t2_values:
    vx, vy = calculate_vx_vy(t2, t_val, tau, 2)
    vx_t2.append(vx)
    vy_t2.append(vy)

# Calculer les valeurs pour t variant de t4-1 à t4
t4_values = np.linspace(t4 - 1, t4, 1000)
vx_t4 = []
vy_t4 = []

for t_val in t4_values:
    vx, vy = calculate_vx_vy(t4, t_val, tau, 4)
    vx_t4.append(vx)
    vy_t4.append(vy)
```

#### <font color="magenta">  4)- Représentation graphique de vx et vy:  </font>  

```python
# Graphique pour vx et vy en fonction de t pour t2
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(t2_values, vx_t2, label='$v_x$')
plt.plot(t2_values, vy_t2, label='$v_y$')
plt.xlabel('Temps $t$')
plt.ylabel('Vitesse')
plt.legend()
plt.title('Vitesse $v_x$ et $v_y$ pour $t$ variant de $t_2-1$ à $t_2$')

# Graphique pour vx et vy en fonction de t pour t4
plt.subplot(1, 2, 2)
plt.plot(t4_values, vx_t4, label='$v_x$')
plt.plot(t4_values, vy_t4, label='$v_y$')
plt.xlabel('Temps $t$')
plt.ylabel('Vitesse')
plt.legend()
plt.title('Vitesse $v_x$ et $v_y$ pour $t$ variant de $t_4-1$ à $t_4$')
plt.tight_layout()
plt.show()
```

#### <font color="magenta">  5)-Trouver les maxima de vx et vy pour t variant de t2-1 à t2, et t4-1 à t4: </font>

```python
# Maxima pour t variant de t2-1 à t2
max_vx_t2 = max(vx_t2)
max_vy_t2 = max(vy_t2)
t_max_vx_t2 = t2_values[vx_t2.index(max_vx_t2)]
t_max_vy_t2 = t2_values[vy_t2.index(max_vy_t2)]

print(f"Maxima pour t2-1 à t2: vx_max={max_vx_t2} à t={t_max_vx_t2}, vy_max={max_vy_t2} à t={t_max_vy_t2}")

# Maxima pour t variant de t4-1 à t4
max_vx_t4 = max(vx_t4)
max_vy_t4 = max(vy_t4)
t_max_vx_t4 = t4_values[vx_t4.index(max_vx_t4)]
t_max_vy_t4 = t4_values[vy_t4.index(max_vy_t4)]

print(f"Maxima pour t4-1 à t4: vx_max={max_vx_t4} à t={t_max_vx_t4}, vy_max={max_vy_t4} à t={t_max_vy_t4}")
```

### Commentaire sur les résultats obtenus: 

Les maxima des composantes de vitesse $v_x$ et $v_y$ pour les intervalles de temps $t_2 - 1$ à $t_2$ et $t_4 - 1$ à $t_4$ montrent que la vitesse du proton varie de manière significative pendant sa trajectoire à l'intérieur du cyclotron. Plus précisément, les valeurs maximales de $v_x$ et $v_y$ sont atteintes lorsque le proton se trouve entre les 'D' et sont influencées par le changement de signe de $\epsilon_n$.
