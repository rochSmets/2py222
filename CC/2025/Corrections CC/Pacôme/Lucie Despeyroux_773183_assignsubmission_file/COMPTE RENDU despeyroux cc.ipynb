 on commence par créer le tableau des tn qui sont définis par: 
 $$t_n=n+\tau \sqrt{n}$$

```python
# on créer le tableau contenant les différents temps
import numpy as np
# valeurs utiles pour la boucle
tau=0.5
n=51 # 51 pour inclure t50
temps=[] #initialisation de la liste
for i in range (n):
    t=i+tau*np.sqrt(i)
    temps.append(t)
temps=np.array(temps)
print(temps)
print(len(temps))
```

On observe que le t0 vaut bien 0 et que nous avons bien 51 temps différents de t0 à t50 inclus. Vérifions à présent les valeurs de t1, t4, t9, t16 et t25:

```python
# sélectionner des élements dans le tableau numpy temps avec leurs indices. l'indice 0 correspondant au temps 0, il n'y a pas de décalage dans les indices. 
print('le temps 1 vaut:',temps[1])
print('le temps 4 vaut:',temps[4])
print('le temps 9 vaut:',temps[9])
print('le temps 16 vaut:',temps[16])
print('le temps 25 vaut:',temps[25])
```

On remarque que les valeurs obtenues par sélection d'élements précis dans le tableau numpy temps coincident bien avec les valeurs mentionnée dans l'énoncé.\
Pour la question suivante, nous sommes dans la configuration ou :
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

```python
# pour des temps allant de t2-1 à t2
# création du nouveau tableau temps:
temps2=np.linspace(temps[2]-1,temps[2],100) #je choisis un tableau de 100 valeurs de temps allant de t(2)-1 à t(2)
# on a n+1=2 soit n=1 donc n est impair, on choisit donc la valeur de epsilon en conséquence:
eps=-1

a=temps2-temps[2] # n=1 donc soustraire t(n+1) revient à soustraire t2
#définition de vx:
v_x=eps*np.sqrt(1+1)*np.sin(np.pi*a) # v_x est un tableau contenant 100 valeurs de v_x
print(len(v_x))

v_y=-eps*np.sqrt(1+1)*np.cos(np.pi*a) # v_y est un tableau contenant 100 valeurs de v_y
print(len(v_y))

# pour des temps allant de t4-1 à t4
# création du nouveau tableau temps:
temps4=np.linspace(temps[4]-1,temps[4],100) #je choisis un tableau de 100 valeurs de temps allant de t(4)-1 à t(4)
# on a n+1=4 soit n=3 donc n est impair, on choisit donc la valeur de epsilon en conséquence:
eps=-1

b=temps4-temps[4] # n=1 donc soustraire t(n+1) revient à soustraire t2
#définition de vx:
v_xprime=eps*np.sqrt(3+1)*np.sin(np.pi*b) # v_x est un tableau contenant 100 valeurs de v_xprime
print(len(v_xprime))

v_yprime=-eps*np.sqrt(3+1)*np.cos(np.pi*b) # v_y est un tableau contenant 100 valeurs de v_yprime
print(len(v_yprime))
```

Nous obtenons donc des tableaux de 100 valeurs pour le calcul de vx et vy pour les deux tranches de temps demandés. Nous nous interresserons maintenant à la représentation graphique de nos résultats:

```python
import matplotlib.pyplot as plt
# représentation graphique pour le tableau temps2
plt.plot(temps2,v_x, label='v_x')
plt.plot(temps2, v_y, label='v_y', color='r')
plt.xlabel('temps allant de t(2)-1 à t(2)')
plt.ylabel('vitesses v_x et v_y')
plt.title('vitesses selon les axes x et y entre des temps allant de t(2)-1 à t(2)')
plt.legend()
plt.grid()
plt.show()

#représentation graphique pour le tableau temps4
plt.plot(temps4,v_xprime, label='v_x')
plt.plot(temps4, v_yprime, label='v_y', color='r')
plt.xlabel('temps allant de t(4)-1 à t(4)')
plt.ylabel('vitesses v_x et v_y')
plt.title('vitesses selon les axes x et y entre des temps allant de t(4)-1 à t(4)')
plt.legend()
plt.grid()
plt.show()
```

Une fois les représentations graphiques obtenues, nous nous interessons aux maximums de ces vitesses sur les différents intervalles de temps étudiés. 

```python
max2x=max(v_x)
print('le maximum de v_x pour t allant de t(2)-1 à t(2) est :',max2x)

max2y=max(v_y)
print('le maximum de v_y pour t allant de t(2)-1 à t(2) est :',max2y)

max4x=max(v_xprime)
print('le maximum de v_x pour t allant de t(4)-1 à t(4) est :',max4x)

max4y=max(v_yprime)
print('le maximum de v_y pour t allant de t(4)-1 à t(4) est :',max4y)
```

On remarque que pour la vitesse selon y, la vitesse maximale pour les deux intervalles de temps est obtenue à l'extrémité soit à t2 et t4 alors que le maximum de la vitesse en x est obtenue pour un temps situé au milieu de l'intervalle. On remarque également que pour un même intervalle de temps, la vitesse maximale obtenue selon les deux axes est identique à l'arrondit près.

On cherche cette fois ci une représentation graphique de vx et vy pour des temps allant de 0 soit t0 à t6. Il faut faire attention aux valeurs de vx et vy qui ne sont pas les mêmes lorsque le proton se situe dans les D inférieur et supérieur ou lorsqu'il entre dans les D.

```python
#définition des fonctions de vx et vy en fonction de la trajectoire du proton
def sup_inf(a,n):
    v_x=(-1)**n*np.sqrt(n+1)*np.sin(np.pi*a)
    v_y=-(-1)**n*np.sqrt(n+1)*np.cos(np.pi*a)
    return v_x , v_y
def D(ta, tb, n):
    v_x=0
    v_y=(-1)**n*(np.sqrt(n)+(ta-tb)/0.5)
    return v_x , v_y
# calcul pour chaque section dans D
#sélection des temps qui nous interresse
t1=np.linspace(temps[0], temps[1]-1, 10)
vx1=[]
for i in t1:
    x=D(t1,temps[0],0)
    vx1.append(x)
vx1=np.array(vx1)
print(vx1)
t2=np.linspace(temps[1], temps[2]-1, 10)
vx2=[]
for i in t2:
    x=D(t2,temps[1],1)
    vx2.append(x)
vx2=np.array(vx2)

t3=np.linspace(temps[2], temps[3]-1, 10)
vx3=[]
for i in t3:
    x=D(t3,temps[2],2)
    vx3.append(x)
vx3=np.array(vx3)

t=np.linspace(temps[3], temps[4]-1, 10)
vx4=[]
for i in t:
    x=D(t,temps[3],3)
    vx4.append(x)
vx4=np.array(vx4)

t4=np.linspace(temps[4], temps[5]-1, 10)
vx5=[]
for i in t4:
    x=D(t4,temps[4],4)
    vx5.append(x)
vx5=np.array(vx5)

# calcul pour les fois où la trajectoire est dans la partie supérieure ou inférieure 
t5=np.linspace(temps[1]-1,temps[1],10)
v1=[]
for i in t5: 
    a=t5-temps[1]
    x=sup_inf(a,0)
    v1.append(x)
v1=np.array(v1)

t6=np.linspace(temps[2]-1,temps[2],10)
v2=[]
for i in t6: 
    a=t6-temps[2]
    x=sup_inf(a,1)
    v2.append(x)
v2=np.array(v2)

t7=np.linspace(temps[3]-1,temps[3],10)
v3=[]
for i in t7: 
    a=t7-temps[3]
    x=sup_inf(a,2)
    v3.append(x)
v3=np.array(v3)

t8=np.linspace(temps[4]-1,temps[4],10)
v4=[]
for i in t8: 
    a=t8-temps[4]
    x=sup_inf(a,3)
    v4.append(x)
v4=np.array(v4)

t9=np.linspace(temps[5]-1,temps[5],10)
v5=[]
for i in t9: 
    a=t9-temps[5]
    x=sup_inf(a,4)
    v5.append(x)
v5=np.array(v5)

t10=np.linspace(temps[6]-1,temps[6],10)
v6=[]
for i in t10: 
    a=t10-temps[6]
    x=sup_inf(a,0)
    v6.append(x)
v6=np.array(v6)


plt.plot(vx1[:,0],t1)
plt.plot(vx2[:,0],t2)
plt.plot(vx3[:,0],t3)
plt.plot(vx4[:,0],t4)
plt.plot(vx5[:,0],t5)

plt.plot(v1[:,0],t6)
plt.plot(v2[:,0],t7)
plt.plot(v3[:,0],t8)
plt.plot(v4[:,0],t9)
plt.plot(v5[:,0],t10)
plt.show()
plt.plot(vx1[:,1],t1)
plt.plot(vx2[:,1],t2)
plt.plot(vx3[:,1],t3)
plt.plot(vx4[:,1],t4)
plt.plot(vx5[:,1],t5)
plt.show()
```
