```python
# on importe les modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## partie 1:

```python
'''exercice 1'''
tau = 0.5
t = np.zeros(51) # tableau des temps 

for i in range (0,51):
    t[i] = i + tau*np.sqrt(i)

'''exercice 2'''    
# on vérifie que les valeurs calculées sont bonnes    
print(t[1]==1.5) 
print(t[4]==5.0)
print(t[9]==10.5)
print(t[16]==18.0)
print(t[25]==27.5)
```

```python
'''exercice 3'''
t2 = np.linspace(t[2]-1,t[2],10)  # initialisation du tableau des temps pour t allant de t2 - 1 à t2
t4 = np.linspace(t[4]-1,t[4],10)  # initialisation du tableau des temps pour t allant de t4 - 1 à t4
vx2 = np.zeros(t2.shape) # tableau des vitesses pour le premier intervalle de temps
vy2 = np.zeros(t2.shape) 
vx4 = np.zeros(t4.shape)  # pareil pour le deuxième
vy4 = np.zeros(t4.shape)

for i in range(0, len(t2)):  # on calcule les vitesses
    vx2[i] = -1*np.sqrt(2)*np.sin(np.pi*(i- t2[i]))
    vy2[i] = 1*np.sqrt(2)*np.cos(np.pi*(i- t2[i]))  

    

for i in range(0, len(t4)):
    vx4[i] = -1*np.sqrt(4)*np.sin(np.pi*(i- t4[i]))
    vy4[i] = 1*np.sqrt(4)*np.cos(np.pi*(i- t4[i])) 

        
```

```python
'''exercice 4'''
# on trace le graphe pour t2
plt.plot(t2,vx2, color = 'r', label = 'vx')
plt.plot(t2,vy2, color = 'g', label = 'vy')
plt.xlabel('valeur de t entre t2-1 et t2')
plt.ylabel('valeur de la vitesse')
plt.title("vitesses selon x et y entre t2 -1 et t2")
plt.legend()
plt.show()
```

Entre les instants t2 - 1 et t2, la valeur maximale de vx semble etre aux alentours de 1.3, atteinte en t = 2.6.

Pour vy,  la valeur maximale semble etre aux alentours de 1.4, atteinte en t = 1.9.

```python
# on fait pareil pour t4
plt.plot(t4,vx4, color = 'r', label = 'vx')
plt.plot(t4,vy4, color = 'g', label = 'vy')
plt.xlabel('valeur de t entre t4-1 et t4')
plt.ylabel('valeur de la vitesse')
plt.title("vitesses selon x et y entre t4 -1 et t4")
plt.legend()
plt.show()
```

Entre les instants t4 - 1 et t4, la valeur maximale de vx semble etre aux alentours de 2.0, atteinte en t = 4.5 .

Pour vy, la valeur maximale semble etre aux alentours de 2.0, atteinte en t = t4-1 et en t4.

### Partie 2

```python
'''exercice 6 '''
t6 = np.zeros(7) # initialisation du tableau des temps entre 0 et t6
for i in range(0,7):
    t6[i] = t[i]

vx = np.zeros(t6.shape) 
vy = np.zeros(t6.shape)

for i in range(0, 7): # calcul des vitesses associées
    N = i+1
    vx[i] = ((-1)**i)*np.sqrt(N)*np.sin(np.pi*(i- t6[i]))
    vy[i] = -((-1)**i)*np.sqrt(N)*np.cos(np.pi*(i- t6[i]))  
```

```python
# on trace les courbes

plt.plot(t6,vx, color = 'r', label = 'vx')
plt.plot(t6,vy, color = 'g', label = 'vy')
plt.xlabel('valeur de t entre 0 et t6')
plt.ylabel('valeur de la vitesse')
plt.title("vitesses selon x et y entre t0 et t6")
plt.legend()
plt.show()
```

Nous avons tracé le graphe représentant les vitesses selon x et y du proton dans le cyclotron. On remarque que les courbes sont continues, ce que signifie qu'à tout instant, on peut déterminer la position du proton.

```python
''' exercice 7'''
v = np.zeros(t6.shape) # tableau du module de la vitesse

for i in range(0, 7):
    v[i] = np.sqrt(vx[i]**2+vy[i]**2)
```

```python
plt.plot(t6, v)
plt.xlabel('temps de t0 à t6')
plt.ylabel('module de la vitesse')
plt.title('évolution du module de la vitesse au cours du temps')
plt.show()
```

Nous pouvons voir sur ce graphe une relation de proportionnalité entre la vitesse et le temps traduite par une droite (ou du moins approximativement), passant par l'origine du repère.

Le cyclotron sert donc à accélérer les protons qui circulent en son sein. En injectant du courant dans le cyclotron, on va accélérer le proton. Il va créer un champ magnétique qui s'intensifie au fur et à mesure que sa vitesse augmente.

```python
'''exercice8'''
# on répète ce que l'on a fait précédemment mais cette fois-ci pour toute la durée de l'expérience
vx0 = np.zeros(t.shape)
vy0 = np.zeros(t.shape)

for i in range(0,51): 
    N = i+1
    vx0[i] = ((-1)**i)*np.sqrt(N)*np.sin(np.pi*(i- t[i]))
    vy0[i] = -((-1)**i)*np.sqrt(N)*np.cos(np.pi*(i- t[i]))
```

Sur le graphe ci-dessus, nous voyons que le proton accélère, cependant, nous l'avons déjà observé sur la figure précédente. c qui nous intéresse ici est le fait que le proton atteint sa vitesse maximale (égale à 7) au bout de t =3.5.
