# Particule dans un cyclotron

Au temps $t_0=0$ **(ici le temps est sans unité)**, un proton est injecté au niveau du D inférieur sans vitesse. Une tension alternative est appliquée entre les D. Elle crée un champ électrique dont la direction et le sens sont toujours **similaires** à ceux de la vitesse du proton lorsqu’il est entre les D. Les instants $t_n$ où le proton sort des D (soit inférieur, soit supérieur, voir la figure) sont donnés, **sans unité**, par
$$t_n=n+\tau \sqrt{n}.$$
On choisit $\tau=0.5$. Le temps passé par le proton dans un D (supérieur ou inférieur) est toujours de $1$.
La trajectoire du proton est plane. On s’intéresse à sa vitesse. Ses composantes sont données, dans un repère et des unités appropriés, **en fonction de sa position dans le cyclotron:**
1) Pour $t$ variant de $t_n$ à $t_{n+1}-1$ c’est-à-dire quand il est entre les D :
$$
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$$
2) Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

## Partie I

### Tableau contenant les valeurs de $t_0$ à $t_5$

```python
#Constantes
import numpy as np
tau = 0.5
t_n = [0]

for i in range(1,50,1):
    t_n.append(i+tau*np.sqrt(i))

#On vérifie qu'on obtient bien "True"
print(t_n[1] == 1.5)
print(t_n[4] == 5)
print(t_n[9] == 10.5)
print(t_n[16] == 18)
print(t_n[25] == 27.5)
```

### Calcul des valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$

```python
import matplotlib.pyplot as plt
v_x2 = []
v_y2 = []

t2 = np.linspace(t_n[2]-1, t_n[2], 100)

for i, t in enumerate(t2):
    v_x2.append( ((-1)**1)*np.sqrt(2)*np.sin(np.pi*(t-t_n[2])) ) #n=1 car on prend n+1 = 2
    v_y2.append( (-(-1)**1)*np.sqrt(2)*np.cos(np.pi*(t-t_n[2])) )

plt.plot(t2, v_x2, label="$v_x$")
plt.plot(t2, v_y2, label="$v_y$")
plt.xlabel("$t$ variant de $t_2 - 1$ à $t_2$")
plt.ylabel("Vitesse")
plt.title("Vitesse selon $t_2 - 1$ à $t_2$")
plt.legend()
plt.grid()
plt.show()
```

### Calcul des valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_4−1$ à $t_4$

```python
v_x4 = []
v_y4 = []
t4 = np.linspace(t_n[4]-1, t_n[4], 100)

for i, t in enumerate(t4):
    v_x4.append( ((-1)**(3))*np.sqrt(4)*np.sin(np.pi*(t-t_n[4])) ) #n=3 car on prend n+1 = 4
    v_y4.append( (-(-1)**(3))*np.sqrt(4)*np.cos(np.pi*(t-t_n[4])) )

plt.plot(t4, v_x4, label="$v_x$")
plt.plot(t4, v_y4, label="$v_y$")
plt.xlabel("$t$ variant de $t_4 - 1$ à $t_4$")
plt.ylabel("Vitesse")
plt.title("Vitesse selon $t_4 - 1$ à $t_4$")
plt.legend()
plt.grid()
plt.show()
```

### Détermination des vitesses maximales selon chaque composante

```python
#Détermination du max entre t2-1 et t2
v_x2_max = np.max(v_x2)
v_y2_max = np.max(v_y2)
for i, t in enumerate(t2):
    if v_y2[i] == v_y2_max:
        t2_ymax = t
    if v_x2[i] == v_x2_max:
        t2_xmax = t
        
print(f"Entre t2-1 et t2, la vitesse maximum selon x est {v_x2_max:.2e} obtenue au temps {t2_xmax:.2f}. Selon y, c'est {v_y2_max:.2e} au temps {t2_ymax:.2f}")

#Détermination du max entre t4-1 et t4
v_x4_max = np.max(v_x4)
v_y4_max = np.max(v_y4)
for i, t in enumerate(t4):
    if v_y4[i] == v_y4_max:
        t4_ymax = t
    if v_x4[i] == v_x4_max:
        t4_xmax = t
        
print(f"Entre t4-1 et t4, la vitesse maximum selon x est {v_x4_max:.2e} obtenue au temps {t4_xmax:.2f}. Selon y, c'est {v_y4_max:.2e} au temps {t4_ymax:.2f}")
```

Par lecture graphique, on constate qu'à une boucle plus grande, la vitesse maximale augmente, ce qui est cohérent par rapport au principe du cyclotron. 

## Partie II
### Représentation de $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$.

```python
v_x6 = []
v_y6 = []
t_global = []

#Il alterne trajectoire droite et D

for i in range (0,6,1):
    #Il fait une droite
    ti = np.linspace(t_n[i], t_n[i+1]-1, 100)
    for k, t in enumerate(ti):
        v_x6.append(0)
        v_y6.append( ((-1)**i)*(np.sqrt(i)+(t-t_n[i])/tau ) )
                
    #Puis la courbe D
    ti2 = np.linspace(t_n[i+1]-1, t_n[i], 100)
    for k, t in enumerate(ti2):
        v_x6.append( ((-1)**i)*np.sqrt(i+1)*np.sin(np.pi*(t-t_n[i+1])) ) #n=1 car on prend n+1 = 2
        v_y6.append( (-(-1)**i)*np.sqrt(i+1)*np.cos(np.pi*(t-t_n[i+1])) )
    
    t_global = np.concatenate((t_global, ti, ti2))
                
plt.plot(t_global, v_x6, label="$v_x$")
plt.plot(t_global, v_y6, label="$v_y$")
plt.xlabel("$t$ variant de $0$ à $t_6$")
plt.ylabel("Vitesse")
plt.title("Vitesse selon le temps")
plt.legend()
plt.grid()
plt.show()
```

Bien que le graphique soit faux à (sûrement par une erreur de formule), on constate que la vitesse maximale (en valeure absolue) à chaque boucle ne fait que croître. Cela confirme notre première instuition obtenue avec seulement t2 et t4.

### Module de la vitesse $\sqrt{v_x^2+v_y^2}$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$

```python
#Module
module = []
for i, t in enumerate(t_global):
    module.append(np.sqrt(v_x6[i]**2 + v_y6[i]**2))
    
plt.plot(t_global, module, label="$Module$")
plt.xlabel("$t$ variant de $0$ à $t_6$")
plt.ylabel("Module de la vitesse")
plt.title("Module de la vitesse selon le temps")
plt.legend()
plt.grid()
plt.show()
```

En faisant le module de la vitesse (bien que l'erreur précédente se répercute ici aussi), on constate effectivement que la vitesse augmente à chaque tour de cyclotron. Cela est cohérent avec son but: accélérer des particules à très grandes vitesses. Pour cela, on leur fait réaliser plusieurs tours grâce à des différences de tensions qui les fait accélérer.

### Module de la vitesse $\sqrt{v_x(t_n)^2+v_y(t_n)^2}$ en fonction de $t_n-n$

```python
temps_passe = []

for i, t in enumerate(t_n):
    temps_passe.append(t_n[i]-i)

v_x_fin = []
v_y_fin = []
modulefin = []

for i in range (0,49,1):
    v_x_fin.append( ((-1)**i)*np.sqrt(i+1)*np.sin(np.pi*(t_n[i]-t_n[i+1])) ) #n=1 car on prend n+1 = 2
    v_y_fin.append( (-(-1)**i)*np.sqrt(i+1)*np.cos(np.pi*(t_n[i]-t_n[i+1])) )

for i in range (len(v_x_fin)): #On ne garde pas la dernière valeur de temps passé pour éviter d'être out of range
    modulefin.append(np.sqrt(v_x_fin[i]**2 + v_y_fin[i]**2))
    
plt.plot(temps_passe[:49], modulefin, label="Module") #On ne garde pas la dernière valeur de temps passé pour éviter d'être out of range
plt.xlabel("$t$ variant de $0$ à $t_{50}$")
plt.ylabel("Module de la vitesse")
plt.title("Module de la vitesse à $t_n$")
plt.legend()
plt.grid()
plt.show()
```

On constate bien que la vitesse augmente.
