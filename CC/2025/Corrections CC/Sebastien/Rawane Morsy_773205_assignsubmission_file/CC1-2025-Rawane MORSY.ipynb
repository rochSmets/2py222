# Controle Continu - Particule dans un cyclotron : 

## Partie A:

1- On crée le tableau contenant les instants de sortie de boucle D, $$t_n=n+\tau \sqrt{n}.$$

```python
# Bibliothéque à importer
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Paramétres du probléme
tau = 0.5

tn = np.ones((50))        # Tableau conetnant les instants tn 
tn[0] = 0 

# Boucle
for i in range(1,49):
    tn[i] = i + tau * np.sqrt(i)
```

Les valeurs des tn sont donc vérifiées.

-----------------------------
3- Pour trouver es valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$.

On va commencer par créer un nouveau tableau de temps telle que t varie de t_2 - 1 à t_2, on crée donc un nouveau tableau de temps qui respecte ces paramétres.

On crée aprés le tableau d'épsilon qui varie entre 1 et -1.

Enfin, on definiera les des deux tableaux $v_x$ et $v_y$ telle que $$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

```python
# Tableau du temps Pour n = 2 

n = 2      # t_2
e = 10000   # échantillonage suffisant pour le temps

temps = np.linspace(tn[n]-1,tn[n],e)       


# Tableau de variation d'epsilon 
epsilon = np.ones (len(tn))

for i in range(len(tn)):
    epsilon [i] = (-1)**i          # On utilisera de ce tableau la valeur e_1 seulement parceque n+1 = 2


# Tableau vx(t):
vx_2 = np.ones(len(temps))

# Boucle : 
for i in range(0,len(temps)-1):
    vx_2[i] = epsilon[n-1] * np.sqrt (n-1+1) * np.sin(np.pi*(temps[i] - tn[n]))
    

# Tableau vy(t):
vy_2 = np.ones(len(temps))

# Boucle : 
for i in range(0,len(temps)-1):
    vy_2[i] = -epsilon[n-1] * np.sqrt (n-1+1) * np.cos(np.pi*(temps[i] - tn[n]))
```

```python
# Tableau du temps Pour n = 4 

s = 4      # t_4
e = 10000   # échantillonage suffisant pour le temps

temps2 = np.linspace(tn[s]-1,tn[s],e)       


# Tableau de variation d'epsilon 
epsilon = np.ones (len(tn))

for i in range(len(tn)):
    epsilon [i] = (-1)**i          # On utilisera de ce tableau la valeur e_1 seulement parceque n+1 = 2


# Tableau vx(t):
vx_4 = np.ones(len(temps2))

# Boucle : 
for i in range(0,len(temps2)-1):
    vx_4[i] = epsilon[s-1] * np.sqrt (s-1+1) * np.sin(np.pi*(temps2[i] - tn[s]))
    

# Tableau vy(t):
vy_4 = np.ones(len(temps2))

# Boucle : 
for i in range(0,len(temps2)-1):
    vy_4[i] = -epsilon[s-1] * np.sqrt (s-1+1) * np.cos(np.pi*(temps2[i] - tn[s]))
    
```

-----------------
4- Representation Graphique : 

```python
# Representation graphique n = 2
plt.plot(temps, vx_2)                                                                          # Creation de la courbe vx
plt.plot(temps, vy_2)                                                                          # Creation de la courbe vy
plt.title('$v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$')           # Titre du Graphique 
plt.ylabel('$v_x$ et $v_y$')                                                                   # Titre des x
plt.xlabel('$t$ variant de $t_2−1$ à $t_2$')                                                    # Titre des y
plt.grid()
plt.show()
```

```python
# Representation graphique n = 4
plt.plot(temps2, vx_4)                                                                          # Creation de la courbe vx
plt.plot(temps2, vy_4)                                                                          # Creation de la courbe vy
plt.title('$v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_4−1$ à $t_4$')           # Titre du Graphique 
plt.ylabel('$v_x$ et $v_y$')                                                                   # Titre des x
plt.xlabel('t$ variant de $t_4−1$ à $t_4$')                                                    # Titre des y
plt.grid()
plt.show()
```

## Partie B:

```python
# Tableau du temps Pour n = 6 

K = 6      # t_2
e = 1000   # échantillonage suffisant pour le temps

temps3 = np.linspace(0,tn[K],e)       


# Tableau de variation d'epsilon 
epsilon = np.ones (len(tn))

for i in range(len(tn)):
    epsilon [i] = (-1)**i          


# Tableau vx(t):
vx_6 = np.ones(len(temps3))

# Boucle : 
for i in range(0,len(temps3)-1):
    vx_6[i] = epsilon[K-1] * np.sqrt (K-1+1) * np.sin(np.pi*(temps[i] - tn[K]))
    

# Tableau vy(t):
vy_6 = np.ones(len(temps3))

# Boucle : 
for i in range(0,len(temps3)-1):
    vy_6[i] = -epsilon[K-1] * np.sqrt (K-1+1) * np.cos(np.pi*(temps3[i] - tn[K]))
    
# Representation graphique n = 6
plt.plot(temps3, vx_6, '-')                                                                          # Creation de la courbe vx
plt.plot(temps3, vy_6, '-')                                                                          # Creation de la courbe vy
plt.title('$v_x$ et $v_y$ en fonction du temps pour $t$ variant de 0 à $t_6$')                 # Titre du Graphique 
plt.ylabel('$v_x$ et $v_y$')                                                                   # Titre des y
plt.xlabel('$t$ variant de 0 à $t_6$')                                                   # Titre des x
plt.grid()
plt.show()
```
