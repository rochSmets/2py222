# Particule dans un cyclotron 

## Partie 1 

```python
#importation des modules Python utiles 

import numpy as np
import matplotlib.pyplot as plt

#constante données dans l'énoncé
 
taux = 0.5
epsilon = 1 #n pair
epsilon1 = -1 #n impair 

#création d'un tableau contenant les instants de sorties de boucle 

n=50

t = np.empty(n+1)

for i in range (0, n+1):
    t[i] = i + taux*np.sqrt(i)
    
print(t)

#vérification par des booléens de certaines valeurs attendues dans le tableau 

print('Les valeurs attendues sont vérifiées ? :',t[1] == 1.5, t[4] == 5, t[9] == 10.5, t[16] == 18, t[25] == 27.5)
```

```python
#calcul de v_x et v_y pour t2 - 1 < t <t2...

t_1 = np.arange(t[2] - 1, t[2], 0.01) #pas de 0.01 pour obtenir une courbe lisse 

v_x1 = np.empty(len(t_1))
v_y1 = np.empty(len(t_1))


for i in range (0, len(t_1)):
    v_x1[i] = epsilon1*np.sqrt(2)*np.sin(np.pi*(t_1[i] - t[2]))
    v_y1[i] = (-1)* epsilon1*np.sqrt(2)*np.cos(np.pi*(t_1[i] - t[2]))
    
#représentation graphique pour v_x et v_y pour les intervalles t2 - 1 < t <t2...
                                        
plt.plot(t_1, v_x1, 'b', label = 'composante v_x')
plt.plot(t_1, v_y1, 'r', label = 'composante v_y')                                         
plt.title('Composantes en x et y de la vitesse du proton pour $t\in [t2 - 1, t2]$')
plt.xlabel('temps')
plt.ylabel('composantes de la vitesse')
plt.legend()
plt.grid()
```

```python
''' la fonction subplot n'a pas tellement fonctionné donc j'ai décidé de faire 2 représentations à part pour ne pas perdre de temps'''

#... et pour t4 - 1 < t < t4  
                                           
t_2 = np.arange(t[4] - 1, t[4], 0.01) #pas de 0.01 pour obtenir une courbe lisse                                            
                                           
v_x2 = np.empty(len(t_2))
v_y2 = np.empty(len(t_2))                                           

for i in range (0, len(t_2)):
    v_x2[i] = epsilon1*np.sqrt(4)*np.sin(np.pi*(t_2[i] - t[4]))
    v_y2[i] = - epsilon1*np.sqrt(4)*np.cos(np.pi*(t_2[i] - t[4]))
    
#... et pour t4 - 1 < t < t4

plt.plot(t_2, v_x2, 'g', label = 'composante v_x')
plt.plot(t_2, v_y2, 'c', label = 'composante v_y')                                         
plt.title('Composantes en x et y de la vitesse du proton pour $t\in [t4 - 1, t4]$')
plt.xlabel('temps')
plt.ylabel('composantes de la vitesse')
plt.legend()
plt.grid()
```

Les maximas des composantes $v_x$ et $v_y$ pour $t\in [t2 - 1, t2]$ semblent être : $v_{xmax} \approx 1.4 $ et $v_{ymax} \approx v_{xmax} $

Les maximas des composantes $v_x$ et $v_y$ pour $t\in [t4 - 1, t4]$ semblent être : $v_{xmax} \approx 2 $ et $v_{ymax} \approx v_{xmax} $

Ceci s'explique car aux temps t2-1, t2, t4-1 et t4 la vitesse du proton est entièrement dirigée selon y. La composante selon y de la vitesse (en valeur absolue) représente donc la vitesse totale du proton. $v_y$ négatif aux temps t2-1 et t4-1 car le proton se déplace selon -y et positif aux temps t2 et t4 car le proton se déplace selon +y.

"Au milieu" de ces deux intervalles de temps soit aux temps $t = 2.2$ et $t = 4.5$, la vitesse du proton est en revanche dirigée entièrement selon x. Cette fois-ci, la composante selon x (toujours en valeur absolue) représente donc la vitesse totale du proton ce qui explique qu'on ait remarqué $v_{ymax} \approx v_{xmax} $ car la vitesse est constante dans les endroits du cyclotron D supérieur et inférieur.

Le proton étant accéléré entre D supérieur et inférieur, on trouve un maximum plus grand pour l'intervalle $t\in [t4 - 1, t4]$ que celui pour $t\in [t2 - 1, t2]$, le proton a été accéléré.

## Partie 2 

```python
#représentaion de v_x et v_y pour 0 < t < t6

temps = np.arange(t[0], t[6], 0.01)
v_x = np.empty(len(temps))
v_y = np.empty(len(temps))

for i in range (0, len(temps)):
    for j in (0, 7):
        if t[j] < temps[i] < t[j+1] -1 :
            v_x[i] = 0
            v_y[i] = (-1)**j*(np.sqrt(j)+(temps[i] - t[j])/taux)
        else :
            v_x[i] = (-1)**j*np.sqrt(j+1)*np.sin(np.pi*(temps[i]-t[j+1]))
            v_y[i] = (-1)*(-1)**j*np.sqrt(j+1)*np.cos(np.pi*(temps[i]-t[j+1]))
                                                      
#représentation graphique 

plt.plot(temps, v_x, 'm', label = 'composante selon x')
plt.plot(temps, v_y, 'k', label = 'composante selon y')
plt.legend()
plt.xlabel('temps')
plt.ylabel('composantes de la vitesse')
plt.title("Composantes de la vitesse du proton sur l'intervalle $t\in[0, t_6]$") 
```

```python
module = np.sqrt(v_x**2 + v_y**2)

plt.plot(temps, module, 'k')
plt.legend()
plt.xlabel('temps')
plt.ylabel('module de la vitese')
plt.title("Module de la vitesse du proton sur l'intervalle $t\in[0, t_6]$")
```
