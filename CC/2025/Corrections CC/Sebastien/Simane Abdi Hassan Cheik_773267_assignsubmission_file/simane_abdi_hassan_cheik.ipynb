```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### question 1

```python

#creations des variables
tau= 0.5
n= np.linspace(1,50,50)  #tableau des valeurs de n choisies arbitrairement
tn= np.zeros(50)  #tableau du temps san unité

for i in range (len(n)) :  
    tn[i]= i +tau*np.sqrt(i)
print(tn) #affiche le tableau des valeurs de t selon n
```

#### question3

```python
t2= np.linspace(tn[2]-1,tn[2],100)  #création du tableau pour t2
vx2= np.zeros(100) #tableau vide vx de dimension 100
vy2= np.zeros(100) #tableau vide vy de dimension 100
for i in range (len(n)-1) :
    vx2[i] = (-1)**i*np.sqrt(i+1)*np.sin(np.pi*(t2[i]-t2[i+1])) 
    vy2[i] = -(-1)**i*np.sqrt(i+1)*np.cos(np.pi*(t2[i]-t2[i+1]))
#print(vx2,vy2) pour vérifier

t4= np.linspace(tn[4-1],tn[4],100)  #création du tableau pou t4
vx4= np.zeros(100) #tableau vide vx de dimension 100
vy4= np.zeros(100) #tableau vide vy de dimension 100
for i in range (len(n)-1) :
    vx4[i] = (-1)**i*np.sqrt(i+1)*np.sin(np.pi*(t4[i]-t4[i+1]))
    vy4[i] = -(-1)**i*np.sqrt(i+1)*np.cos(np.pi*(t4[i]-t4[i+1]))
#print(vx4,vy4) pour vérifier
```

#### question4

```python
#entre t2-1 et t2
plt.plot(t2,vx2, label = "Vx")
plt.plot(t2,vy2, label = "Vy")
plt.title("Graphique des positions x et y en fonctions du temps entre t2-1 et t2")
plt.xlabel("Temps sans unité")
plt.ylabel("Position")
plt.grid()
plt.legend()
plt.show()
```

```python
# entre t4-1 et t4
plt.plot(t4,vx4, label = "Vx")
plt.plot(t4,vy4, label = "Vy")
plt.title("Graphique des positions x et y en fonctions du temps entre t4-1 et t4")
plt.xlabel("Temps sans unité")
plt.ylabel("Position")
plt.grid()
plt.legend()
plt.show()
```

#### question 6

```python
t6= np.linspace(tn[0],tn[6],100)
vx6= np.zeros(100)
vy6= np.zeros(100)
#print(t6)
for i in range (len(n)-1) :
    if t6[i+1]-1 < t6[i] < t6[i+1]:  # dans le cas 2
        vx6[i] = (-1)**i*np.sqrt(i+1)*np.sin(np.pi*(t6[i]-t6[i+1]))
        vy6[i] = -(-1)**i*np.sqrt(i+1)*np.cos(np.pi*(t6[i]-t6[i+1]))       
    elif t6[i] < t6[i] < t6[i+1]-1:   #dans le cas 1
        vx6[i] = 0
        vy6[i] = (-1)**i*(np.sqrt(n)+(t6[i]-t6[i])/tau)
```

```python
plt.plot(t6,vx6,label = "Vx")
plt.plot(t6,vy6, label = "Vy")
plt.title("Graphique des positions x et y en fonctions du temps entre 0 et t6")
plt.xlabel("Temps sans unité")
plt.ylabel("Position")
plt.grid()
plt.legend()
plt.show()
```

#### question 7

```python
vitesse = np.sqrt( vy6**2+vx6**2) #expression de la vitesse soit le module de vx et vy

plt.plot(t6, vitesse, label = "vitesse")
plt.title("Graphique des positions x et y en fonctions du temps entre 0 et t6")
plt.xlabel("Temps sans unité")
plt.ylabel("Position")
plt.grid()
plt.legend()
plt.show()
```
