# Particule dans un cyclotron

Un cyclotron comporte trois parties, voir la figure ci-dessous. Les parties supérieure et inférieure, appelées "D" à cause de la forme de la trajectoire, sont plongées dans un champ magnétique.

![](Figure3.png)

Au temps $t_0=0$ **(ici le temps est sans unité)**, un proton est injecté au niveau du D inférieur sans vitesse. Une tension alternative est appliquée entre les D. Elle crée un champ électrique dont la direction et le sens sont toujours **similaires** à ceux de la vitesse du proton lorsqu’il est entre les D. Les instants $t_n$ où le proton sort des D (soit inférieur, soit supérieur, voir la figure) sont donnés, **sans unité**, par
$$t_n=n+\tau \sqrt{n}.$$
On choisit $\tau=0.5$. Le temps passé par le proton dans un D (supérieur ou inférieur) est toujours de $1$.
La trajectoire du proton est plane. On s’intéresse à sa vitesse. Ses composantes sont données, dans un repère et des unités appropriés, **en fonction de sa position dans le cyclotron:**
1) Pour $t$ variant de $t_n$ à $t_{n+1}-1$ c’est-à-dire quand il est entre les D :
$$
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$$
2) Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

```python
#importer les modules necessaires
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#1. 
t0 = 0 #(sans unité) proton injecté au niveau du D inferieur sans vitesse (v0=0)
tau = 0.5 #valeur de tau

t = [] #liste vide

#boucle pour 50 insants
for n in range(49):
    t.append(n + (tau * np.sqrt(n))) #formule pour les instants t[n] ou le proton sort des D (sans unité)

tn = np.array(t) #convertir la liste en tableau numpy

#2.

if (tn[1] == 1.5 and tn[4] == 5 and tn[9] == 10.5 and tn[16] == 18 and tn[25] == 27.5): #donne True si tous les comparaisons sont vrais
    print("Les valeurs sont vérifiés et sont correctes")

#3. 

n = 1 #pour la partie t_2−1 à t_2

#fonction pour determiner la valeur de e selon ses proprietés
def e(n):
    "' pour determiner la valeur de e selon ses proprietés '"
    return (-1)**n

#boucle pour determiner la valeur de vx(t) et vy(t) avec n=1
for t in range(len(tn)):
    vx1 = e(n) * np.sqrt(n + 1) * np.sin(np.pi * (tn - tn[n + 1]))
    vy1 = -e(n)* np.sqrt(n + 1)* np.cos(np.pi * (tn - tn[n + 1]))

n = 3 #pour la partie t_4−1 à t_4

#boucle pour determiner la valeur de vx(t) et vy(t) avec n=3
for t in range(len(tn)):
    vx2 = e(n) * np.sqrt(n + 1) * np.sin(np.pi * (tn - tn[n + 1]))
    vy2 = -e(n)* np.sqrt(n + 1)* np.cos(np.pi * (tn - tn[n + 1]))

#4.

#création du graphique pour n=1 aved vx et vy en fonction de temps
plt.plot(tn, vx1, vy1)
plt.xlabel('temps (sans unité)')
plt.ylabel('la vitesse des protons')
plt.title(' la vitesse des protons en fonction de temps passé pour n=1')
plt.legend(['vitesse x', 'vitesse y'])
plt.grid()
plt.show()

#création du graphique pour n=3 aved vx et vy en fonction de temps
plt.plot(tn, vx2, vy2)
plt.xlabel('temps (sans unité)')
plt.ylabel('la vitesse des protons')
plt.title(' la vitesse des protons en fonction de temps passé pour n=3')
plt.legend(['vitesse x', 'vitesse y'])
plt.grid()
plt.show()

#5. 
#calculer les maximums valeurs de vitesse quand n=1 
#j'ai utilisé la valeur de vitesse absolut (np.abs comme avec sans direction négatif ou positif) car je pense que c'est plus répresentatif du vrai vitesse maximal. 
#si on utilise pas la valeur absolut, on va manquer les valeurs qui sont peut etre vite, mais dans la direction opposé
maximum_vx1 = str(max(np.abs(vx1)))
maximum_vy1 = max(np.abs(vy1))

#pour trouver le valeur de temps qui correspond à la vitesse maximale
dfx1 = pd.DataFrame(data = vx1, index = tn)
print(dfx1)

dfy1 = pd.DataFrame(data = vy1, index = tn)
print(dfy1)

print("La valeur maximale de vx = " + str(maximum_vx1) + " à t=44.2 et vy = " + str(maximum_vy1) + " à t=2.71 quand n = 1")
      
#calculer les maximums valeurs de vitesse quand n=3
maximum_vx2 = max(np.abs(vx2))
maximum_vy2 = max(np.abs(vy2))

#pour trouver le valeur de temps qui correspond à la vitesse maximale
dfx2 = pd.DataFrame(data = vx2, index = tn)
print(dfx2)

dfy2 = pd.DataFrame(data = vy2, index = tn)
print(dfy2)

print("La valeur maximale de vx = " + str(maximum_vx2) + " à t=0.5 et vy = " + str(maximum_vy2) + " à t= 0 quand n = 3")
```
