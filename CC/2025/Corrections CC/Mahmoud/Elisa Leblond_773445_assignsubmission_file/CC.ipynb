### Partie 1

On a : $$t_n=n+\tau \sqrt{n}.$$

```python
#création du tableau des temps 
D=[] #initialisation du tableau 
for n in range(0,50):
    values=n+taux*np.sqrt(n)
    D.append(values)
print(D[1]==1.5) # vérification des valeurs
print(D[16]==18)
print(D[25]==27.5)
```

On a vérifié que les valeurs t1, t16, et t25, correspondent bien avec les valeurs théoriques attendues.

On veut calculer les valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$ et pour $t$ variant de $t_4-1$ à $t_4$.

```python
#définition des intervals de temps
t2=D[2]
t2_1=t2-1

temps2=np.linspace(t2_1, t2, 50)
b=np.array(temps2)

t4=5
t4_1=4
temps4=np.linspace(t4,t4_1, 50)
c=np.array(temps4)
```

```python
#on calcule les vitesse pour le premier interval
Vx2=0*temps2
Vy2=(np.sqrt(2))+((D-temps2)/taux)
```

On répresente les vitesses en fonction du temps

```python
#Ajout du graphique et des légendes pour Vx2
plt.plot(D, Vx2 , label="Vitesse" , color="blue") 
plt.xlabel("temps (s)") 
plt.ylabel( "Vitesse selon x")
plt.title( "Vx en fonction du temps dans l'intervalle t2-1 à t2") 
plt.grid()
plt.legend() 
plt.show()


#Ajout du graphique et des légendes pour Vy2
plt.plot(D, Vy2 , label="Vitesse" , color="blue") 
plt.xlabel("temps (s)") 
plt.ylabel( "Vitesse selon y")
plt.title( "Vy en fonction du temps dans l'intervalle t2-1 à t2") 
plt.grid()
plt.legend() 
plt.show()

          
```

```python
On a bien une vitesse qui est nulle selon x et croissante selon y.
```

```python
#on calcule les vitesse pour le deuxième interval
Vx4=0*temps4
Vy4=2+((D-temps4)/taux)
```

```python
plt.plot(D, Vx4 , label="Vitesse" , color="blue") 
plt.xlabel("temps (s)") 
plt.ylabel( "Vitesse selon x")
plt.title( "Vx en fonction du temps dans l'intervalle t4-1 à t4") 
plt.grid()
plt.legend() 
plt.show()


plt.plot(D, Vy4 , label="Vitesse" , color="blue") 
plt.xlabel("temps (s)") 
plt.ylabel( "Vitesse selon y")
plt.title( "Vy en fonction du temps dans l'intervalle t4-1 à t4") 
plt.grid()
plt.legend() 
plt.show()
```

```python
#détermination des miximas pour Vx2 et Vy2

max_Vx2=max(Vx2)
max_Vy2=max(Vy2)

print(max_Vx2)
print(max_Vy2)
```

```python
#détermination des miximas pour Vx4 et Vy4
max_Vx4=max(Vx4)
max_Vy4=max(Vy4)

print(max_Vx4)
print(max_Vy4)
```
