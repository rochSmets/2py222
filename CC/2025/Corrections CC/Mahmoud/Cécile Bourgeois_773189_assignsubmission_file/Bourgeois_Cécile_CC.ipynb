## Partie I

1. Créer un tableau contenant les instants de sortie des boucles "D" : $t_0$, $t_1$, ... jusqu'à $t_{50}$ inclus. 
2. Vérifier que $t_1=1.5$, $t_4=5$, $t_9=10.5$, $t_{16}=18$, et $t_{25}=27.5$ avec des opérateurs booléens.
3. Calculer les valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$. Vous utiliserez un échantillonage suffisant pour le temps $t$, dans un tableau distinct de celui de la question précédente. Faire de même pour $t$ variant de $t_4-1$ à $t_4$. 
4. Représenter dans un graphique $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $t_2-1$ a $t_2$. Faire de même pour $t$ variant de $t_4-1$ à $t_4$.
5. Déterminer des valeurs approchées des maxima de $v_x$ et $v_y$ pour $t$ variant de $t_2-1$ à $t_2$, ainsi que les instants $t$ correspondants. Un maximum peut être en bord d'intervalle. Faire de même pour $t$ variant de $t_4-1$ à $t_4$. Commenter.

```python
#Questions 1 et 2

t=[] #création d'une liste vide pouvant stocker les valeurs de t à la sortie de la boucle
tau=0.5
for i in range(50):
    instant=i+tau*np.sqrt(i)
    t.append(instant)
    
t=np.array(t) #conversion de la liste des temps en un tableau numpy


print(f"La première condition t1=1.5 est {t[1]==1.5}")
print(f"La deuxième condition t4=5 est {t[4]==5}")
print(f"La troisième condition t9=10.5 est {t[9]==10.5}")
print(f"La quatrième condition t16=18 est {t[16]==18}")
print(f"La cinquième condition t25=27.5 est {t[25]==27.5}")
```

```python
#Question 3
delta_t2=np.linspace(t[2]-1,t[2],50) #création du tableau avec les valeurs de t comprises entre t_2-1 et t_2
En=1 # car n=2 et n=4 sont des nombres paires

v_x=[]
v_y=[]
#même méthode que pour le temps on crée une boucle pour calculer la valeur de v_x et v_y pour chaque valeur de t puis on les stocke dans les listes correspondantes.
for i in delta_t2: 
    v_x.append(En*np.sqrt(2+1)*np.sin(np.pi*(i)))
    v_y.append(-En*np.sqrt(2+1)*np.cos(np.pi*(i)))

#conversion des listes en tableau numpy
v_x=np.array(v_x)
v_y=np.array(v_y) 


#même méthode que pour delta_2
delta_t4=np.linspace(t[4]-1,t[4],50)
v_x4=[]
v_y4=[]
for i in delta_t4:
    v_x4.append(En*np.sqrt(4+1)*np.sin(np.pi*(i)))
    v_y4.append(-En*np.sqrt(4+1)*np.cos(np.pi*(i)))

v_x4=np.array(v_x4)
v_y4=np.array(v_y4)
```

```python
#question 4

plt.figure(figsize=(13,6)) #augmentation de la taille de la figure pour améliorer la visibilité avec la fonction subplot
plt.subplot(1,2,1) #permet d'afficher les deux graphiques côte à côte 
plt.plot(delta_t2,v_x, label='vitesse selon x')
plt.plot(delta_t2,v_y, label='vitesse selon y')
plt.title("Vitesses d'une particule pour t variant de t_2-1 à t_2")
plt.xlabel("Temps (sans unité)")
plt.ylabel("Vitesses (sans unité)")
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(delta_t4,v_x4, label='vitesse selon x')
plt.plot(delta_t4,v_y4, label='vitesse selon y')
plt.title("Vitesses d'une particule pour t variant de t_4-1 à t_4")
plt.xlabel("Temps (sans unité)")
plt.ylabel("Vitesses (sans unité)")
plt.grid()
plt.legend() #permet d'afficher les légendes des fonctions 
plt.show()
```

```python
#Question 5

#pour t variant de t_2-1 à t2
v_xmax=np.max(v_x)
print(f"Le maximum de la vitesse selon x est {v_xmax:.2f} pour t variant de t_2-1 à t_2.")
v_ymax=np.max(v_y)
print(f"Le minimum de la vitesse selon y est {v_ymax:.2f} pour t variant de t_2-1 à t_2.", '\n')


#pour t variant de t_4-1 à t4
v_xmax4=np.max(v_x4)
print(f"Le maximum de la vitesse selon x est {v_xmax4:.2f} pour t variant de t_4-1 à t4.")
v_ymax4=np.max(v_y4)
print(f"Le minimum de la vitesse selon y est {v_ymax4:.2f} pour t variant de t_4-1 à t4.")
```

Par lecture graphique nous trouvons approximativement les temps maximaux suivants: $t_{xmax}= 2.5$ pour le premier graphique. Nous trouvons $t_{xmax}= 4.5$ et $t_{ymax}=5.0$ pour le deuxième graphique. Nous constatons que les temps correspondants aux maximaux se rapprochent lorsque le temps augmente. 

## Partie II
6. Représenter $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$. On utilisera une couleur unique pour la courbe par morceaux. On remarque que $v_x$ et $v_y$ sont des fonctions continues du temps $t$. Commenter.
7. Représenter le module de la vitesse $\sqrt{v_x^2+v_y^2}$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$. On utilisera une couleur unique pour la courbe par morceaux. Commenter. A quoi sert un cyclotron ?
8. Pour $n$ variant de $0$ à $50$, représenter le module de la vitesse $\sqrt{v_x(t_n)^2+v_y(t_n)^2}$ en fonction de $t_n-n$, qui est le temps total passé par le proton entre les D à l'instant $t_n$. Commenter.

```python
#Question 6
temps=np.linspace(0,t[6],175) #creation d'un  tableau de temps pour 175 valeurs entre 0 et t_6 pour avoir une courbe continue: choix du pas de discrétisation
En=1 #car n=6 est un nombre paire

vx=[]
vy=[]
for i in range(len(temps)):
    vx.append(En*np.sqrt(i+1)*np.sin(np.pi*(temps[i-1]+temps[i])))
    vy.append(-En*np.sqrt(i+1)*np.cos(np.pi*(temps[i-1]+temps[i])))
vx=np.array(vx)
vy=np.array(vy)


plt.plot(temps,vx, label='vitesse selon x')
plt.plot(temps,vy, label='vitesse selon y')
plt.title("Vitesses d'une particule pour t variant de 0 à t_6")
plt.xlabel("Temps (sans unité)")
plt.ylabel("Vitesses (sans unité)")
plt.legend()
plt.grid()
plt.show()
```

On remarque que les fonctions $v_x$ et $v_y$ voient leur amplitude augmenter et se rapprocher en même temps que le temps augmente. On remarque aussi que les courbes tracées pour t variant de $t_2$-1 à $t_2$ et t variant de $t_4$-1 à $t_4$ correspondent bien à des parties de la courbe pour t variant de 0 à $t_6$. On retrouve bien une fonction sinus pour $v_x$ et une fonction cosinus pour $v_y$.

```python
#Question 7

module=np.sqrt(vx**2+vy**2) #calcul du module de la vitesse 
plt.plot(temps,module, label="$v_x^{2}+v_y^{2}$")
plt.xlabel("Temps (sans unité)")
plt.ylabel("Module (sans unité)")
plt.title("Module de la vitesse  pour t variant de  0 à  t_6")
plt.grid()
plt.legend()
plt.show()
```

On constate avec le graphique que la courbe du module de la vitesse est croissante en fonction du temps. Ainsi la vitesse d'une particule va être considerablement augmenter dans un cyclotron. Un cyclotron permet d'augmenter la vitesse d'une particule au cours du temps, c'est un accélérateur de particules. 

```python
#Question 8
En=1 #car n=50 est un nombre paire

vx_n=[]
vy_n=[]
for i in range(50):
    vx_n.append(En*np.sqrt(i+1)*np.sin(np.pi*(t[i-1]+t[i])))
    vy_n.append(-En*np.sqrt(i+1)*np.cos(np.pi*(t[i-1]+t[i]))) 
vx_n=np.array(vx_n)
vy_n=np.array(vy_n)

module_n=np.sqrt(vx_n**2+vy_n**2)

plt.plot(t, module_n, label="$v_x (t_n)^{2}+v_y (t_n)^{2}$")
plt.xlabel("Temps (sans unité)")
plt.ylabel("Module de la vitesse (sans unité)")
plt.title("Module de la vitesse  pour t variant de  0 à 50")
plt.grid()
plt.legend()
plt.show()
```

On constate que la vitesse augmente avec le temps ce qui est le résultat attendu d'une particule dans un accélérateur cyclotron. Le module atteint à 50 secondes n'est pas le même que celui d'avant car il n'y a pas le même nombre de points dans notre tableau temps. Nous avons été obligé de rajouter des points pour t variant de 0 à $t_6$ pour avoir une fonction qui est bien discrétisée.
