```python
# Importation des bibliothèques

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Question 1

t = []
tau = 0.5

for n in range(51):
     t.append(n + tau * np.sqrt(n))
        
tableau = np.array(t)
```

```python
#Question 2

print(t[1] == 1.5)
print(t[4] == 5)
print(t[9] == 10.5)
print(t[16] == 18)
print(t[25] == 27.5)
```

```python
#Question 3

# Pour t variant de t2-1 à t2

n = 1

temps = np.linspace(t[n+1] - 1, t[n + 1], 100)
V_x = []
V_y = []

for i in temps:    
    v1 = (-1)**n * np.sqrt(n + 1) * np.sin(np.pi*(i - t[n+1]))
    V_x.append(v1)

for i in temps:    
    v2 = - (-1)**n * np.sqrt(n + 1) * np.cos(np.pi*(i - t[n+1]))
    V_y.append(v2)
    
Tab_Vx = np.array(V_x)
Tab_Vy = np.array(V_y)
```

```python
#Question 6

# Attention : Pour avoir le bon affichage il faut relancer la cellule une deuxième fois, ainsi que les suivantes (je ne sais pas pourquoi)

# On définit 3 temps différents

nv_temps = np.linspace(0,t[6],600) #Temps total

temps_1_partie = np.linspace(t[n], t[n+1]-1,50) # Temps où on applique les premières formules données dans l'énoncé

temps_2_partie = np.linspace(t[n+1]-1, t[n+1],50) # Temps où on applique les deuxièmes formules données dans l'énoncé

liste_n = [0,1,2,3,4,5]
v_x = []
v_y = []

for n in liste_n:
    for i in temps_1_partie:
        m = 0
        v_x.append(m)
        m1 = (-1)**n * (np.sqrt(n) +( i - t[n] )/  tau)
        v_y.append(m1)
    for i in temps_2_partie:
        m2 = (-1)**n * np.sqrt(n + 1) * np.sin(np.pi*(i - t[n+1]))
        v_x.append(m2)
        m3 = - (-1)**n * np.sqrt(n + 1) * np.cos(np.pi*(i - t[n+1]))
        v_y.append(m3)
```

```python
#Tracé du graphe des deux vitesses selon le temps 

plt.plot(nv_temps, v_x, label = "Vx")
plt.plot(nv_temps, v_y, label = "Vy")

plt.xlabel("Temps(s)")
plt.ylabel("Vitesse (m/s)")
plt.title("Graphique représentant Vx et Vy en fonction de t pour t variant de 0 à t6")
plt.legend()
```

On remarque sur le graphe que Vy et Vx sont bien continues selon t. On voit aussi que Vy varie de manière beaucoup plus prononcé que Vx au début mais que plus on avance dans le temps, plus ses fluctuations diminuent et se rapprochent de celles de Vx. 

```python
#Question 7

#Transformation des listes en tableau pour pouvoir mettre au carré pour calculer le modulo

tab_vx = np.array(v_x) 
tab_vy = np.array(v_y)

Module_vitesse = np.sqrt(tab_vx**2 + tab_vy**2)

#Tracé du modulo en fonction du temps

plt.plot(nv_temps, Module_vitesse)

plt.xlabel("Temps(s)")
plt.ylabel(" Modulo vitesse (m/s)")
plt.title("Graphique représentant Le modulo de la vitesse en fonction de t pour t variant de 0 à t6")
```

On observe sur le graphe que le modulo de la vitesse diminue. On peut en déduire qu'un cyclotron sert à ralentir les protons
