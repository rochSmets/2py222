```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

1)

```python
# tableau de n valeur 
n = 51 
N = np.ones(n)

p = 0.5   # taux 

# tableau de tn

D = np.ones(n)
D[0] = 0     # initialisation 

for i in range(n) : 
    D[i] = i + p * (i)**(1/2)
```

2)

```python
# regarder certaine valeurs 
print("t1 = {}".format(D[1]))
print("t4 = {}".format(D[4]))
print("t9 = {}".format(D[9]))
print("t16 = {}".format(D[16]))
print("t25 = {}".format(D[25]))
```

On retrouve les bonnes valeurs pour les tn donnée.

3)

```python
print("t2 = {}".format(D[2]))
```

```python
# definition du temps pour t variant de t2 - 1 a t2

temps_initial = D[2] - 1   
temps_final = D[2]  
pas_temps = 0.01

intervalle = temps_final - temps_initial
num_points = int(intervalle / pas_temps) + 1   # Nombre d'éléments
t = np.linspace(temps_initial, temps_final, num_points)

# Vx et Vy 
Vx = np.ones(num_points)
Vy = np.ones(num_points)


for i in range (2) : 
    D[i] = i + p * (i)**(1/2)
    for j in range(num_points) :
        Vx[j] = (-1)**(i) * (i + 1)**(1/2) *  np.sin(np.pi * (t[j] - D[i]))
        Vy[j] = -(-1)**(i) * (i + 1)**(1/2) *  np.cos(np.pi * (t[j] - D[i]))
```

```python
# definition du temps

temps_initial = D[4] - 1   
temps_final = D[4]  
pas_temps = 0.01

intervalle = temps_final - temps_initial
num_points1 = int(intervalle / pas_temps) + 1   # Nombre d'éléments
t1 = np.linspace(temps_initial, temps_final, num_points1)

# Vx et Vy pour t variant de t4 - 1 a t4
Vx1 = np.ones(num_points1)
Vy1 = np.ones(num_points1)


for i in range (4) : 
    D[i] = i + p * (i)**(1/2)
    for j in range(num_points1) :
        Vx1[j] = (-1)**(i) * (i + 1)**(1/2) *  np.sin(np.pi * (t1[j] - D[i]))
        Vy1[j] = -(-1)**(i) * (i + 1)**(1/2) *  np.cos(np.pi * (t1[j] - D[i]))
```

4)

```python
# graphe pour t variant de t2 - 1 a t2
plt.plot(t,Vx,'r')
plt.plot(t,Vy,'b')
plt.xlabel('t variant de t2 - 1 a t2')
plt.ylabel('Vx et Vy')
plt.title('Vx et Vy en fonction de t variant de t2 - 1 a t2')
plt.legend(['Vx' ,'Vy'])
plt.grid
```

```python
# graphe pour t variant de t4 - 1 a t4
plt.plot(t1,Vx1,'r')
plt.plot(t1,Vy1,'b')
plt.xlabel('t variant de t4 - 1 a t4')
plt.ylabel('Vx et Vy')
plt.title('Vx et Vy en fonction de t')
plt.legend(['Vx' ,'Vy'])
plt.grid
```
