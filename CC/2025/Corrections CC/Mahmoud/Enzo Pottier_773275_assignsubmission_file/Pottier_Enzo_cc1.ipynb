### Pottier Enzo - CC Modélisation numérique en physique


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Partie 1

# 1)

```python

t=[n+0.5*np.sqrt(n) for n in range(0,51)]
```

# 2)

```python

if t[1]==1.5:
    print(True)
if t[4]==5:
    print(True)
if t[9]==10.5:
    print(True)
if t[16]==18:
    print(True)
if t[25]==27.5:
    print(True)
    
```

# 3)
Pour t variant de $t_2−1$ à $t_2$, on prend $n=1$. Ainsi $\epsilon_n=-1$.

```python

t_test1=np.linspace(t[2]-1,t[2],100)
vx1=[-1*np.sqrt(1+1)*np.sin(np.pi*(t_test1[i]-t[2])) for i in range(0,len(t_test1))]

vy1=[1*np.sqrt(1+1)*np.cos(np.pi*(t_test1[i]-t[2])) for i in range(0,len(t_test1))]
```

De même pour t variant de $t_4−1$ à $t_4$,on prend $n=3$ et $\epsilon_n=-1$

```python
t_test2=np.linspace(t[4]-1,t[4],100)
vx2=[-1*np.sqrt(3+1)*np.sin(np.pi*(t_test2[i]-t[4])) for i in range(0,len(t_test2))]

vy2=[1*np.sqrt(3+1)*np.cos(np.pi*(t_test2[i]-t[4])) for i in range(0,len(t_test2))]


```

# 4)

```python
plt.plot(t_test1,vx1,label="vx")
plt.plot(t_test1,vy1,label="vy")
plt.title("graphique de la vitesse en fonction du temps pour t compris entre t2 et t2-1")
plt.xlabel("temps ")
plt.ylabel("vitesse")
plt.legend()
plt.show()
```

```python
plt.plot(t_test2,vx2,label="vx")
plt.plot(t_test2,vy2,label="vy")
plt.title("graphique de la vitesse en fonction du temps pour t compris entre t4 et t4-1")
plt.xlabel("temps ")
plt.ylabel("vitesse")
plt.legend()
plt.show()
```

# 5)

```python
max_x1=vx1[0]
max_y1=vy1[0]
indice_x1=0
indice_y1=0
for i in range(len(vx1)):
    if vx1[i]>max_x1:
        max_x1=vx1[i]
        indice_x1=i
    if vy1[i]>max_y1:
        max_y1=vy1[i]
        indice_y1=i
print("la vitesse maximale selon x est",max_x1, "atteinte à t=",t_test1[indice_x1])
print("la vitesse maximale selon y est",max_y1,"atteinte à t =",t_test1[indice_y1])
```

```python
max_x2=vx2[0]
max_y2=vy2[0]
indice_x2=0
indice_y2=0
for i in range(len(vx2)):
    if vx2[i]>max_x2:
        max_x2=vx2[i]
        indice_x2=i
    if vy2[i]>max_y2:
        max_y2=vy2[i]
        indice_y2=i
print("la vitesse maximale selon x est",max_x2, "atteinte à t=",t_test1[indice_x2])
print("la vitesse maximale selon y est",max_y2,"atteinte à t =",t_test1[indice_y2])
```

On peut voir que dans les deux cas la vitesse maximale selon x est atteinte au même momment. On peut faire la même observation pour la vitesse selon y.

## Partie 2

# 6)

```python

temps=[]
for i in range(6):
    for j in np.linspace(t[i],t[i+1]-1):        # pour t variant de t_n à t_(n+1)-1
        temps.append(j)
    for k in np.linspace(t[i+1]-1,t[i+1]):    # pour t variant de t_(n+1)-1 à t_(n+1)
        temps.append(k)                       # sans indication, la fonction linspace renvoie 50 valeurs

vx=[]
vy=[]

compteur=0                                    # compteur servant à se repérer dans la liste temps
for n in range(6):
    for i in range(compteur,compteur+50):     # calcul séparer des valeurs de vx et vy dépendant de si t est dans t_n à t_(n+1)-1 ou bien dans t_(n+1)-1 à t_(n+1)
        vx.append(0)
        vy.append(((-1)**n)*(np.sqrt(n)+(temps[i]-t[n])/0.5))
    compteur+=50
    for j in range(compteur,compteur+50):
        vx.append((-1)**n*np.sqrt(n+1)*np.sin(np.pi*(temps[i]-t[n+1])))
        vy.append(-(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(temps[i]-t[n+1])))

```

```python
plt.plot(temps,vx,label="vx")
plt.plot(temps,vy,label="vy")
plt.title("graphique de la vitesse en fonction du temps pour t vatiant de t0 à t6")
plt.xlabel("temps")
plt.ylabel("vitesse")
plt.legend()
plt.show()
```

On peut voir que les protons accélèrent et déccélère sans arrêt.

# 7)

```python
vitesse=[]

for i in range(len(temps)):
    vitesse.append(np.sqrt(vx[i]**2+vy[i]**2))

plt.plot(temps,vitesse)
plt.xlabel("temps")
plt.ylabel("vitesse")
plt.title("graphique du module de la vitesse en fonction du temps")

```

On peut voir que la vitesse de proton augemnte par palier.On retrouve un forte diminution de vitesse avant de voir une nouvelle accélération.Le cyclotron permet donc accélérer un proton  

# 8)


```python
v=[]
t_bis=[]
for n in range(0,50):      
    v.append(np.sqrt(((-1)**n*np.sqrt(n+1)*np.sin(np.pi*(temps[n]-t[n+1])))**2 +(-(-1)**n*np.sqrt(n+1)*np.cos(np.pi*(temps[n]-t[n+1])))**2))
    t_bis.append(temps[n]-n)

plt.plot(t_bis,v)
plt.xlabel("t_n -n")
plt.ylabel("vitesse")
plt.title("graphique du module de la vitesse en fonction de t_n - n pour n compris entre 0 et 50")
```
