## Partie 1

```python
#Question 1
import numpy as np
import math as mt
tau=0.5
n=np.arange(0,51,1)
t=n+tau*n**(1/2)
#Question 2
print(t[1]==1.5,t[4]==5,t[9]==10.5,t[16]==18,t[25]==27.5)
t[2]
```

```python
#Question 3
def vitesses(n):
    """permet de calculer les vitesses vx et vy pour t allant de tn - 1 à tn"""
    t2=np.linspace(t[n]-1,t[n],20) #crée tableau échantillonage
    vx=(-1)**(n-1)*n**(1/2)*np.sin(np.pi*(t2-t[n])) 
    vy=-(-1)**(n-1)*(n)**(1/2)*np.cos(np.pi*(t2-t[n]))
    return "vx:",vx,"vy:",vy
```

```python
vitesses(2) 
```

```python
vitesses(4)
```

```python
#Question 4
import matplotlib.pyplot as plt
#pour t2-1 à t2
plt.figure()
t2=np.linspace(t[2]-1,t[2],20)
plt.xlabel("t(s)")
plt.ylabel("vitesse(m.s^-1)")
plt.plot(t2,vitesses(2)[1],label="vx")
plt.plot(t2,vitesses(2)[3],label="vy")
plt.legend()
#pour t4-1 à t4
plt.figure()
t4=np.linspace(t[4]-1,t[4],20)
plt.plot(t4,vitesses(4)[1],label="vx")
plt.plot(t4,vitesses(4)[3],label="vy")
plt.legend()
plt.xlabel("t(s)")
plt.ylabel("vitesse(m.s^-1)")
```

On peut déterminer graphiquement que pour $t$ variant de $t_2−1$ à $t_2$, $v_{xmax}\approx1.7$ et $v_{ymax}\approx1.5 m.s^-1$.

pour $t$ variant de $t_4-1$ à $t_4$, $v_{xmax}\approx v_{ymax}\approx2.2 m.s^-1$ 

On peut remarqué que lorsque n augmente, c'est à dire le nombre de fois où il sort d'un D, la vitesse maximale augmente ce qui est cohérent puisque le temps passé par un proton dans une couche est constant, cependant lorsque n augmente,  la distance que le proton doit parcourir augmente par conséquent sa vitesse augmente.

## Partie II

```python
#Question 6
def vitesses2(n):
    """retourne les vitesses vx et vy pour t variant de tn à t_n+1 -1"""
    te=np.linspace(t[n],t[n+1]-1,10)
    vx=np.zeros(len(te))
    vy=(-1)**(n)*(np.sqrt(n)+(te-t[n])/tau)
    return vx,vy
vx=[]
vy=[]
for i in range(0,7):
    vx.extend(vitesses2(i)[0])
    vy.extend(vitesses2(i)[1])
    vx.extend(vitesses(i+1)[1])
    vy.extend(vitesses(i+1)[3])
t6=np.linspace(0,t[6],len(vx))
plt.figure()
plt.xlabel("t(s)")
plt.ylabel("vitesse(m.s^-1)")
plt.plot(t6,vx,label="vx",color="r")
plt.plot(t6,vy,label="vy",color='b')
plt.legend()
```

On peut remarqué que lorsque le proton est en dehors d'un D,la vitesse vy est plus élevées que lorsque est dans un D ce qui est explique pourquoi le proton retourne dans un D lorsqu'il n'y est pas.


```python
#Question 7
vx1=np.array(vx)
vy=np.array(vy)
mod=np.sqrt(vx1**2+vy**2)
plt.plot(t6,mod,color="g")
plt.xlabel("t(s)")
plt.ylabel("vitesses(m.s^-1)")
```

On peut voir que le module de la vitesse augmente, un cyclotron sert donc à accéler les protons.

```python
#Question 8
vx=[]
vy=[]
for i in range(0,50):
    vx.extend(vitesses2(i)[0])
    vy.extend(vitesses2(i)[1])
    vx.extend(vitesses(i+1)[1])
    vy.extend(vitesses(i+1)[3])
    
mod=np.sqrt(vx1**2+vy**2)
plt.plot(t6,mod,color="g")
plt.xlabel("t(s)")
plt.ylabel("vitesses(m.s^-1)")
```
