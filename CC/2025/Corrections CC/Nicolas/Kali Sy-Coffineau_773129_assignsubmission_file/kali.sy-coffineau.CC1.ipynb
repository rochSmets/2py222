```python
def t(n):
    return n+0.5*n**(1/2)

t(1)==1.5
t(4)==5
t(9)==10.5
t(16)==18
t(25)==27.5
```

```python
D = np.array([t(n) for n in range(51)])  #tableau des temps de sorties de boucle

print(D)
```

```python
def vx(t,n,tn):
    #vx de tn+1-1 a tn+1
    return ((-1)**(n-1))*np.sqrt(n)*np.sin(np.pi*(t-tn))



def vy(t,n,tn):
    #vy de tn+1-1 a tn+1
    return -((-1)**(n-1))*np.sqrt(n)*np.cos(np.pi*(t-tn))
```

```python
#tablau des vitesse de t2-1 à t2
t_2 = np.linspace(t(2)-1,t(2),100)


vx_t2 = vx(t_2,2,t(2))
vy_t2 = vy(t_2,2,t(2))
```

```python
#tablau des vitesse de t4-1 à t4
t_4 = np.linspace(t(4)-1,t(4),100)


vx_t4 = vx(t_2,4,t(4))
vy_t4 = vy(t_2,4,t(4))
```

```python
### representtaion des vitesses

# Création de la figure et des axes
plt.subplot(2,1,1)
plt.plot(t_2, vx_t2, 'g', label = 'v_x')
plt.plot(t_2, vy_t2, 'y',label = 'v_y')
plt.title('vitesse de t2-1 à t2')
plt.legend()
plt.ylabel('vitesse')



plt.subplot(2,1,2)
plt.plot(t_4, vx_t4, 'g', label = 'v_x')
plt.plot(t_4, vy_t4, 'y',label = 'v_y')
plt.title('vitesse de t4-1 à t4')
plt.ylabel('vitesse')
plt.xlabel('temps')
plt.legend()
```
