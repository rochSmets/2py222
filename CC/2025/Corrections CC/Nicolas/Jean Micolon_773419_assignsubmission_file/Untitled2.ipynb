```python
t_2= np.linspace(t_n(2)-1., t_n(2), 50)
tableau2=np.array(t_2)
t_4= np.linspace(t_n(4)-1., t_n(4), 50)
tableau4=np.array(t_4)
print(tableau4)
print(tableau4[4:10])
```

```python
def v_x2(k):
    return np.sqrt(3)*np.sin(np.pi*(tableau2[k]-t_n(2)))
def v_y2(k):
    return -np.sqrt(3)*np.cos(np.pi*(tableau2[k]-t_n(2)))


Vx2=[]
Vy2=[]
for k in range (50):
    t=tableau2[k]
    
    Vx2=Vx2+[v_x2(k)]
    Vy2=Vy2+[v_y2(k)]
    
print(Vx2)
print(Vy2)
```

```python
def v_x4(k):
    return np.sqrt(3)*np.sin(np.pi*(tableau4[k]-t_n(2)))
def v_y4(k):
    return -np.sqrt(3)*np.cos(np.pi*(tableau4[k]-t_n(2)))


Vx4=[]
Vy4=[]
for k in range (50):
    t=tableau4[k]
    
    Vx4=Vx4+[v_x4(k)]
    Vy4=Vy4+[v_y4(k)]
    
print(Vx4)
print(Vy4)
```

```python
import matplotlib.pyplot as plt

# Création de la figure
plt.plot(tableau2, Vx2, '-*')
plt.plot(tableau2, Vy2, '-*')
plt.plot(tableau4, Vx4, '-*')
plt.plot(tableau4, Vy4, '-*')
# Style
plt.xlabel('temps [s]')
plt.ylabel('vitesse [m.s$^{-1}$]')
plt.grid()
```
