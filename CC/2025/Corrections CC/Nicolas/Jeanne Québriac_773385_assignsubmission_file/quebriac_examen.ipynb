## Introduction: particule dans un cyclotron

Un cyclotron comporte trois parties:
* La partie supérieur appelée D supérieur plongée dans un champs magnétique créé par une tension alternative
* La partie inférieur appelée D inférieur plongée dans un champs magnétique créé par une tension alternative
* La partie entre D supérieur et D inférieur où la tension alternative est appliquée

Voici une représentation sous forme de schéma:
![](Figure3.png)





Au temps $t_0=0$ **(ici le temps est sans unité)**, un proton est injecté au niveau du D inférieur sans vitesse. Le champ électrique à une direction et un sens toujours **similaires** à ceux de la vitesse du proton lorsqu’il est entre les D. Les instants $t_n$ où le proton sort des D (soit inférieur, soit supérieur, voir la figure) sont donnés, **sans unité**, par
$$t_n=n+\tau \sqrt{n}.$$
On choisit $\tau=0.5$. Le temps passé par le proton dans un D (supérieur ou inférieur) est toujours de $1$.
La trajectoire du proton est plane. On s’intéresse à sa vitesse. Ses composantes sont données, dans un repère et des unités appropriés, **en fonction de sa position dans le cyclotron:**
1) Pour $t$ variant de $t_n$ à $t_{n+1}-1$ c’est-à-dire quand il est entre les D :
$$
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$$
2) Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

### On importe les bibliothèques utiles

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Partie 1:
### On s'intéresse à la vitesse de la particule

#### On créer un tableau des instants où la particule sort de D supérieur où D inférieur

```python
#On définit les valeurs connues
t_0=0
tau=0.5

#On créer une liste que l'on remplira avec les différents instants et on insert la valeur initiale déjà connue
temps=[t_0]

n=51 
for i in range(1,n):
    values=i+tau*np.sqrt(i)
    temps.append(values)
print("Liste des instants où la particule sort des D:",temps)
```

#### On vérifie certaine valeur calculée avec les valeurs connues à l'avance

```python
print("La valeur est correcte pour t_1 ?" ,temps[1]==1.5)
print("La valeur est correcte pour t_4 ?" ,temps[4]==5)
print("La valeur est correcte pour t_9 ?" ,temps[9]==10.5)
print("La valeur est correcte pour t_16 ?" ,temps[16]==18)
print("La valeur est correcte pour t_25 ?" ,temps[25]==27.5)
```

#### On calcule les vitesses selon x et y $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$

```python
#On initialise les valeurs numérique

n_1=1
# Tableu de 50 valeur entre t_2-1 et t_2
t_2=np.linspace(temps[2]-1,temps[2],50) 
# On créer des listes vide des vitesses selon x et y que nous remplirons
v_x_1=[]
v_y_1=[]


for i in range(len(t_2)-1):
    v_x_1_values=(-1)**n_1*np.sqrt(n_1+1)*np.sin(np.pi*(t_2[i]-t_2[i+1]))
    v_y_1_values=-(-1)**n_1*np.sqrt(n_1+1)*np.cos(np.pi*(t_2[i]-t_2[i+1]))
    v_x_1.append(v_x_1_values)
    v_y_1.append(v_y_1_values)
```

#### On calcule les vitesses selon x et y $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_4−1$ à $t_4$

```python
n_3=3
# Tableu de 50 valeur entre t_4-1 et t_4
t_4=np.linspace(temps[4]-1,temps[4],50)
# On créer des listes vide des vitesses selon x et y que nous remplirons
v_x_3=[]
v_y_3=[]


for i in range(len(t_4)-1):
    v_x_3_values=(-1)**n_3*np.sqrt(n_3+1)*np.sin(np.pi*(t_4[i]-t_4[i+1]))
    v_y_3_values=-(-1)**n_3*np.sqrt(n_3+1)*np.cos(np.pi*(t_4[i]-t_4[i+1]))
    v_x_3.append(v_x_3_values)
    v_y_3.append(v_y_3_values)
```

#### On représente graphiquement ses vitesse pour t variant $t_2−1$ à $t_2$

```python
# Vitesse selon x
plt.plot(t_2[0:49],v_x_1)
plt.title("Evolution de la vitesse selon x dans le temps")
plt.ylabel("$v_x$")
plt.xlabel("temps entre $t_2$-1 et $t_2$")
```

```python
# Vitesse selon y
plt.plot(t_2[0:49],v_y_1)
plt.title("Evolution de la vitesse selon y dans le temps")
plt.ylabel("$v_y$")
plt.xlabel("temps entre $t_2$-1 et $t_2$")
```

#### On représente graphiquement ses vitesse pour t variant $t_4−1$ à $t_4$

```python
#Vitesse selon x
plt.plot(t_4[0:49],v_x_3)
plt.title("Evolution de la vitesse selon x dans le temps")
plt.ylabel("$v_x$")
plt.xlabel("temps entre $t_4$-1 et $t_4$")
```

```python
# Vitesse selon y
plt.plot(t_4[0:49],v_y_3)
plt.title("Evolution de la vitesse selon y dans le temps")
plt.ylabel("$v_y$")
plt.xlabel("temps entre $t_4$-1 et $t_4$")
```

#### On regarde les minima des vitesses selon x et y

```python
#Entre t_2-1 et t_2

#Pour x
indice_minimum_x_1=np.argmin(v_x_1)
print("Vitesse minimale selon x:",v_x_1[indice_minimum_x_1])
print("Temps correspondant:",t_2[indice_minimum_x_1])

#Pour y
indice_minimum_y_1=np.argmin(v_y_1)
print("Vitesse minimale selon y:",v_y_1[indice_minimum_y_1])
print("Temps correspondant:",t_2[indice_minimum_y_1])
```

On peut remarquer que la vitesse minimale selon x est 100 fois plus petite que celle selon y et que les temps pour atteindre ses vitesses sont presques identiques

```python
#Entre t_4-1 et t_4

#Pour x
indice_minimum_x_3=np.argmin(v_x_3)
print("Vitesse minimale selon x:",v_x_3[indice_minimum_x_3])
print("Temps correspondant:",t_4[indice_minimum_x_3])

#Pour y
indice_minimum_y_3=np.argmin(v_y_3)
print("Vitesse minimale selon y:",v_y_3[indice_minimum_y_3])
print("Temps correspondant:",t_4[indice_minimum_y_3])
```

On peut remarquer que la vitesse minimale selon x est 10 fois plus petite que celle selon y et que les temps pour atteindre ses vitesses sont presques identiques

## Partie 2:
### On s'interesse ici au module des vitesses

On représente $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$

```python
#On calculeles valeurs
n_5=5
# Tableu de 50 valeur entre t_0 et t_6
t_6=np.linspace(0,temps[6],50) 
# On créer des listes vide des vitesses selon x et y que nous remplirons
v_x_5=[]
v_y_5=[]


for i in range(len(t_6)-1):
    v_x_5_values=(-1)**n_1*np.sqrt(n_1+1)*np.sin(np.pi*(t_2[i]-t_2[i+1]))
    v_y_5_values=-(-1)**n_1*np.sqrt(n_1+1)*np.cos(np.pi*(t_2[i]-t_2[i+1]))
    v_x_5.append(v_x_5_values)
    v_y_5.append(v_y_5_values)
    
#On représente les vitesses
plt.subplot(2,1,2)
plt.plot(t_6[0:49],v_y_5)
plt.title("Evolution de la vitesse selon y dans le temps")
plt.ylabel("$v_y$")
plt.xlabel("temps entre $t_0$ et $t_6$")

plt.subplot(2,1,1)
plt.plot(t_6[0:49],v_x_5)
plt.title("Evolution de la vitesse selon x dans le temps")
plt.ylabel("$v_x$")
plt.xlabel("temps entre $t_0$ et $t_6$")

plt.subplots_adjust(top=2)
```

On remarque visuellement que la vitesse selon y est constante tandis que la vitesse selon x semble constante pour les temps inferieru à deux puis cyclique.

#### On représente le module de la vitesse $\sqrt{v_x^2+v_y^2}$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$. 

```python
# On transforme les listes en tableau pour pouvoir effectuer des opérations dessus
np_v_x_5=np.array(v_x_5)
np_v_y_y=np.array(v_y_5)

#On calcule le module de la vitesse
module_vitesse=np.sqrt(np_v_x_5**2+np_v_y_y**2)

# Représentation
plt.plot(t_6[0:49],module_vitesse)
plt.title("Evolution du module de la vitesse en fonction du temps")
plt.xlabel("Temps entre 0 et $t_6$")
plt.ylabel("Module de la vitesse")
```

La vitesse semble constante.

On peut stipuler qu'un cyclotron sert à faire tendre une particule vers une vitesse constante
