```python
# Question 6 : On g√©n√©ralise les calculs de vx et vy dans une fonction

def vx_ext(a,pas=n):
    """retourne la valeur de vx de la particule situ√© dans la partie D sup√©rieur/inf√©rieur en fonction de l'indice a du tableau t"""
    tech=np.linspace(t[a]-1,t[a],pas)
    vx=np.empty(pas)
    for i in range(pas):
        vx[i]=(-1)**(a-1)*np.sqrt(a)*np.sin(np.pi*(tech[i]-t[a]))
    return tech,vx

def vy_ext(a,pas=n):
    """retourne la valeur de vy de la particule situ√© dans la partie D sup√©rieur/inf√©rieur en fonction de l'indice a du tableau t"""
    tech=np.linspace(t[a]-1,t[a],pas)
    vy=np.empty(pas)
    for i in range(pas):
        vy[i]=(-1)**(a)*np.sqrt(a)*np.cos(np.pi*(tech[i]-t[a]))
    return tech,vy


def vy_int(a,pas=n):
    """retourne la valeur de vy de la particule situ√© dans la partie entre les D  en fonction de l'indice a du tableau t"""
    tech=np.linspace(t[a],t[a+1]-1,pas)
    vy=np.empty(pas)
    for i in range(pas):
        vy[i]=(-1)**a*(np.sqrt(a)+(tech[i]-t[a])/ùúè)
    return tech,vy

vx_int=np.zeros(n)
for i in range(6):
    plt.plot(vy_int(i)[0],vx_int,vx_ext(i+1)[0],vx_ext(i+1)[1])
plt.plot(vy_int(6)[0],vx_int)
plt.title("Graphique des vitesses selon x entre t=0 et t6")
plt.xlabel("Temps")
plt.ylabel("Vx")
plt.show()
#Toujours sans unit√©
```

```python
for i in range(6):
    plt.plot(vy_int(i)[0],vy_int(i)[1],vy_ext(i+1)[0],vy_ext(i+1)[1])
    
plt.plot(vy_int(6)[0],vy_int(6)[1])
plt.title("Graphique des vitesses selon y entre t=0 et t6")
plt.xlabel("Temps")
plt.ylabel("Vy")
plt.show()
#Toujours sans unit√©
```

On constate qu'effectivement vx et vy sont continues. Se sont des m√©langes de fonctions lin√©aires et sinuso√Ædales.

```python
for i in range(6):
    plt.plot(vy_int(i)[0],np.sqrt(vx_int**2+vy_int(i)[1]**2),vx_ext(i+1)[0],np.sqrt(vx_ext(i+1)[1]**2+vy_ext(i+1)[1]**2))
plt.plot(vy_int(6)[0],vy_int(6)[1])
plt.title("Graphique de la norme des vitesses selon x entre t=0 et t6")
plt.xlabel("Temps")
plt.ylabel("Norme de la vitesse")
plt.show()
#Toujours sans unit√©
```

On constate que la norme de la vitesse augmente lors de chaque passage entre les D et reste constante "dans" les D. Dans les D, la vitesse ne varie pas en norme mais varie en direction pour √™tre dirig√©e entre les D et pouvoir r√©augmenter sa norme. Un cyclotron permet donc d'augmenter la vitesse (et donc l'√©nergie cin√©tique) d'une particule en r√©p√©tant ce sch√©ma en boucle.
