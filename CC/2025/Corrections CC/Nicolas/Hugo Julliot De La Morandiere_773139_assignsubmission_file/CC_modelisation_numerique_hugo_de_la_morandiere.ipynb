### Partie 1
1 ) pour calculer toute les valeur de $t_n$ je commence par crée une liste d'entier qui poura nous servire tout au long des exercice puis j'asocie a un nouveau tableau $ t_n $ toute les valeur calculer a partire de la liste

```python
import matplotlib.pyplot as plt
import numpy as np

entier = np.arange(0,100,1)
tn = entier + 0.5*(entier**(1/2))

print(tn[:10])
```

ce resultat parait corecte mais comme l'énoncé nous donne certaine valeur de $ t_n $ vérifion ceci

on sait que 
$t_1=1.5$, $t_4=5$, $t_9=10.5$, $t_{16}=18$

```python
if tn[1] == 1.5 and tn[4] == 5 and tn[9] == 10.5 and tn[16] == 18:
    print("toute les valeur donner dans l'énoncé coresponde a celle ci ")
```

maintenant que nos valeurs sont vérifier on peut passer à la suite qui consiste a calculer certaine valeur de $v_x$ et $v_y$ en fonction du temps

commencons pour $t$ variant de $t_2−1$ à $t_2$ pour cela on va commencer par crée des fonctions pour $v_x$ et $v_y$

```python
# Pour t variant de t_{n+1}-1 à t_{n+1}
def Vx(n = 1, t = 1):
     return (-1)**n * (n + 1)**(1/2) * np.sin(np.pi*(t - tn[n+1]))

def Vy(n = 1, t = 1):
    return -(-1)**n * (n + 1)**(1/2) * np.cos(np.pi*(t - tn[n+1]))
```

maintenant que nous avons nos 2 fonctions nous devons faire varier la valeur de t pour calculer $v_x$ et $v_y$

on comencera par crée un nouveau tableau avec 100 valeur de t puis on les passera dans nos fonctions


```python
#pour t entre t2-1 et t2
Vx_2 = []
Vy_2 = []
Vx_4 = []
Vy_4 = []
val_t_2 = []
val_t_4 = []
for i in range(0,105,1):
    val_t_2.append(tn[2]-1+(1/100)*i)
    Vx_2.append(Vx(1, val_t_2[i]))
    Vy_2.append(Vy(1, val_t_2[i]))
    val_t_4.append(tn[4]-1+(1/100)*i)
    Vx_4.append(Vx(3, val_t_4[i]))
    Vy_4.append(Vy(3, val_t_4[i]))
```

on a bien toute nos valeur de $v_x$ et $v_y$ pour $t$ variant de $t_2−1$ à $t_2$ et pour $t$ variant de $t_4−1$ à $t_4$ 

on cherche maintenant a afficher $v_x$ et $v_y$ en fonctions de t

```python
plt.title('représentations de Vx et Vy en fonctions de t') 
plt.xlabel('t')
plt.ylabel('Vx et Vy')
plt.plot(val_t_2,Vx_2)
plt.plot(val_t_2,Vy_2)
```

on voie bien ici que selon x la vitesse part de 0 et augmente avant de ralentire alors que selon y elle est négative au départ et ne fait que augmenter jusqu'a avoir la meme vitesse mais cette fois positive

```python
plt.title('représentations de Vx et Vy en fonctions de t') 
plt.xlabel('t')
plt.ylabel('Vx et Vy')
plt.plot(val_t_4,Vx_4)
plt.plot(val_t_4,Vy_4)
```

on voie bien que les 2 graffes se resemble énormement car l'objet effectue a peut près le même mouvement.

pour trouver les maxima on regarde a quel moment dans la liste il y a un changement de signe de la dériver 


```python
def extrema(fonctions,a=1):
    e = 0
    for i in range(0,103,1):
        if fonctions[i+1] > fonctions[i] and a == 1:
            print("maxima en ", fonctions[i+1])
            a = 2
            e = e + 1
        elif fonctions[i+1] < fonctions[i] and a == 2:
            print("maxima en ", fonctions[i+1])
            a = 1
            e = e + 1
    if e == 0:
        print("aucun maxima pour cette fonction")

print("Pour Vx entre t2-1 et t2")
extrema(Vx_2,2)
print("")
print("Pour Vy entre t2-1 et t2")
extrema(Vy_2,1)
print("")
print("Pour Vx entre t4-1 et t4")
extrema(Vx_4,2)
print("")
print("Pour Vy entre t4-1 et t4")
extrema(Vy_4,1)
```

Il y a donc des maxima pour $V_x$ au centre des graphique et pour $V_y$ au extremiter de ceux ci

### Partie 2

dans cette questions nous allons respresenter $ V_x $ et $ V_y $ en fonction de $ t $ pour $ t $ variant de $ 0 $ à $ t_6 $ ou de $ 0 $ à $ 7.22 $

```python
Vx_6 = []
Vy_6 = []
val_t_6 = []
for i in range(0,723,1):
    val_t_6.append((1/100)*i)
    Vx_6.append(Vx(1, val_t_6[i]))
    Vy_6.append(Vy(1, val_t_6[i]))

plt.title('représentations de Vx et Vy en fonctions de t') 
plt.xlabel('t')
plt.ylabel('Vx et Vy')
plt.plot(val_t_6,Vx_6)
plt.plot(val_t_6,Vy_6)
```

on voie bien que chaque courbe se répete dans le temps se sont donc des fontions continue du temps.

on cherche maintenant à afficher le module de la vitesse en fonctions du temps.

```python
modu_t = []
for i in range(0,723,1):
    modu_t.append( (Vx_6[i]**2  +  Vy_6[i]**2)**(1/2))
plt.title('représentations du module de la vitesse en fonctions de t') 
plt.xlabel('t')
plt.ylabel('module de la vitesse')
plt.plot(val_t_6,modu_t)
```

le module de la vitesse ne varie pas au cours du temps, la vitesse est donc constante, on peut dire qu'un cyclotron permet de gardger une particule a une haute vitesse, et dans certaine cas de l'accélerer.
