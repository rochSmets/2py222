## Partie I

```python
#QUESTION 1
import numpy as np

liste_instants_sorties = []
for i in range(51): #Pour bien avoir les 50 éléments
    liste_instants_sorties.append(i+0.5*np.sqrt(i))
```

```python
#QUESTION 2

# Si les égalités sont bien vérifiées on aura un tableau avec seulement des True
verification = [liste_instants_sorties[1] == 1.5,liste_instants_sorties[4] == 5,liste_instants_sorties[9] == 10.5,liste_instants_sorties[16] == 18,liste_instants_sorties[25] == 27.5]
print(verification)
```

```python
#QUESTION 3

#On pose les bornes des intervalles
t2 = liste_instants_sorties[2]
t2_depart = t2-1

t4 = liste_instants_sorties[4]
t4_depart = t4-1

#On définit les fonctions v_xD et v_yD
def v_xD(n,t_fin,t):
    return((-1)**n*np.sqrt(n+1)*np.sin(np.pi*(t-t_fin)))

def v_yD(n,t_fin,t):
    return((-1)**(n)*np.sqrt(n+1)*np.cos(np.pi*(t-t_fin)))

#On crée les tableaux des t et on en déduit les vitesses

temps2 = np.linspace(t2_depart,t2,100)
temps4 = np.linspace(t4_depart,t4,100)

liste_vitesse_2_x = v_xD(1,t2,temps2)
liste_vitesse_4_x = v_xD(3,t4,temps4)

liste_vitesse_2_y = v_yD(1,t2,temps2)
liste_vitesse_4_y = v_yD(3,t4,temps4)
```

```python
#QUESTION 4

import matplotlib.pyplot as plt

plt.plot(temps2,liste_vitesse_2_x,label='v_x')
plt.plot(temps2,liste_vitesse_2_y,label = 'v_y')
plt.grid()
plt.title("Vitesse entre t_2-1 et t_2")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.legend()
plt.show()
```

```python
#QUESTION 4

plt.plot(temps4,liste_vitesse_4_x,label='v_x')
plt.plot(temps4,liste_vitesse_4_y,label = 'v_y')
plt.grid()
plt.title("Vitesse entre t_4-1 et t_4")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.legend()
plt.show()
```

```python
#QUESTION 5

def maxima_fonction_vitesse(f,a,b,n_points,n): #fonction,point de départ de l'intervalle, point de fin, nombre de points d'évalutation, parametre n de la fonction
    """fonction qui va calculer les valeurs des vitesses pour un certain nombre de points données afin de déterminer une valeur approchée du max"""
    points_evaluation = np.linspace(a,b,n_points)
    max = -10000
    for x_n in points_evaluation:
        if f(n,b,x_n)>max:
            max = f(n,b,x_n) #reassigne le max si il est dépassé
    return max

max_vx_2 = maxima_fonction_vitesse(v_xD,t2_depart,t2,100,1)
max_vy_2 = maxima_fonction_vitesse(v_yD,t2_depart,t2,100,1) 

max_vx_4 = maxima_fonction_vitesse(v_xD,t4_depart,t4,100,3) 
max_vy_4 = maxima_fonction_vitesse(v_yD,t4_depart,t4,100,3) 

print("Les max des vitesses x et y, 2 et 4, respectivemen, sont:")
print(max_vx_2)
print(max_vy_2)
print(max_vx_4)
print(max_vy_4)
```

Si on néglige l'erreur induite par le fait que l'on ne calcuule qu'une valeur approchée, on peut considérer que les maximas pour vx et vy sont égaux.

```python
#QUESTION 6
#On définit les fonctions v_x et v_y correspondant aux vitesses entre les D

def v_x(n,t_depart,t):
    return 0*t

def v_y(n,t_depart,t):
    return((-1)**n*(np.sqrt(n)+2*(t-t_depart)))

#On renote les autres fonctions pour plus de clarté
def v_xD(n,t_fin,t):
    return((-1)**n*np.sqrt(n+1)*np.sin(np.pi*(t-t_fin)))

def v_yD(n,t_fin,t):
    return((-1)**(n)*np.sqrt(n+1)*np.cos(np.pi*(t-t_fin)))

liste_t_jusqua_6 = liste_instants_sorties[:7] 
liste_t_moins1 = [i-1 for i in liste_t_jusqua_6[1:]]

for i in range(6):
    t_dep = liste_t_moins1[i]
    t_fin = liste_t_jusqua_6[i+1]
    
    liste_temps = np.linspace(t_dep,t_fin,100)
    plt.plot(liste_temps,v_xD(i,t_dep,liste_temps),'r',label = 'v_x dans D')
    plt.plot(liste_temps,v_yD(i,t_dep,liste_temps),'r', label = 'v_y dans D')
        
for i in range(6):
    t_dep = liste_t_jusqua_6[i]
    t_fin = liste_t_moins1[i]
    liste_temps = np.linspace(t_dep,t_fin,100)
    
    plt.plot(liste_temps,v_x(i,t_dep,liste_temps),'b-',label='v_x entre D')
    plt.plot(liste_temps,v_y(i,t_dep,liste_temps),'b-', label='v_y entre D')
    
plt.title("Vx et Vy en fonction du temps")
plt.ylabel('Vitesses')
plt.xlabel('Temps')
plt.grid()
plt.show()
```

On pouvait s'attendre à ce type de graphe en regardant la forme de la trajectoire de la particule dans le cyclotron. La trajectoire est continue donc les vitesses le seront aussi et les parties linéaires ou constantes par morceaux viennent des trajectoires rectilignes entre les D inférieurs et supérieurs.

```python
#QUESTION 7

def norme(a,b):
    return(np.sqrt(a**2+b**2))

for i in range(6):

    t_dep = liste_t_moins1[i]
    t_fin = liste_t_jusqua_6[i+1]
    liste_temps = np.linspace(t_dep,t_fin,100)
    v_x_dans_D = v_xD(i,t_dep,liste_temps)
    v_y_dans_D = v_yD(i,t_dep,liste_temps)
    plt.plot(liste_temps,norme(v_x_dans_D,v_y_dans_D),'r',label = 'v_x dans D')
    
for i in range(6):
    t_dep = liste_t_jusqua_6[i]
    t_fin = liste_t_moins1[i]
    liste_temps = np.linspace(t_dep,t_fin,100)
    v_x_hors_D = v_x(i,t_dep,liste_temps)
    v_y_hors_D = v_y(i,t_dep,liste_temps)
    
    plt.plot(liste_temps,norme(v_x_hors_D,v_y_hors_D),'b-',label='v_x entre D')
    
plt.title("Norme de la vitesse en fonction du temps")
plt.ylabel('Norme de la vitesse')
plt.xlabel('Temps')
plt.grid()
plt.show()
```

La norme de la vitesse est constante dans les zones D et croit de manière linéaire entre les zones D. On a donc une accélération positive et constante par morceaux. Le cyclotron permet donc d'accélérer une particule. En effet, d'apres l'expression de la force de Lorentz, l'effet d'un champ magnétique sur une particule chargé change la direction de sa vitesse mais pas sa norme. Ainsi le cyclotron permettra d'accelerer la particule autant de fois que nécéssaires dans une zone relativement petite.

```python
#QUESTION 8


#V_x est continue et toujours nul donc la norme revient juste a la valeur absolue de v_y, de meme v_y est continue donc on prendra l'expression de vy dans D

t_n_moins_n = [liste_instants_sorties[i]-i for i in range(51)]
def v_y_tn(n):
    return((-1)**n*(np.sqrt(n)))

vitesse = []
for i in range(51):
    vitesse.append(np.abs(v_y_tn(i)))
    
plt.plot(t_n_moins_n,vitesse)
plt.title("Norme de la vitesse en fonction de n")
plt.ylabel('Norme de la vitesse')
plt.xlabel('$t_n - n$')
plt.grid()
plt.show()
```

On remarque que la relation entre $\sqrt{v_x(t_n)^2+v_y(t_n)^2}$ et $t_n-n$ est linéaire. La particule accélère donc d'une quantité constante à chaque fois qu'elle passe entre les zones D. Encore une fois avec la force de Lorentz, on voit que c'est le champ éléctrique généré par les tensions entre les D qui va accélérer la particule chargée.
