```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
to=0.5
N=50
t=[]
for i in range(N+1):#boucle qui associe une valeur de t pour un n donnée
    t.append(i+to*np.sqrt(i))
np_tn=np.array(t)
dic_t={'tn':np_tn}
df_t = pd.DataFrame(data = dic_t)
print(df_t)
```
