## Partie I

1. Créer un tableau contenant les instants de sortie des boucles "D" : $t_0$, $t_1$, ... jusqu'à $t_{50}$ inclus. 
2. Vérifier que $t_1=1.5$, $t_4=5$, $t_9=10.5$, $t_{16}=18$, et $t_{25}=27.5$ avec des opérateurs booléens.
3. Calculer les valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$. Vous utiliserez un échantillonage suffisant pour le temps $t$, dans un tableau distinct de celui de la question précédente. Faire de même pour $t$ variant de $t_4-1$ à $t_4$. 
4. Représenter dans un graphique $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $t_2-1$ a $t_2$. Faire de même pour $t$ variant de $t_4-1$ à $t_4$.
5. Déterminer des valeurs approchées des maxima de $v_x$ et $v_y$ pour $t$ variant de $t_2-1$ à $t_2$, ainsi que les instants $t$ correspondants. Un maximum peut être en bord d'intervalle. Faire de même pour $t$ variant de $t_4-1$ à $t_4$. Commenter.

```python
n = 50    # n max du temps
# Paramètres initiaux
tau = 0.5   
tn = np.empty(n+1)   # initialisation du tableau de sortie de boucles
tn[0] = 0

for i in range(len(tn)) :
    tn[i] = i + tau*np.sqrt(i) 

# Vérification
print(f"t_1=1.5 est {tn[1] == 1.5}")
    
print(f"t_4=5 est {tn[4] == 5}")

print(f"t_9=10.5 est {tn[9] == 10.5}")

print(f"t_16=18 est {tn[16] == 18}")

print(f"t_25=27.5 est {tn[25] == 27.5}")
```

```python
N = 100   # longueur de l'échantillonage

# Initialisation des paramètres
t2 = np.linspace(tn[2]-1, tn[2], N)
t4 = np.linspace(tn[4]-1, tn[4], N)

vx_2 = np.empty(t2.shape)
vy_2 = np.empty(t2.shape)

vx_4 = np.empty(t4.shape)
vy_4 = np.empty(t4.shape)

epsilon = -1

# Calcul des vitesses
# Comme on est dans le cas t_(n+1)-1 à t_(n+1), notre n vaut respectivement 1 et 3 donc
# Epsilon_n = -1 dans les deux cas

for i,t in enumerate(t2) :
    n = 1
    vx_2[i] = (epsilon)**n*np.sqrt(n+1)*np.sin(np.pi*(t-tn[n+1]))
    vy_2[i] = -(epsilon)**n*np.sqrt(n+1)*np.cos(np.pi*(t-tn[n+1]))
    
for i,t in enumerate(t4) :
    n = 3
    vx_4[i] = (epsilon)**n*np.sqrt(n+1)*np.sin(np.pi*(t-tn[n+1]))
    vy_4[i] = -(epsilon)**n*np.sqrt(n+1)*np.cos(np.pi*(t-tn[n+1]))

# Affichage des graphes
plt.plot(t2,vx_2, label="Vitesse selon x")
plt.plot(t2,vy_2, label="Vitesse selon y")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.legend()
plt.title(f"Vitesses selon les axes en fonction du temps entre $t_2 -1$ et $t_2$")
plt.show()

plt.plot(t4,vx_4, label="Vitesse selon x")
plt.plot(t4,vy_4, label="Vitesse selon y")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.legend()
plt.title(f"Vitesses selon les axes en fonction du temps entre $t_4 -1$ et $t_4$")
plt.show()
```

```python
# Vitesses maxima approchées et leurs instants t correspondant
vx_max2 = max(vx_2)
t_vxmax2 = t2[np.argmin(vx_2)]
vy_max2 = max(vy_2)
t_vymax2 = t2[np.argmin(vy_2)]

vx_max4 = max(vx_4)
t_vxmax4 = t4[np.argmin(vx_4)]
vy_max4 = max(vy_4)
t_vymax4 = t4[np.argmin(vy_4)]

print(f"Entre t_2-1 et t_2 Les valeurs des vitesses maximales sont : \nPour vx = {vx_max2:.3f} à t = {t_vxmax2:.3f},\nPour vy = {vy_max2:.3f} à t = {t_vymax2:.3f}.")
print()
print(f"Entre t_4-1 et t_4 Les valeurs des vitesses maximales sont : \nPour vx = {vx_max4:.3f} à t = {t_vxmax4:.3f},\nPour vy = {vy_max4:.3f} à t = {t_vymax4:.3f}.")
```

On remarque que les vitesses selon x et y atteignent le même maxima à une seconde d'écart dans les deux intervalles de temps choisi. On remarque aussi que les vitesses maximales ont augmenté entre $t_2$ et $t_4$ ce qui est cohérent pour un cyclotron qui accélère les particules.

## Partie II
6. Représenter $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$. On utilisera une couleur unique pour la courbe par morceaux. On remarque que $v_x$ et $v_y$ sont des fonctions continues du temps $t$. Commenter.
7. Représenter le module de la vitesse $\sqrt{v_x^2+v_y^2}$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$. On utilisera une couleur unique pour la courbe par morceaux. Commenter. A quoi sert un cyclotron ?
8. Pour $n$ variant de $0$ à $50$, représenter le module de la vitesse $\sqrt{v_x(t_n)^2+v_y(t_n)^2}$ en fonction de $t_n-n$, qui est le temps total passé par le proton entre les D à l'instant $t_n$. Commenter.

```python
n_max = 6   # On définit jusqu'à quelle t nous souhaitons étudier la situation
N = 100   # longueur des échantillonages

# On définit des fonctions permettant de faciliter le calcul
def vitesse_interD(t, n) :
    """
    Calcul les composantes des vitesses x et y lorsque le proton est entre les dés
    Prends une liste de temps entre t_n et t_n+1 -1 et la valeur n
    """
    vx = np.zeros(t.shape)
    vy = np.empty(t.shape)
    for i, a in enumerate(t) :
        vy[i] = epsilon**n*(np.sqrt(n)+(a-tn[n])/tau)
    return vx, vy
    
def vitesse_dansD(t, n) :
    """
    Calcul les composantes des vitesses x et y lorsque le proton est dans les dés
    Prends une liste de temps entre t_n+1 -1 et t_n+1 et la valeur n
    """
    vx = np.empty(t.shape)
    vy = np.empty(t.shape)
    for i, a in enumerate(t) :
        vx[i] = (epsilon)**n*np.sqrt(n+1)*np.sin(np.pi*(a-tn[n+1]))
        vy[i] = -(epsilon)**n*np.sqrt(n+1)*np.cos(np.pi*(a-tn[n+1]))
    return vx, vy

def temps(n) :
    """
    Crée deux tableaux de temps correspondants aux deux sections de la trajectoire
    """
    t_s1 = np.linspace(tn[n], tn[n+1]-1, N)
    t_s2 = np.linspace(tn[n+1]-1, tn[n+1], N)
    return t_s1, t_s2

T = np.array([temps(r) for r in range(n_max)])  # Notre tableau de toutes le valeurs de temps

# Calcul des vitesses
vx_6 = np.array([[vitesse_interD(T[i,0],i)[0], vitesse_dansD(T[i,1],i)[0]] for i in range(n_max)])
vy_6 = np.array([[vitesse_interD(T[i,0],i)[1], vitesse_dansD(T[i,1],i)[1]] for i in range(n_max)])

# Affichage du graphe
for i in range(n_max) :
    plt.plot(T[i,0],vx_6[i,0], color = "blue")
    plt.plot(T[i,0],vy_6[i,0], color = "red")
    plt.plot(T[i,1],vx_6[i,1], color = "blue")
    plt.plot(T[i,1],vy_6[i,1], color = "red")
plt.title("Composante de la vitesse selon x et y en fonction du temps")
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.show()
```

En rouge, nous avons la vitesse selon x et en bleu, nous avons la vitesse selon y. Nous remarquons effectivement que la vitesse selon les deux axes est continue et qu'elle est périodique avec une amplitude qui augmente avec le temps.

```python
# Calcul de la norme de la vitesse
v = np.array(np.sqrt(vx_6**2 + vy_6**2))

# Affichage du graphe
for i in range(n_max) :
    plt.plot(T[i,0],v[i,0], color = "blue")
    plt.plot(T[i,1],v[i,1], color = "blue")
    
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.title("Vitesse en fonction du temps")
plt.show()
```

On voit que la vitesse du proton en norme augmente avec le temps, on en déduit qu'un cyclotron sert à accélérer les protons.

```python
# On définit notre nouveau n max
n_max = 50

# On recalcule le temps et on calcule également le temps total passé passé par le proton entre les D
T = np.array([temps(r) for r in range(n_max)])  # Notre tableau de toutes le valeurs de temps
T50 = np.array([np.array(temps(r))-r for r in range(n_max)])

# Recalcule des vitesses
vx_50 = np.array([[vitesse_interD(T[r,0],r)[0], vitesse_dansD(T[r,1],i)[0]] for r in range(n_max)])
vy_50 = np.array([[vitesse_interD(T[r,0],r)[1], vitesse_dansD(T[r,1],i)[1]] for r in range(n_max)])
v50 = np.array(np.sqrt(vx_50**2 + vy_50**2))

# Affichage du graphe
for i in range(n_max) :
    plt.plot(T50[i,0],v50[i,0], color = "blue")
    plt.plot(T50[i,1],v50[i,1], color = "blue")
    
plt.xlabel("Temps")
plt.ylabel("Vitesse")
plt.title("Vitesse en fonction du temps")
plt.show()
```
