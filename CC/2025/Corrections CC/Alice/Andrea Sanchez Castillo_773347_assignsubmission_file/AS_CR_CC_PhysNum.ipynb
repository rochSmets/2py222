### Exercice 1

```python
def tn(n):
    """Instants de sortie du proton"""
    return n + tau*np.sqrt(n)
```

Con calcule un tableau avec les instants de sortie des boucles "D" $t_n$:

```python
n =50
t_n = np.arange(0.,50.,1)

for i in range(n):
    t_n[i] = tn(i)

print("Tableau des instants t_n:\n",t_n)
```

### Exercice 3

```python
def epsilon (n):
    """Fonction qu"""
    return (-1)**n
```

```python
def vx (n,t):
    return epsilon(n)*np.sqrt(n+1)*np.sin(np.pi*(t-tn(n+1)))
```

```python
def vy (n,t):
    return -epsilon(n)*np.sqrt(n+1)*np.cos(np.pi*(t-tn(n+1)))
```

```python
def vy_D (n,t):
    return epsilon(n)*(np.sqrt(n) + (t-tn(n+1))/tau)
```

```python
n2 = 2
n4 = 4
```

```python
# Plage de temps
t2 =  np.linspace(t_n[2]-1,t_n[2],50) 
t4 =  np.linspace(t_n[4]-1,t_n[4],50)

# Calcul de la vitesse pour t2
v_x2 = vx(n2,t2)
#print(v_x2)
v_y2 = vy(n2,t2)
#print(v_y2)

# Calcul de la vitesse pour t4
v_x4 = vx(n4,t4)
#print(v_x4)
v_y4 = vy(n4,t4)
#print(v_y4)
```

```python
vitesse_x = []
vitesse_y = []
time = []

for i in range(0,6):
    tempsD = np.linspace(t_n[i],t_n[i+1]-1,50)
    temps = np.linspace(t_n[i+1]-1,t_n[i+1],50)
    time.append(tempsD)
    time.append(temps)
    
    v_xD = np.zeros(50)
    v_yD = vy_D(i,tempsD)
                   
    v_x= vx(i,temps)
    v_y = vy(i,temps)
    
    vitesse_y.append(v_yD)
    vitesse_y.append(v_y)
    vitesse_x.append(v_xD)
    vitesse_x.append(v_x)

vitesse_x = np.array(vitesse_x)
vitesse_y = np.array(vitesse_y)
time = np.array(time)

print(time.shape)
```

```python
# plot de la vitesse selon x
for i in range(0,12,2):
    plt.plot(time[i], vitesse_x[i], color ="blue")
    plt.plot(time[i+1], vitesse_x[i+1], color ="orange")

plt.title("Vitesse selon x")
plt.xlabel("temps (sans unité)") ; plt.ylabel("vitesse (sans unité)")
plt.grid()
```

```python
# plot de la vitesse selon y
for i in range(0,12,2):    
    plt.plot(time[i], vitesse_y[i],color ="orange")
    plt.plot(time[i+1], vitesse_y[i+1], color ="blue")

plt.title("Vitesse selon y")
plt.xlabel("temps (sans unité)") ; plt.ylabel("vitesse (sans unité)")
plt.grid()
```
