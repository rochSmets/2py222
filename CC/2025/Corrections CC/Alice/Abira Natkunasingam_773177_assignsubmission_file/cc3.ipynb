# Introduction

Dans ce projet nous allons étudier la vitesse d'un proton dans un cyclotron en fonction de sa position dans celui-ci (c'est à dire selon si il est entre les D ou s'il est dans un D supérieur ou inférieur).

## Partie 1

```python
# Import des modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Création d'un tableau avec les instants de sortie des boucles "D" 

# Introduction des variables
tau = 0.5
n = 50

# Tableau des instants 
t = np.zeros(n) # Création d'un tableau vide
for i in range (n) :
    t[i] = tau*(i)**(1/2) + i
    
print (t)
    
```

```python
# Vérification des valeurs de t1, t4, t16, et t25

print (t[1] == 1.5)
print (t[4] == 5)
print (t[16] == 18)
print (t[25] == 27.5)
```

### Étude théorique
Pour les instants demandées, le proton se trouve dans D supérieur ou inférieur. Ainsi sa vitesse se note :


$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad 
$$

$$
v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$
    

```python
# Calcul de la vitesse Vx et Vy
# Nouvel échantillonage du temps
t_2 = np.linspace(t[2] -1, t[2], 20)

# Introduction des variables 
n =1
epsilon = (-1)**n

# Équations des vitesses
Vx = np.zeros(len(t_2))
Vy = np.zeros(len(t_2))
for i in range (len(t_2)-1) :
    Vx[i] =  epsilon*(n+1)**(1/2)*np.sin(np.pi*(t_2[i]-t[2]))
    Vy[i] =  (-1)* epsilon *(n+1)**(1/2)*np.cos(np.pi*(t_2[i]-t[2]))
print ("Les valeurs de Vx sont: " , Vx)
print ("Les valeurs de Vy sont: " , Vy)  
```

```python
# Tracé de Vx et Vy en foncton de t_2

plt.plot(t_2[:-1], Vx[:-1],  label = 'Vitesse selon x')
plt.plot(t_2[:-1], Vy[:-1],  label = 'Vitesse selon y', color = 'r')
plt.title ( "La vitesse en fonction du temps lorsque le proton est dans D supérieur ou inférieur" )
plt.ylabel(" Vitesse (m) ") # Le temps est sans unité
plt.xlabel(" Temps ")
plt.grid()
plt.legend()
plt.show()
```

```python
# Déterminer les valeurs approchées des maxima de Vx et Vy pour t_2

Vxmax  =np.max(Vx)
Vymax = np.max(Vy)

print ("La valeur de Vx est :" ,Vxmax)
print ("La valeur de Vy est :", Vymax)

for i in range (len(t_2)-1):
    if Vx[i] == Vxmax :
        print ("La valeur de t en Vxmax est :", t_2[i])
    if Vy[i] == Vymax :
        print("La valeur de t en Vymax est :", t_2[i])
```

```python
# Calcul de la vitesse Vx et Vy
# Nouvel échantillonage du temps
t_4 = np.linspace(t[4] -1, t[4], 50)

# Introduction des variables 
n =3
epsilon = (-1)**n

# Équations des vitesses
Vx1 = np.zeros(len(t_4))
Vy1 = np.zeros(len(t_4))
for i in range (len(t_4)-1) :
    Vx1[i] =  epsilon*(n+1)**(1/2)*np.sin(np.pi*(t_4[i]-t[4]))
    Vy1[i] =  (-1)* epsilon *(n+1)**(1/2)*np.cos(np.pi*(t_4[i]-t[4]))
print ("Les valeurs de Vx sont: " , Vx1)
print ("Les valeurs de Vy sont: " , Vy1) 
```

```python
# Tracé de Vx et Vy en foncton de t_4

plt.plot(t_4[:-1], Vx1[:-1],  label = 'Vitesse selon x')
plt.plot(t_4[:-1], Vy1[:-1],  label = 'Vitesse selon y', color = 'r')
plt.title ( "La vitesse en fonction du temps lorsque le proton est dans D supérieur ou inférieur" )
plt.ylabel(" Vitesse (m) ") # Le temps est sans unité
plt.xlabel(" Temps ")
plt.grid()
plt.legend()
plt.show()
```

```python
# Déterminer les valeurs approchées des maxima de Vx et Vy pour t_4

Vx1max  =np.max(Vx1)
Vy1max = np.max(Vy1)

print ("La valeur de Vx est :", Vx1max)
print("La valeur de Vy est :", Vy1max)

for i in range (len(t_4)-1):
    if Vx1[i] == Vx1max :
        print ("La valeur de t en Vxmax est :" , t_4[i])
    if Vy1[i] == Vy1max :
        print("La valeur de t en Vymax est :", t_4[i])
```

## Partie II

Dans cette partie il faut disitnguer les diffrentes position du proton.

1) Pour $t$ variant de $t_n$ à $t_{n+1}-1$ c’est-à-dire quand il est entre les D :
$$
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$$
2) Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$

Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

```python
# Étude de la vitesse dans le cas 1

# Calcul de la vitesse Vx et Vy
# Nouvel échantillonage du temps

t_6 = np.linspace(0, t[6], 50)

# Équations des vitesses
Vx2 = np.zeros(len(t_6)) * 0
Vy2 = np.zeros(len(t_6))

for i in range (6) :
    for k in range (len(t_6)):
        Vy2[k] = (-1)**(i)*(i**(1/2) + (t_6[k]-t[i])/tau)

print ("Les valeurs de Vx sont: " , Vx2)
print ("Les valeurs de Vy sont: " , Vy2) 
```

```python
# Tracé de Vx et Vy en foncton de t_6 dans le cas 1

plt.plot(t_6[:-1], Vx2[:-1],  label = 'Vitesse selon x')
plt.plot(t_6[:-1], Vy2[:-1],  label = 'Vitesse selon y', color = 'r')
plt.title ( "La vitesse en fonction du temps lorsque le proton est entre les D " )
plt.ylabel(" Vitesse (m) ") # Le temps est sans unité
plt.xlabel(" Temps ")
plt.grid()
plt.legend()
plt.show()
```

# Équation théorique du module de la vitesse
La norme de la vitesse se note :
$\sqrt{v_x^2+v_y^2}$

```python
# Équation du module dans le cas 1

module = (Vx2**2 +Vy2**2) **(1/2)
print (module)
```

```python
# Représentation du module de la vitesse en fonction de t dans le cas 1

plt.plot(t_6[:-1], module[:-1],  color = 'g')
plt.title ( "La module de la vitesse en fonction du temps lorsque le proton est entre les D " )
plt.ylabel(" Module de la vitesse ") 
plt.xlabel(" Temps ")
plt.grid()
plt.show()
```

```python
# Étude de la vitesse dans le cas 2

Vx3 = np.zeros(len(t_6)) 
Vy3 = np.zeros(len(t_6))

for i in range (6) :
    for k in range (len(t_6)):
        Vx3[k] =  (-1)**(i)*(n+1)**(1/2)*np.sin(np.pi*(t_6[k]-t[i]))
        Vy3[k] =  (-1)* (-1)**(i) *(n+1)**(1/2)*np.cos(np.pi*(t_6[k]-t[i]))
print ("Les valeurs de Vx sont: " , Vx3)
print ("Les valeurs de Vy sont: " , Vy3) 
```

```python
# Tracé de Vx et Vy en foncton de t_6 dans le cas 2

plt.plot(t_6[:-1], Vx3[:-1],  label = 'Vitesse selon x')
plt.plot(t_6[:-1], Vy3[:-1],  label = 'Vitesse selon y', color = 'r')
plt.title ( "La vitesse en fonction du temps lorsque le proton est dans D supérieur ou inférieur") 
plt.ylabel(" Vitesse (m) ") # Le temps est sans unité
plt.xlabel(" Temps ")
plt.grid()
plt.legend()
plt.show()
```

```python
# Équation du module dans le cas 2

module1 = (Vx3**2 +Vy3**2) **(1/2)
print (module1)
```

```python
# Représentation du module de la vitesse en fonction de t dans le cas 2

plt.plot(t_6[:-1], module1[:-1],  color = 'g')
plt.title ( "La module de la vitesse en fonction du temps lorsque le proton est entre les D " )
plt.ylabel(" Module de la vitesse ")
plt.xlabel(" Temps ")
plt.grid()
plt.show()
```

### Obervations 

Lorsque le proton se trouve entre les D le module de la vitesse varie contrairement à lorsqu'il est dans un D supérieur ou inérieur où alors là son module est constant c'est à dire que sa vitesse est constante.

Ainsi un cyclotron sert donc à modifier la vitesse du proton selon sa position.
