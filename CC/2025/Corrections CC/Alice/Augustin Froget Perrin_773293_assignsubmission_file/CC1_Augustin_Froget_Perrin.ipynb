## partie I

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1/
tau = 0.5
T = np.zeros(51) #tableau des temps initialisation

for i in range(len(T)) : #boucle calculant tout les tn
    T[i] = i + tau*np.sqrt(i)
```

```python
# 2/ Vérifications de quelque valeurs de t
#cette commande renvoie True si la valeur de tn est égale à la valeur recherchée
print(T[1] == 1.5)
print(T[4] == 5)
print(T[9] == 10.5)
print(T[16] == 18)
print(T[25] == 27.5)
```

```python
T
```

```python
#3/ vitesse

# fonctions vitesses pour t allant de tn -1 à tn
def vx(t) :
    return (-1)**n * np.sqrt(n+1) * np.sin(np.pi*(t - T[n+1]))
def vy(t) :
    return -(-1)**n * np.sqrt(n+1) * np.cos(np.pi*(t - T[n+1]))
```

```python
# de t2 - 1 à t2
n = 2
#intervalle de temps entre les deux bornes considérés
t2 = np.linspace(T[2]-1, T[2], 100)

#calcul des vx
v2x = np.zeros(len(t2))

for i in range(len(t2)) :
    v2x[i] = vx(t2[i])
    
#calcul des vy
v2y = np.zeros(len(t2))

for i in range(len(t2)) :
    v2y[i] = vy(t2[i])
```

```python
# de t4-1 à t4
n = 4
#intervalle de temps entre les deux bornes considérés
t4 = np.linspace(T[4]-1, T[4], 100)

#calcul des vx
v4x = np.zeros(len(t4))

for i in range(len(t4)) :
    v4x[i] = vx(t4[i])
    
#calcul des vy
v4y = np.zeros(len(t4))

for i in range(len(t4)) :
    v4y[i] = vy(t4[i])
```

```python
#print('Tableau des vitesses de t2 -1 à t2', v2x, 'selon x', v2y, 'selon y, en m/s')
```

```python
#4/ Tableau vitesse en fonction du temps
#5/ Point des maximas de vx et vy

#de t2 -1 à t2

plt.plot(t2, v2x, label = 'vitesse selon x')
plt.plot(t2, v2y, label = 'vitesse selon y')
plt.grid()

plt.xlabel('temps ')
plt.ylabel('vitesses ')
plt.title('Vitesses selon x et y en fonction du temps dans le D entre t2 - 1 et t2')

#point des vitesses maximales 
plt.plot(t2[65], np.max(v2x), label = 'maxima de vx', marker = 'o', linewidth = 0)
plt.plot(t2[99], np.max(v2y), label = 'maxima de vy', marker = 'o', linewidth = 0)

plt.plot()
plt.legend()
```

```python
# de t4 -1 à t4

plt.plot(t4, v4x, label = 'vitesse selon x')
plt.plot(t4, v4y, label = 'vitesse selon y')
plt.grid()

plt.xlabel('temps ')
plt.ylabel('vitesses ')
plt.title('Vitesses selon x et y en fonction du temps dans le D entre t4 - 1 et t4')

#point de vitesses maximales
plt.plot(t4[59], np.max(v4x), label = 'maxima de vx', marker = 'o', linewidth = 0)
plt.plot(t4[99], np.max(v4y), label = 'maxima de vy', marker = 'o', linewidth = 0)
plt.legend()
```

```python
#5/ Vitesses maximales

# t2 - 1 à t2
print('la vitesse maximale selon x est : ',np.max(v2x),' à t = ', t2[65])
print('la vitesse maximale selon y est : ',np.max(v2y),'à t = ', t2[99])
```

```python
#t4 - 1 à t4
print('la vitesse maximale selon x est : ',np.max(v4x),'à t = ', t4[59])
print('la vitesse maximale selon y est : ',np.max(v4y),'à t = ', t4[99])
```

on remarque que les valeurs des vitesses maximales augmentent plus n augmente, ce qui est logique dans un accélérateur de particules

```python
#6/ Vitesses 

#initialisation d'un tableau à n colonnes, qui pour chaque colonne contient toute les valeurs de vx (pour le n correspondant)
VX = np.zeros((6, 100))
nmax = 6
def VitesseX (nmax) :
    n = 0
    for i in range(nmax) : #boucle pour chaque n
        tn = np.linspace(T[n]-1, T[n], 50)
        tN = np.linspace(T[n], T[n+1], 50)
        for j in range(len(tn)) : #calcule pour chaque n la vitesse dans le D
            VX[i][j] = vx(tn[j])
        for k in range(len(tN)) : #calcule pour chaque n la vitesse entre les D
            VX[i][50+k] = 0
        n+=1
    return VX
```

```python

#initialisation d'un tableau à n colonnes, qui pour chaque colonne contient toute les valeurs de vx (pour le n correspondant)
VY = np.zeros((6, 100))
nmax = 6

#fonction qui calcule la vitesse selon Y entre deux D
def vY (t) :
    return (-1)**n * (np.sqrt(n)+ (t - T[n])/0.5)


def VitesseY (nmax) :
    n = 0
    for i in range(nmax) : #boucle pour chaque n
        tn = np.linspace(T[n]-1, T[n], 50)
        tN = np.linspace(T[n], T[n+1], 50)
        for j in range(len(tn)) : #calcule pour chaque n la vitesse dans le D
            VY[i][j] = vy(tn[j])
        for k in range(len(tN)) : #calcule pour chaque n la vitesse entre les D
            VY[i][50+k] = vY(tn[k])
        n+=1
    return VY
```

```python
plt.plot(np.linspace(T[0], T[1], 100), VitesseX(6)[0])
plt.plot(np.linspace(T[1], T[2], 100), VitesseX(6)[1])
plt.plot(np.linspace(T[2], T[3], 100), VitesseX(6)[2])
plt.plot(np.linspace(T[3], T[4], 100), VitesseX(6)[3])
plt.plot(np.linspace(T[4], T[5], 100), VitesseX(6)[4])
plt.plot(np.linspace(T[5], T[6], 100), VitesseX(6)[5])
plt.grid()
plt.title('Vitesse en fonction du temps')
```

```python
# 7/ Module de vitesse

#initialisation d'un tableau des modules
M = np.zeros((6, 100))

X = VitesseX(6)
Y = VitesseY(6)

#calcule le module pour chaque valeur de vitesse
for j in range(6) :
    for i in range(100) :
        M[j][i] = np.sqrt(X[j][i]**2 + Y[j][i]**2)
```
