- 1. Cr√©er un tableau contenant les instants de sortie des boucles "D" :  ùë°0,  ùë°1, ... jusqu'√†  ùë°50 inclus.
On a $t_n=n+ùúè‚àön$.

```python
#code pour calculer les temps 1 √† 50 :
import numpy as np

n = 50
to = 0.5

# D√©finition de la fonction t
def t(n):
    ''' Retourne l'√©l√©ment tn de la suite '''
    return n + to*np.sqrt(n)

# Liste des √©l√©ments
liste_elements = [t(k) for k in range(n)]

#appel de la fonction
t(n) 

#v√©rification qu'il y'a bien 50 √©l√©ments
print(len(liste_elements))

#cr√©ation du tableaux des valeurs du temps avec la liste d'√©l√©ments
temps = np.array(liste_elements)

print(temps)
```

- 3. Calculer les valeurs de  $ùë£_ùë•$ et  $ùë£_ùë¶$ en fonction du temps pour  ùë° variant de  $ùë°_2‚àí1$ √†  $ùë°_2$. Vous utiliserez un √©chantillonage suffisant pour le temps  ùë°, dans un tableau distinct de celui de la question pr√©c√©dente. Faire de m√™me pour  ùë° variant de  $ùë°_4‚àí1$ √†  $ùë°_4$.
avec $
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$
avec : $
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$

De $t_2-1$ √† $t_2$ le proton se situe dans le D inf√©rieur :

on doit donc utiliser la deuxi√®me formule des vitesses: $
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$

```python
#calcul des vitesses de t2-1 √† t2
t = np.array(temps-1)
e = (-1)**2
pi = np.pi
n=50
# Nombre de termes
N = 10

# D√©finition de la fonction F
def vx(n):
    ''' Retourne la formule de la vitesse vx '''
    return e*(np.sqrt(n+1)*np.sin(pi*(n + to*np.sqrt(n)-n+1+to*np.sqrt(n+1))))

def vy(n): 
    ''' Retourne la formule de la vitesse vy '''
    return e*(np.sqrt(n+1)*np.cos(pi*(n + to*np.sqrt(n)-n+1+to*np.sqrt(n+1))))

# Liste des valeurs de vx
val_vx = [vx(k) for k in range(N)]
print("quelques valeurs de vx:",val_vx)

#liste des valeurs de vy 
val_vy = [vy(k) for k in range(N)]
print("quelques valeurs de vy:",val_vy)

```

```python
import matplotlib.pyplot as plt
plt.plot(temps[:10], val_vx)
plt.plot(temps[:10], val_vy)
plt.title("les vitesses en fonction du temps")
plt.legend(["vx", "vy"])
plt.show()
```
