## Partie 1

- Création d'un tableau des valeurs de temps jusqu'à $t_{50}$ avec la formule :
$$t_n=n+\tau \sqrt{n}.$$


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math as mt
from math import sqrt
tau=0.5
t=[0]*50 #Création d'un tableau vide 

for i in range(0,50):
    t[i]=i+tau*sqrt(i)
print(t)
```

- Vérification du tableau de temps à l'aide de valeurs données et d'opérateurs booléens.

```python
print(t[1]==1.5)
print(t[4]==5)
print(t[9]==10.5)
print(t[16]==18)
print(t[25]==27.5)
```

- Tracer des vitesses, avec les formules:

$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$




 - Tracer de $v_{x}$ et de $v_{y}$ en fonction de $t_{x}$ (intervalle $t_2-1,t_2$).

```python
t2=2.7071067811865475
tx=np.linspace(t2-1,t2,50) # tableau de temps dan sl'intervalle

vx=[0]*50 # tableau des vitesse 
vy=[0]*50

vx[0]=np.sin(np.pi*(-t2)) # valeur en zero pour simplifier la boucle 
vy[0]=np.cos(np.pi*(-t2))

i=0
for t in tx[1:]: # calcule de vx 
    i+=1
    vx[i]=(((-1)**i)*sqrt(i+1)*np.sin(np.pi*(t-t2)))
    
i=0
for t in tx[1:]: # calcule de vy 
    i+=1
    vy[i]=((-1)**i*sqrt(i+1)*np.cos(np.pi*(t-t2)))
import matplotlib.pyplot as plt

plt.plot(tx,vx)
plt.plot(tx,vy)
plt.grid()
plt.show()
```

- Tracer de $v_{x}$ et de $v_{y}$ en fonction de $t_{y}$ (intervalle $t_4-1,t_4$).

```python
t4=5
ty=np.linspace(t4-1,t4,50) # tableau de temps dan sl'intervalle

vx=[0]*50
vy=[0]*50

vx[0]=np.sin(np.pi*(-t4))
vy[0]=np.cos(np.pi*(-t4))

i=0
for t in ty[1:]:
    i+=1
    vx[i]=(((-1)**i)*sqrt(i+1)*np.sin(np.pi*(t-t4)))

i=0
for t in tx[1:]:
    i+=1
    vy[i]=((-1)**i*sqrt(i+1)*np.cos(np.pi*(t-t4)))

import matplotlib.pyplot as plt

plt.plot(ty,vx)
plt.plot(ty,vy)
plt.grid()
plt.show()
```

## Partie 2

- Représentation de  $v_{x}$ et de $v_{y}$ en fonction de $t$ dans l'intervale $0,t_6$.

D'abord nous calculons les nouvelles valeurs de $v_{x}$ et de $v_{y}$.

```python
t6=6.118033988749895
tm=np.linspace(0,t6,50)


vx=[0]*50
vy=[0]*50

vx[0]=np.sin(np.pi*(-t6))
vy[0]=np.cos(np.pi*(-t6))

i=0
for t in tm[1:]:
    i+=1
    vx[i]=((-1)**i*sqrt(i+1)*np.sin(np.pi*(t-t6)))

i=0
for t in tm[1:]:
    i+=1
    vy[i]=((-1)**i*sqrt(i+1)*np.cos(np.pi*(t-t6)))
```

**Enfin nous tracons les vitesses**

```python
import matplotlib.pyplot as plt

plt.plot(tm,vy,label="vy")
plt.plot(tm,vx,label="vx")
plt.xlabel('Temps') 
plt.ylabel('vitesse')
plt.legend()
plt.grid()
plt.show()
```

- Tracer de la norme N= $\sqrt{v_x^2+v_y^2}$ en fonction de t dans l'intervalle $0$ à $t_6$.

```python
import matplotlib.pyplot as plt
N=[0]*50
for i in range(len(vx)):
    N[i]=np.sqrt(vx[i]**2+vy[i]**2) # calcule de la norme des vitesses
plt.plot(tm,N,label="Norme")
plt.xlabel('Temps') 
plt.ylabel('Norme de la vitesse ')
plt.legend()
plt.grid()
plt.show()
```

- On constate que la norme de la vitesse est croissante en fonction du temps, un cyclotron agit comme un accelerateur de particules. 

#### Représentation du module de la vitesse $\sqrt{v_x(t_n)^2+v_y(t_n)^2}$ en fonction de $t_n-n$.
