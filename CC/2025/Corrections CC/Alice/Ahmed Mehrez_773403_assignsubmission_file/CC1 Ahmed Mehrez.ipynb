#### 1. Créer un tableau contenant les instants de sortie des boucles "D" : $t_0$, $t_1$, ... jusqu'à $t_{50}$ inclus.

On rappelle la formule des instants $t_n$ de sorties:  $t_n=n+\tau \sqrt{n}.$
On utilise la bibliothèque `numpy` pour creer le tableau et utiliser les fonctions mathématiques comme `sqrt`.

```python
import numpy as np

taux=0.5
#Création du tableau
t_sorti=np.ones(51)
for i in range(51):
    t_sorti[i]=i+taux*np.sqrt(i)
```

#### 3. Calculer les valeurs de $v_x$ et $v_y$ en fonction du temps pour $t$ variant de $t_2−1$ à $t_2$. Vous utiliserez un échantillonage suffisant pour le temps $t$, dans un tableau distinct de celui de la question précédente. Faire de même pour $t$ variant de $t_4-1$ à $t_4$. 

On va dans un premier temps créer un tableau qui contiendra les deux valeurs qui nous intèressent, puis nous allons calculer $v_x$ et $v_y$ correspondant.

On rappelle: 

Pour $t$ variant de $t_{n+1}-1$ à $t_{n+1}$ c’est-à-dire quand il est dans D supérieur ou D inférieur.
$$
v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) \qquad v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$$
Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.

```python
### Pour n=2

#On initialise les tableaux
def tab_D_in(n):
    nb_element=50 # c'est le nombre de t distincts entre t(n+1)-1 et t(n+1)
    tab=np.linspace(t_sorti[n]-1,t_sorti[n],num=nb_element)
    vx=np.zeros(nb_element)
    vy=np.zeros(nb_element)

    #On affectes les differentes vitesses aux tableaux en utilisant une boucle for
    for i in range(nb_element):
        vx[i]= (-1)**n*np.sqrt(n+1)*np.sin(np.pi*(tab[i]-t_sorti[n]))
        vy[i]= (-1)**(n+1)* np.sqrt(n+1)*np.cos(np.pi*(tab[i]-t_sorti[n]))
    return tab,vx,vy

t2,Vx2,Vy2=tab_D_in(2)
print("Les valeurs de Vx pour n=2 sont: \n",Vx2,"\n Les valeurs de Vy sont:",Vy2,"\n")

#Pour n=4, on reutilisera la fonction du dessus ainsi:

t4,Vx4,Vy4=tab_D_in(4)

print("Les valeurs de Vx pour n=4 sont: \n",Vx4,"\n Les valeurs de Vy sont:",Vy4)
```

#### 6. Représenter $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $0$ à $t_6$. On utilisera une couleur unique pour la courbe par morceaux. On remarque que $v_x$ et $v_y$ sont des fonctions continues du temps $t$. Commenter.

Pour cela on va calculer par morceaux les vitesse $v_x$ et $_y$ puisqu'elles ont différentes fonctions selon $t_n$. 

On rappelle:
1) Pour $t$ variant de $t_n$ à $t_{n+1}-1$ c’est-à-dire quand il est entre les D :
$$
v_x(t) = 0  \qquad  v_y(t) = \epsilon_n \left( \sqrt{n} + (t-t_n)/\tau \right)
$$

$$

Dans ces expressions, $\epsilon_n$ vaut $1$ pour $n$ pair et $-1$ pour $n$ impair. On pourra prendre $\epsilon_n = (-1)^n$.
