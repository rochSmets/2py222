## Particule dans un cyclotron

## Partie I


```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Création d'un tableau contenant les instants de sortie des boucles "D" (t_0 jusqu'à t_50 inclu)

t_sortie = []

for i in range(51):
    
    t_sortie.append(i + (0.5*np.sqrt(i)))
    

# 2) Création d'un tableau contenant des opérateurs booléens pour vérifier que t_1, t_4, t_9, t_16 et t_25 ont les bonnes valeurs   
    
verif = [t_sortie[1] == 1.5, t_sortie[4] == 5, t_sortie[9]==10.5, t_sortie[16]==18, t_sortie[25]==27.5 ]

print("Le tableau de vérification par booléens est :",verif) #Le tableau ne contient que des "True", les valeurs sont bonnes
```

On a bien : $t_1=1.5$, $t_4=5$, $t_9=10.5$, $t_{16}=18$, et $t_{25}=27.5$

```python
# 3) Calcul de vx et vy pour t variant de t_2 - 1 à t_2 

# Création d'un échantillonage de valeurs du temps entre t_2 - 1 et t_2 

t2 = np.linspace(t_sortie[2] - 1, t_sortie[2], 50)

# Création de tableaux contenant les valeurs de vx et vy pour t entre t_2 - 1 et t_2, qu'on calcule à l'aide de la formule donnée

vx2 = []
vy2 = []

for i in t2:
    
    vx2.append((-1)*np.sqrt(2)*np.sin(np.pi*(i - t_sortie[2])))  
    vy2.append(np.sqrt(2)*np.cos(np.pi*(i - t_sortie[2])))
    
# On fait la même chose pour t variant de t_4 - 1 à t_4 :

t4 = np.linspace(t_sortie[4] - 1, t_sortie[4], 50)

vx4 = []
vy4 = []

for i in t4:
    
    vx4.append((-1)*np.sqrt(4)*np.sin(np.pi*(i - t_sortie[4]))) 
    vy4.append(np.sqrt(4)*np.cos(np.pi*(i - t_sortie[4])))
    
    
# 4) Nous allons désormais représenter vx et vy en fonction de t pour t variant de t_2 -  1 à t_2 

plt.plot(t2,vx2,'.-', label ="Vx")
plt.plot(t2,vy2,'.-',label="Vy")
plt.title("Vx et Vy en fonction de t (pour t variant de t_2 - 1 à t_2)")
plt.xlabel("Temps (entre t_2 - 1 et t_2)")
plt.ylabel("Vitesse")
plt.legend()
```

```python
# On fait la même chose pour t variant de t_4 - 1 à t_4 :

plt.plot(t4,vx4,'.-', label ="Vx")
plt.plot(t4,vy4,'.-',label="Vy")
plt.title("Vx et Vy en fonction de t (pour t variant de t_4 - 1 à t_4)")
plt.xlabel("Temps (entre t_4 -1 et t_4)")
plt.ylabel("Vitesse")
plt.legend()
```

```python
# 5) Détermination des valeurs approchées des maximas de Vx et Vy pour t variant de t_2 - 1 à t_2

max_vx2 = max(vx2)
min_vx2 = min(vx2)

print("Une valeur approchée du maximum de Vx pour t entre t_2 - 1 et t_2 est :",round(max(vx2),1))
print("Le minimum de Vx pour t entre t_2 - 1 et t_2 est :", np.abs(round(min(vx2),1)))

max_vy2 = max(vy2)
min_vy2 = min(vy2)

print("Une valeur approchée du maximum de Vy pour t entre t_2 - 1 et t_2 est :", round(max_vy2,2))
print("Une valeur approchée du minimum de Vy pour t entre t_2 - 1 et t_2 est :", round(min_vy2,2))

# Détermination des instants correspondants :

T2 = [i for i in range(0,50)]

tx2_max = 0

ty2_max = 0
ty2_min = 0

for i in T2:
    
    if vx2[i] == max_vx2:
        tx2_max = t2[i]
        
    if vy2[i] == max_vy2:
        ty2_max = t2[i]
        
    if vy2[i] == min_vy2:
        ty2_min = t2[i]
        
print("Le maximum de vx2 est atteint (environ) en t =", round(tx2_max,2))
print("Le minimum de vx2 est atteint en t =", round(t2[0],2),"ainsi qu'en t =",round(t2[len(t2)-1],2))    
print("Le maximum de vy2 est atteint (environ) en t =", round(ty2_max,2))    
print("Le minimumm de vy2 est atteint (environ) en t =", round(ty2_min,2))
```

On remarque que le minimum (atteint 2 fois) de Vx est bien 0, et qu'il est à chaque fois atteint aux mêmes instants que les maxima de Vy. Vx semble atteindre son maximum à la moitié du temps.

```python
# On fait la même chose pour t entre t_4 - 1 et t_4 :

max_vx4 = max(vx4)
min_vx4 = min(vx4)

print("Une valeur approchée du maximum de Vx pour t entre t_4 - 1 et t_4 est :",round(max(vx4),1))
print("Le minimum de Vx pour t entre t_4 - 1 et t_4 est :", np.abs(round(min(vx4),1)))

max_vy4 = max(vy4)
min_vy4 = min(vy4)

print("Une valeur approchée du maximum de Vy pour t entre t_4 - 1 et t_4 est :", round(max_vy4,2))
print("Une valeur approchée du minimum de Vy pour t entre t_4 - 1 et t_4 est :", round(min_vy4,2))

# Détermination des instants correspondants :

T4 = [i for i in range(0,50)]

tx4_max = 0

ty4_max = 0
ty4_min = 0

for i in T4:
    
    if vx4[i] == max_vx4:
        tx4_max = t4[i]
        
    if vy4[i] == max_vy4:
        ty4_max = t4[i]
        
    if vy4[i] == min_vy4:
        ty4_min = t4[i]
        
print("Le maximum de vx4 est atteint (environ) en t =", round(tx4_max,2))
print("Le minimum de vx4 est atteint en t =", round(t4[0],2),"ainsi qu'en t =",round(t4[len(t4)-1],2))    
print("Le maximum de vy4 est atteint (environ) en t =", round(ty4_max,2))    
print("Le minimumm de vy4 est atteint (environ) en t =", round(ty4_min,2))
```

On remarque que pour t entre t_4 - 1 et t_4 les valeurs absolues des maxima sont plus élevées que pour t entre t_2 - 1 et t_2, ce qui prouve que la particule **accélère**.
