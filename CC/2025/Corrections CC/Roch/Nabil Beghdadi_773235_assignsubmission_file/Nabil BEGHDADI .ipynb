# Particule dans un cyclotron

## Partie 1

```python
#1
import numpy as np
f=[]
n=[]
for i in range(51):
    u=i+0.5*(np.sqrt(i))
    f.append(u)
t=np.array(f)
print(t)
#2
print(t[1]==1.5)
print(t[4]==5)
print(t[9]==10.5)
print(t[16]==18)
print(t[25]==27.5)
#3
tnew2=np.linspace(t[2]-1,t[2],50)
tnew4=np.linspace(t[4]-1,t[4],50)
Vx2=[]
Vy2=[]
Vx4=[]
Vy4=[]
for i in range(51):
    n.append(i)
print(n)
for i in range(len(n)-1):
    vx2=((-1)**i)*np.sqrt(i+1)*np.sin(np.pi*(tnew2[i]-(tnew2[i]+1)))
    Vx2.append(vx2)
    vy2=((-1)**i)*np.sqrt(i+1)*np.cos(np.pi*(tnew2[i]-(tnew2[i]+1)))
    Vy2.append(vy2)
    vx4=((-1)**i)*np.sqrt(i+1)*np.sin(np.pi*(tnew4[i]-(tnew4[i]+1)))
    Vx4.append(vx4)
    vy4=((-1)**i)*np.sqrt(i+1)*np.cos(np.pi*(tnew4[i]-(tnew4[i]+1)))
    Vy4.append(vy4)
#4
import matplotlib.pyplot as plt
plt.plot(tnew2,Vx2)
plt.plot(tnew2,Vy2)
plt.xlabel('temps entre les instants t[2]-1 et t[2]')
plt.ylabel('vitesses vx et vy du proton entre les instants t[2]-1 et t[2]')
plt.title('Evolution des vitesses vx et vy entre les instants t[2]-1 et t[2] ')
plt.grid()
plt.show()
plt.plot(tnew4,Vx4)
plt.plot(tnew4,Vy4)
plt.xlabel('temps entre les instants t[4]-1 et t[4]')
plt.ylabel('vitesses vx et vy du proton entre les instants t[4]-1 et t[4]')
plt.title('Evolution des vitesses vx et vy entre les instants t[4]-1 et t[4] ')
plt.grid()
plt.show()
```

## Partie 2

```python
import numpy as np
f=[]
n=[]
for i in range(51):
    u=i+0.5*(np.sqrt(i))
    f.append(u)
t=np.array(f)
#6
tnew6=np.linspace(t[0],t[6],50)
n=[]
for i in range(50):
    n.append(i)
print(n)
Vx6=[]
Vy6=[]
for i in range(len(n)-1):
    if tnew6[i]<=tnew6[i]<=tnew6[i+1]-1:
        vx6=0
        vy6=((-1)**i)*(np.sqrt(i)+(tnew6[i]-tnew6[i])/0.5)
    elif tnew6[i+1]-1<=tnew6[i]<=tnew6[i+1]:
        vx6=((-1)**i)*np.sqrt(i+1)*np.sin(np.pi*(tnew6[i]-(tnew6[i]+1)))
        vy6=((-1)**i)*np.sqrt(i+1)*np.cos(np.pi*(tnew6[i]-(tnew6[i]+1)))
    Vx6.append(vx6)
    Vy6.append(vy6)
import matplotlib.pyplot as plt
plt.plot(tnew6,Vx6)
plt.plot(tnew6,Vy6)
plt.xlabel('temps entre les instants t[0] et t[6]')
plt.ylabel('vitesses vx et vy du proton entre les instants t[0] et t[6]')
plt.title('Evolution des vitesses vx et vy entre les instants t[0] et t[6] ')
plt.grid()
plt.show()
```
