```python
import matplotlib.pyplot as plt
import numpy as np
```

#### 1/ Création d'un tableau:

```python
tau = 0.5
tab = np.empty(51)     #tableau contenant les instants de sortie des boucles d
indices = np.arange(0,51,1)    #tableau contenant les valeurs de n de 0 à 50
for i in indices:
    tab[i] = i+tau*np.sqrt(i)
print(tab)
```

#### 2/ Vérification des valeurs 

Nous voulons vérifier que $t_1=1.5$, $t_4=5$, $t_9=10.5$, $t_{16}=18$, et $t_{25}=27.5$ avec des opérateurs booléens:

```python
print(tab[1]==1.5 and tab[4]==5 and tab[9]==10.5 and tab[16]==18 and tab[25]==27.5)
```

Nous avons obtenu une valeur booléen "True", ce qui indique que les valeurs sont vérifiés.

#### 3/ Calcul des vitesses $v_x$ et $v_y$ 


```python
#Création d'un tableau contenant les t entre t_n+1 - 1 et t_n+1:
def temps(n):
    """Renvoie un tableau contenant 100 valeurs de t entre t_n+1 - 1 et t_n+1"""
    return np.linspace(tab[n+1] - 1,tab[n+1],101)

#Fonctions de vitesses
def vx(n,t,tn):
    """Renvoie la vitesse x des du proton dans le cas ou il est dans D supérieur ou inférieur"""
    e_n = (-1)^n
    return e_n * np.sqrt(n+1) * np.sin(np.pi*(t-tn))

def vy(n,t,tn):
    """Renvoie la vitesse y des du proton dans le cas ou il est dans D supérieur ou inférieur"""
    e_n = (-1)^n
    return -e_n * np.sqrt(n+1) * np.cos(np.pi*(t-tn))

#Calcul des vitesses entre t_2 − 1 à t_2:
vx2=vx(1,temps(1),tab[2])
vy2=vy(1,temps(1),tab[2])

#Calcul des vitesses entre t_4 − 1 à t_4:
vx4=vx(3,temps(3),tab[4])
vy4=vy(3,temps(3),tab[4])
```

#### 4/ Représentation graphique de $v_x$ et $v_y$ en fonction de $t$ pour $t$ variant de $t_2-1$ à $t_2$ et pour $t$ variant de $t_4-1$ à $t_4$:

Pour $t_2-1$ à $t_2$ :

```python
t_int2=temps(1)

#Vx
plt.plot(t_int2, vx2, label="vx")
plt.xlabel("temps t")
plt.ylabel("v_x")

#Vy
plt.plot(t_int2, vy2, label="vy")
plt.xlabel("temps t")
plt.ylabel("v_y")
plt.legend()
plt.grid()
plt.title("Représentation graphique de v_x et v_y en fonction du temps t variant de t_2 - 1 à t_2")

plt.show()
```

Pour $t_4-1$ à $t_4$ :

```python
t_int4=temps(3)

#Vx
plt.plot(t_int4, vx4, label="vx")
plt.xlabel("temps t")
plt.ylabel("v_x")

#Vy
plt.plot(t_int4, vy4, label="vy")
plt.xlabel("temps t")
plt.ylabel("v_y")
plt.legend()
plt.grid()
plt.title("Représentation graphique de v_x et v_y en fonction du temps t variant de t_4 - 1 à t_4")

plt.show()
```

Nous observons que l'allure des représentations graphiques sont similaires mais avec des amplitudes différents.

#### 5/ Déterminer les maxima de $v_x$ et $v_y$ et les temps associés dans les 2 cas:

```python
#Cas pour t entre t_2 - 1 et t_2:
vx2_max=np.max(vx2) #Valeur maximum de vx 
t2x = t_int2[vx2 == vx2_max]  #Temps associé à cette valeur

vy2_max=np.max(vy2)  #Valeur maximal de vy
t2y = t_int2[vy2 == vy2_max]  #Temps associé à cette valeur

print(f"La vitesse maximal de vx est {vx2_max:.2f} à un instant de t = {t2x[0]:.2f}")
print(f"La vitesse maximal de vy est {vy2_max:.2f} à un instant de t = {t2y[0]:.2f}")
```

```python
#Cas pour t entre t_4 - 1 et t_4:
vx4_max=np.max(vx4) #Valeur maximum de vx 
t4x = t_int4[vx4 == vx4_max]  #Temps associé à cette valeur

vy4_max=np.max(vy4)  #Valeur maximal de vy
t4y = t_int4[vy4 == vy4_max]  #Temps associé à cette valeur

print(f"La vitesse maximal de vx est {vx4_max:.2f} à un instant de t = {t4x[0]:.2f}")
print(f"La vitesse maximal de vy est {vy4_max:.2f} à un instant de t = {t4y[0]:.2f}")
```
