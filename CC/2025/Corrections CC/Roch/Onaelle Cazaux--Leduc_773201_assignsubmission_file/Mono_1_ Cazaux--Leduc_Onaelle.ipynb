### Question 3

```python
# calcul des valeurs de vx et vy

# definition de epsilonn
def epsilon(n):
    epsilon = 0
    if epsilon % 2 == 0:
        epsilon = 1
    if epsilon % 2 != 0:
        epsilon = -1
    return epsilon
    
```

```python
N = 50
M = 2
P = 4


# fonction vx
def vx(n, m):
    temps_t = np.linspace(t[m]-1, t[m], n+1)
    vx = np.zeros(n+1)
    for i in range (0, n+1):
        vx[i] = epsilon(m-1) * (m+1 **0.5)*np.sin(np.pi*(temps_t[i] - t[m]))
    return vx, temps_t



# fonction vy
def vy(n, m):
    temps_t = np.linspace(t[m]-1, t[m], n+1)
    vy = np.zeros(n+1)
    for i in range (0, n+1):
        vy[i] = -epsilon(m-1) * (m+1 **0.5)*np.cos(np.pi*(temps_t[i] - t[m]))
    return vy, temps_t

```

```python
vx2, temps_t1 = vx(N,M)
vy2, temps_t2 = vy(N,M)


vx4, temps_t3 = vx(N,P)
vy4, temps_t4 = vy(N,P)
```
