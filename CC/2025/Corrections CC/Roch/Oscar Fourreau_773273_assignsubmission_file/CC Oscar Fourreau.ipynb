```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau=0.5   #on définit la valeur de la variable tau
tn=[]   #On crée une  liste vide tn où on ajoutera les temps de sorties de boucles
for i in range(0,51):             #on calcule tous les tn qu'on ajoute à la liste grçace à une boucle for
    tn.append(i+tau*np.sqrt(i))
```

```python
print(tn[1]==1.5)
print(tn[4]==5)
print(tn[9]==10.5)
print(tn[16]==18)
print(tn[25]==27.5)
```

On observe donc que tous les tn testés sont bons l'opérateur booléen nous retourne bien la valeur True

```python
t2=np.linspace(tn[2]-1,tn[2],100)    #On crée un tableau contenant un échantillonnage de 100 points entre t2-1 et t2
vx2=[]
vy2=[]
for i in t2:                  #On calcule les vitesses avec une boucle for et on les stockes dans une liste 
    vx2.append((1)*np.sqrt(2+1)*np.sin(np.pi*(i-tn[2])))
    vy2.append((-1)*np.sqrt(2+1)*np.cos(np.pi*(i-tn[2])))
```

```python
t4=np.linspace(tn[4]-1,tn[4],100)    #On crée un tableau contenant un échantillonnage de 100 points entre t4-1 et t4
vx4=[]
vy4=[]
for i in t4:                           #On calcule les vitesses avec une boucle for et on les stockes dans une liste 
    vx4.append((1)*np.sqrt(4+1)*np.sin(np.pi*(i-tn[4])))
    vy4.append((-1)*np.sqrt(4+1)*np.cos(np.pi*(i-tn[4])))
```

```python
plt.plot(t2,vx2,label="Vx")
plt.plot(t2,vy2,label="Vy")
plt.grid()
plt.xlabel("Le temps t")
plt.ylabel("La Vitesse V")
plt.title("Les composantes x et y de la vitesse pour t allant de t2-1 à t2")
plt.legend()
```

```python
plt.plot(t4,vx4,label="Vx")
plt.plot(t4,vy4,label="Vy")
plt.grid()
plt.xlabel("Le temps t")
plt.ylabel("La Vitesse V")
plt.title("Les composantes x et y de la vitesse pour t allant de t4-1 à t4")
plt.legend()
```

```python
print(f"De t2-1 à t2, la valeur maximale de vx est {max(np.absolute(vx2))} atteinte pour t={t2[vx2.index(-max(np.absolute(vx2)))]} et la valeur maximale de vy est {max(np.absolute(vy2))} atteinte pour t={t2[vy2.index(-max(np.absolute(vy2)))]}")
```

```python
print(f"De t4-1 à t4, la valeur maximale de vx est {max(np.absolute(vx4))} atteinte pour t={t4[vx4.index(-max(np.absolute(vx4)))]} et la valeur maximale de vy est {max(np.absolute(vy4))} atteinte pour t={t4[vy4.index(-max(np.absolute(vy4)))]}")
```

On remarque que plus n augmente plus les vitesses vx et vy sont importantes, en effet la distance parcourue par le proton dans les d inférieur et supérieur augmente tandis que le temps passé dans un d reste de 1. Cela implique que la vitesse du  proton augmente forcément.

# Partie 2

```python
t6=[]
for i in range(0,6):                               #On crée une liste de liste qui contient les temps de chaque étape de façon séparée(le temps entre t0 et t1-1, le temps entre t1-1 et t1.....)
    t6.append(np.linspace(tn[i],tn[i+1]-1,10))
    t6.append(np.linspace(tn[i+1]-1,tn[i+1],10))
```

```python
vx=[]
vy=[]
b=0
for i in t6:                                                           #On calcule les vitesses correspondantes grâce à une double boucle fort où on différencie deux cas : entre les 2 D ou dans un des 2 D
    a=int(b)
    for j in i:
        if b%1==0:
            vx.append(0)
            vy.append(((-1)**a)*(np.sqrt(a)+(j-tn[a])/tau))
        else:        
            vx.append(((-1)**a)*np.sqrt(a+1)*np.sin(np.pi*(j-tn[a])))
            vy.append((-(-1)**a)*np.sqrt(a+1)*np.cos(np.pi*(j-tn[a])))
    b=b+0.5
```

```python
t=np.linspace(0,tn[6],120)
```

```python
plt.plot(t,vx,label="Vx")
plt.plot(t,vy,label="Vy")
plt.legend()
plt.grid()
```

```python
v=[]
for i in range(0,120):                  #On calcule la norme de la vitesse que l'on met dans une liste v
    v.append(np.sqrt((vx[i]**2)+(vy[i]**2)))
```

```python
plt.plot(t,v)
plt.grid()
plt.xlabel("Le temps t")
plt.ylabel("La norme de la vitesse v")
plt.title("Evolution de la norme de la vitesse pour t allant de 0 à t6")
```

On observe donc que la norme de la vitesse augmente par palier au cours du temps,un cyclotron sert donc à accélérer un proton.

```python
vx=[]
vy=[]
b=0
for i in t6:                                                           #On calcule les vitesses correspondantes grâce à une double boucle fort où on prend maintenant qu'un seul cas : dans un des 2 D
    a=int(b)
    for j in i:
        if b%1!=0:       
            vx.append(((-1)**a)*np.sqrt(a+1)*np.sin(np.pi*(j-tn[a])))
            vy.append((-(-1)**a)*np.sqrt(a+1)*np.cos(np.pi*(j-tn[a])))
    b=b+0.5
```

```python
v=[]
for i in range(0,60):                  #On calcule la norme de la vitesse que l'on met dans une liste v
    v.append(np.sqrt((vx[i]**2)+(vy[i]**2)))
```

```python
t=np.linspace(0,tn[6],60)
```

```python
plt.plot(t,v)
plt.grid()
plt.xlabel("Le temps t")
plt.ylabel("La norme de la vitesse v")
plt.title("Evolution de la norme de la vitesse pour t allant de 0 à t6")
```
