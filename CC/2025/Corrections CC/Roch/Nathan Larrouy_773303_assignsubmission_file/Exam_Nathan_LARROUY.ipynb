# Etude d'une particule dans un cyclotron

```python
import numpy as np
import matplotlib.pyplot as plt



tau = 0.5
t = np.empty(0)
for n in range (0, 51):                    #boucle remplissant un tableau avec les valeurs de tn selon la formule donnée pour n allant de 0 à 50 inclus
    t_n = n + tau*np.sqrt(n)
    t = np.append(t, t_n)
  
                                           #vérification des valeurs de t avec des booléens
if (t[1] == 1.5) :
    print("t1 check")
else :
    print("t1 error")

if (t[4] == 5) :
    print("t4 check")
else :
    print("t4 error")
    
if (t[9] == 10.5) :
    print("t9 check")
else :
    print("t9 error")
    
if (t[16] == 18) :
    print("t16 check")
else :
    print("t16 error")
    
if (t[25] == 27.5) :
    print("t25 check")
else :
    print("t25 error")
    
"""
fonction traitant du mouvement d'une particule lorsqu'elle se trouve dans un D dans l'inervalle de temps t_n+1 - 1 , t_n+1
elle prends pour argument la valeur de n+1 (attention à ne pas se tromper d'indice)
la fonction calcul vx et vy durant cet inerval de temps, affiche le maximum des valeurs abs de chacune, et trace les valeurs sur un graphe

"""    
def vitesse_D(n): 
    T = np.linspace(t[n]-1, t[n], 20)
    vx = np.empty(0)
    vy = np.empty(0)
    for i in T :                                                      #implémentation des vitesses avec les formules données
        vitx = (-1)**(n-1)*np.sqrt(n)*np.sin(np.pi*(i - t[n]))
        vity = -(-1)**(n-1)*np.sqrt(n)*np.cos(np.pi*(i - t[n]))
        vx = np.append(vx, vitx)
        vy = np.append(vy, vity)
    
    if abs(vx[0])>abs(vx[1]):                               #les boucles "for" qui suivent permettent d'obtenir les valeurs max de vx et vy
        vxmax = abs(vx[0])
    else :
        vxmax = abs(vx[1])
        
    if abs(vy[0])>abs(vy[1]):
        vymax = abs(vy[0])
    else :
        vymax = abs(vy[1])
        
    for i in range(1, np.size(vx)):
        if abs(vx[i-1]) > vxmax:
            vxmax = abs(vx[i-1])
            txmax = T[i-1]
            
    for i in range(1, np.size(vx)):
        if abs(vy[i-1]) >= vymax:
            vymax = abs(vy[i-1])
            tymax = T[i-1]

     
    print ("vx_max en norme = ",vxmax, "et t_vxmax = ", txmax)    #affichage des valeurs max trouvées
    print ("vy_max en norme = ",vymax, "et t_vymax = ", tymax)
    
    print( "graphe de vx et vy entre (t",n,")-1 et (t",n,")")     #affichage des courbes et mise en page 
    plt.plot(T, vx)
    plt.plot(T, vy)
    plt.legend(['vx','vy'])
    plt.grid()
    plt.xlabel("temps t")
    plt.ylabel("vitesse v")

vitesse_D(2)   
```

```python
vitesse_D(4)
```

Sur les deux graphes on observe que la vitesse selon x est toujours nulle à l'entrée et à la sortie du D (ce qui est de bonne augure pour la continuité de la fonction vx pour le mouvement complet car vx en dehors des D est nulle ). Concernant vy sa valeurs d'arrivée vaut l'opposée de sa valeure de départ et elle vaut 0 au milieu de l'interval, ce qui correspond à ce qu'on observe : la particule fait un quart de cercle de gauche à droite "en descendant" donc vy<0 puis fait un deuxième quart de cercle de gauche à droite "en montant" donc vy devient positive . On remarque également qui plus on avance dans les intervals de temps tn, plus l'amplitude des vitesses est grande.

```python
"""
fonction traitant du mouvement d'une particule lorsqu'elle se trouve entre deux D l'inervalle de temps t_n , t_n+1 - 1 
elle prends pour argument la valeur de n
la fonction calcul vx et vy durant cet inerval de temps et trace les valeurs sur un graphe

"""    
def vitesse_inter_D(n):
    T = np.linspace(t[n], t[n+1]-1, 20)
    vx = np.empty(0)
    vy = np.empty(0)
    for i in T :
        vitx = 0
        vity = (-1)**n*(np.sqrt(n)+(i-t[n])/tau)
        vx = np.append(vx, vitx)
        vy = np.append(vy, vity)
    
    plt.plot(T, vx, 'r')
    plt.plot(T, vy, 'k')
    plt.legend(['vx','vy'])

"""
Variante de la fonction vitesse_D2 qui trace uniquement les graphes de vx et vy"

"""
def vitesse_D2(n):
    T = np.linspace(t[n]-1, t[n], 20)
    vx = np.empty(0)
    vy = np.empty(0)
    for i in T :
        vitx = (-1)**(n-1)*np.sqrt(n)*np.sin(np.pi*(i - t[n]))
        vity = -(-1)**(n-1)*np.sqrt(n)*np.cos(np.pi*(i - t[n]))
        vx = np.append(vx, vitx)
        vy = np.append(vy, vity)
    plt.plot(T, vx, 'r')
    plt.plot(T, vy, 'k')
    plt.legend(['vx','vy'])
    plt.xlabel("temps t")
    plt.ylabel("vitesse v")
    
plt.title("graphe de vx et vy entre t_0 et t_6")

for i in range (0, 7):
    vitesse_inter_D(i)
    vitesse_D2(i+1)
```

```python
def vitesse_inter_D_mod(n):
    T = np.linspace(t[n], t[n+1]-1, 20)
    v = np.empty(0)
    for i in T :
        vitx = 0
        vity = (-1)**n*(np.sqrt(n)+(i-t[n])/tau)
        vit = np.sqrt(vitx**2 + vity**2)
        v = np.append(v, vit)
        
    
    plt.plot(T, v, 'r')
    
    
def vitesse_D2_mod(n):
    T = np.linspace(t[n]-1, t[n], 20)
    v = np.empty(0)
    for i in T :
        vitx = (-1)**(n-1)*np.sqrt(n)*np.sin(np.pi*(i - t[n]))
        vity = -(-1)**(n-1)*np.sqrt(n)*np.cos(np.pi*(i - t[n]))
        vit = np.sqrt(vitx**2 + vity**2)
        v = np.append(v, vit)
        
    plt.plot(T, v, 'r')
    

plt.title("module de la vitesse en fonction du temps") 
plt.ylabel("module v")
plt.xlabel("temps t")


for i in range (0, 7):
    vitesse_inter_D_mod(i)
    vitesse_D2_mod(i+1)
```
