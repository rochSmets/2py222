#  Particule dans un cyclotron

## Introduction

 Un cyclotron est un acc√©l√©rateur de particule constitu√© de deux parties appel√©es "D" √† cause de leur forme. A l'aide d'une tension alternative appliqu√©e entre les D quisont plong√©s dans un champs magn√©tiques, le principe d'un cyclotron consiste √† g√©n√©rer un champs √©l√©ctrique dont le sens et la direction sont similaires √† ceux de la vitesse du proton lorsqu‚Äôil est entre les D.   
L'objectif de ce travail sera de'√©tudier l'√©volution de la vitesse d'un proton dans un cyclotron. Pour ce faire, nous t√¢cherons de mod√©liser la position du proton en focntion du temps afin d'en tirer des conclusions sur sa vitesse dans cgaque phase de son mouvement √† ttravers le cyclotron.  
Nous proc√®derons selon deux parties :
1. **Partie I** : Etude des vitesses du proton selon x et y pour deux intervalles de temps
2. **Partie II** : Etude de la norme de la vitesse du proton dans le cyclotron durant tout le mouvement

### Partie I : Etude des vitesses du proton selon x et y pour deux intervalles de temps

Cr√©ons tous d'abord un tableau repr√©sentant les instants diff√©rentes sorties des boucles D :

```python
import numpy as np #importation de la bibliioth√®que n√©cessaire √† la cr√©ation de tableaux
tau=0.5 #d√©claration des param√®tres du probl√®me
t=[] #initialisation de la liste des valeurs de t
for n in range(0,51):
    t_n= n + tau*np.sqrt(n) #calcul des instants t
    t.append(t_n) #ajout √† la liste de chaque valeur de t
tab_t=np.array(t) #conversion de la liste en tableau
print(tab_t[2])
```

Verifions qu'on a bien $t_1=1.5$ , $t_4=5$ , $ùë°_9=10.5$ , $t_{16}=18$ , et $t_{25}=27.5$

```python
print(tab_t[1]==1.5)
print(tab_t[4]==5)
print(tab_t[9]==10.5)
print(tab_t[16]==18) 
print(tab_t[25]==27.5)
```

L'op√©rateur bool√©√©n renvoie bien "True" pour chacune de ces affirmations, nous avons donc bien v√©rifi√© les valeurs demand√©es.

Calculons √† pr√©sent les valeurs de $v_x$ et $v_y$ gr√¢ce aux formules donn√©es pour t variant de $t_2 -1 $ √† $t_2$ 

Les formules donn√©es et adapt√©es √† cette situation sont :  
$v_x(t) = \epsilon_n \sqrt{n+1} \sin \left( \pi (t-t_{n+1}) \right) $ 

et 
$v_y(t) = -\epsilon_n \sqrt{n+1} \cos \left( \pi (t-t_{n+1}) \right)
$

avec $\epsilon=1$ ici car $n$ est pair 

```python
tab_t_2=np.linspace(tab_t[2]-1, tab_t[2],50) #Cr√©ons d'abord un tableau consignant un √©chantillonnage de $t_2 -1$ √† $t_2$
epsilon=1 #param√®tre du probl√®me

list_vx_t2=[] #initialisation de la liste des valeurs de v_x
list_vy_t2=[] #initialisation de la liste des valeurs de v_y

#calcul de v_x : 

for n in range(len(tab_t_2)): 
    a=epsilon*np.sqrt(n+1)*np.sin(np.pi*(tab_t_2[n-1]-tab_t_2[n])) #calcul de chaque √©l√©ment individuellement
    list_vx_t2.append(a) #ajout √† la liste initiale
vx_t2=np.array(list_vx_t2) #conversion de la liste en tableau

# calcul de v_y :

for n in range(len(tab_t_2)): 
    b=-epsilon*np.sqrt(n+1)*np.cos(np.pi*(tab_t_2[n-1]-tab_t_2[n])) #calcul de chaque √©l√©ment individuellement
    list_vy_t2.append(b) #ajout √† la liste initiale
vy_t2=np.array(list_vy_t2) #conversion de la liste en tableau
```

On proc√®de de m√™me pour t variant de $t_4 -1 $ √† $t_4$  

```python
tab_t_4=np.linspace(tab_t[4]-1, tab_t[4],50) 
epsilon=1 

list_vx_t4=[]
list_vy_t4=[]

#calcul de v_x :

for n in range(len(tab_t_4)): 
    a=epsilon*np.sqrt(n+1)*np.sin(np.pi*(tab_t_4[n-1]-tab_t_4[n])) 
    list_vx_t4.append(a) 
vx_t4=np.array(list_vx_t4) 

#calcul de v_y :
for n in range(len(tab_t_4)): 
    b=-epsilon*np.sqrt(n+1)*np.cos(np.pi*(tab_t_4[n-1]-tab_t_4[n])) #calcul de chaque √©l√©ment individuellement
    list_vy_t4.append(b) #ajout √† la liste initiale
vy_t4=np.array(list_vy_t4) #conversion de la liste en tableau
```

Repr√©sentons graphiquement les variations de $v_x$ et $v_y$ en fonction de t, respectivement pour $t_2$ et $t_4$

```python
import matplotlib.pyplot as plt #importation de la biblioth√®que n√©cessaire

#Pour t_2 : 

#Trac√©
plt.plot(tab_t_2,vx_t2, label = 'vitesse selon x')
plt.plot(tab_t_2, vy_t2, label='vitesse selon y')

#Pour faire joli
plt.xlabel('Valeurs de t (entre $t_2-1$ et t_2)')
plt.ylabel('Vitesse de la particule')
plt.title('Evolution de la vitesse de la particule en fonction du temps')
plt.legend()
plt.grid()
```

```python
#Pour t_4 : 

plt.plot(tab_t_4,vx_t4, label='vitesse selon x')
plt.plot(tab_t_4, vy_t4, label='vitesse selon y')

plt.xlabel('Valeurs de t (entre $t_4-1$ et t_4)')
plt.ylabel('Vitesse de la particule')
plt.title('Evolution de la vitesse de la particule en fonction du temps')
plt.legend()
plt.grid()
```

D√©terminons les maxima des vitesses $v_x$ et $v_y$.  

Nous nous servons des listes de valeurs des vitesses utilis√©es pr√©c√©demment pour cr√©er les tableaux, auxquelles nous appliquons la fonction "max" de python

```python
#pour t_2 :
max_vx_t2=max(list_vx_t2)
print('Le maximum de la vitesse selon x entre t_2-1 et t_2 est', max_vx_t2)
max_vy_t2=max(list_vy_t2)
print('Le maximum de la vitesse selon y entre t_2-1 et t_2 est', max_vy_t2)
```

```python
#pour t_4 :
max_vx_t4=max(list_vx_t4)
print('Le maximum de la vitesse selon x entre t_4-1 et t_4 est', max_vx_t4)
max_vy_t4=max(list_vy_t4)
print('Le maximum de la vitesse selon y entre t_4-1 et t_4 est', max_vy_t4)
```

On remarque que les maxima des vitesses sont les m√™mes quelque soit l'intervalle de valeurs de t, ce qui souligne le caract√®re p√©riodique de la vitesse dans le cyclotron qu'on avait conjectur√© √† travers la forme de la courbe du louvement de la particule ainsi que par l'utilisation des focntions sin et cos pour mod√©liser la vitesse
