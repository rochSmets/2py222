## Introduction :

Pour ce travail nous allons nous intéresser au comportement d'une particule dans un cyclotron afin d'étudier sa vitesse selon les axes x et y et de les représenter au cours du temps.


```python
import matplotlib.pyplot as plt
import numpy as np

#importation des bibliothèques utiles
```

```python
temps = []

for i in range(50):
    temps.append(i + 0.5*(i)**(1/2))   #création d'un tableau contenant les instants de sortie des boucles
    
print(temps[1] == 1.5)
print(temps[4] == 5)
print(temps[9] == 10.5)
print(temps[16] == 18)
print(temps[25] == 27.5) #on vérifie que les tn correspondent bien aux valeurs de l'énoncé avec des booléens
```

Grâce au mot "True", (vrai), on voit que les valeurs de l'énoncé sont bien les valeurs contenues dans le tableau temps.

```python
t2 = np.linspace(temps[2]-1,temps[2],100)   #création d'un tableau contenant 100 valeurs espacées régulièrement entre t2-1 et t2

def vx2(t):
    vx2 = []
    for i in range(len(t)-1):
        vx2.append((-1)**1*(2)**(1/2)*np.sin(np.pi*(t[i]-temps[2])))
    return vx2 

vitesse_x_t2 = vx2(t2)       #définition de la fonction pour calculer la vitesse selon x entre t2-1 et t2 pour 100 valeurs de t et à l'aide de la formule donnée

def vy2(t):
    vy2 = []
    for i in range(len(t)-1):
        vy2.append(-(-1)**1*(2)**(1/2)*np.cos(np.pi*(t[i]-temps[2])))
    return vy2

vitesse_y_t2 = vy2(t2)      #définition de la fonction pour calculer la vitesse selon y entre t2-1 et t2 pour 100 valeurs de t et à l'aide de la formule donnée

t4 = np.linspace(temps[4]-1,temps[4],100)   #création d'un tableau contenant 100 valeurs espacées régulièrement entre t4-1 et t4

def vx4(t):
    vx4 = []
    for i in range(len(t)-1):
        vx4.append((-1)**3*(4)**(1/2)*np.sin(np.pi*(t[i]-temps[4])))
    return vx4

vitesse_x_t4 = vx4(t4)     #définition de la fonction pour calculer la vitesse selon x entre t4-1 et t4 pour 100 valeurs de t et à l'aide de la formule donnée

def vy4(t):
    vy4 = []
    for i in range(len(t)-1):
        vy4.append(-(-1)**3*(4)**(1/2)*np.cos(np.pi*(t[i]-temps[4])))
    return vy4

vitesse_y_t4 = vy4(t4)     #définition de la fonction pour calculer la vitesse selon y entre t4-1 et t4 pour 100 valeurs de t et à l'aide de la formule donnée
```

```python
plt.plot(t2[0:99],vitesse_x_t2,label="vitesse selon x")
plt.plot(t2[0:99],vitesse_y_t2,label="vitesse selon y")
plt.legend()
plt.title("Vitesse de la particule entre t2-1 et t2")
plt.xlabel("temps")
plt.ylabel("vitesse de la particule")
plt.grid()
plt.show()
                                                         #tracage des courbes des vitesse x et y pour les deux intervalles de temps
plt.plot(t4[0:99],vitesse_x_t4,label="vitesse selon x")
plt.plot(t4[0:99],vitesse_y_t4,label="vitesse selon y")
plt.legend()
plt.title("Vitesse de la particule entre t4 -1 et t4")
plt.xlabel("temps")
plt.ylabel("vitesse de la particule")
plt.grid()
plt.show()
```

```python


max_x_t2 = vitesse_x_t2[50]
max_tx_t2 = t2[97]

max_y_t2 = vitesse_y_t2[97]
max_ty_t2 = t4[97]

print("la vitesse max en x est",max_x_t2, "et celle en y est", max_y_t2, "au temps ",max_tx_t2)

max_x_t4 = vitesse_x_t4[50]
max_tx_t4 = t4[97]
print(max_x_t4)

max_y_t4 = vitesse_y_t4[97]
max_ty_t4 = t4[97]
print("la vitesse max en x est",max_x_t4, "et celle en y est", max_y_t4, "au temps ",max_tx_t4)

#Les maximas se trouvent sur les bords des intervalles, après vérification les valeurs obtenues ici sont bien les plus grandes valeurs que l'on observe sur les graphiques
 
```

Dans les deux cas on remarque que la vitesse selon x est accélérée puis décélérée en raison du changement de direction de la particule affectée par le cyclotron.
Pour les vitesse selon y on voit qu'elles sont toujours accélérées au cours des temps passées dans les parties supérieures ou inférieures.

De plus, la vitesse en x et en y pour l'intervalle de temps t4-1 à t4, est plus grande, il y a donc une accélération de la particule plus elle passe de fois dans le cyclotron.

```python
t6 = np.linspace(0,temps[6],600)  #création d'un tableau contenant des valeurs régulièrement espacées entre 0 et t6

for u in range(1,7):
    vx = []
    vy = []
    
    t = np.linspace(temps[u]-1,temps[u],100)   #création d'un tableau contenant 100 valeurs espacées régulièrement entre tn-1 et tn
    
    for a in range(u):
        for i in range(len(t)):
            x = ((-1)**(u-1)*(u)**(1/2)*np.sin(np.pi*(t[i]-temps[u])))
            vx.append(x)
        
    for a in range(u):
        for i in range(len(t)):
            y =(-(-1)**(u-1)*(u)**(1/2)*np.cos(np.pi*(t[i]-temps[u])))
            vy.append(y)        #ajoute les valeur de vx et vy en fonction des intervalles tn -1 et tn pour n allant de 1 à 6
            
#il manque les intervalles entre tn et tn+1 -1, manque de temps pour les rajouter dans le code            
    
        
        
plt.plot(t6, vx, label="vitesse x")
plt.plot(t6, vy, label="vitesse y")
plt.legend()
plt.grid()
plt.xlabel("temps")
plt.ylabel("vitesse de la particule")
plt.title("Vitesses de la particule entre 0 et t6")
plt.show()
```

Ici la courbe n'est pas continue comme il manque les intervalles tn à tn+1 -1, normalement elle l'est.

```python
module = []

for i in range(len(t6)):
    module.append((vx[i]**+vy[i]**2)**(1/2))
    
plt.plot(t6, module)
plt.title("module des vitesse de 0 à t6")
plt.xlabel("temps")
plt.ylabel("module")
plt.grid()
plt.show()
```

Le but d'un cyclotron est d'accélérer une particule chargée grâce à un champ magnétique. Ici le module de la vitesse est sensé être accéléré, le problème vient surement du calcul de vx et vy dans la cellule précédente.
Il manque à calculer les périodes où la particule est entre les parties du cyclotron.
