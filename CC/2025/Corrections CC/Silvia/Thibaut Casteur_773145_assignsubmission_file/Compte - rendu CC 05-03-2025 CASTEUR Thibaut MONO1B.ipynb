## I/. Partie 1

```python
# importation des modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_values = [] #cr√©ation de la liste des temps
tho = 0.5

for i in range (0, 50):
    t_values.append(i + tho * np.sqrt(i)) #application de la formule : tn = n + tho.sqrt(n)

#Graphe
print(t_values)
print()
print(f't1 = 1.5 ? : {t_values[1] == 1.5}')
print(f't4 = 5 ? : {t_values[4] == 5}')
print(f't9 = 10.5 ? : {t_values[9] == 10.5}')
print(f't16 = 19 ? : {t_values[16] == 18}')
print(f't25 = 27.5 ? : {t_values[25] == 27.5}')
```

```python
t2 = np.linspace(t_values[2]-1, t_values[2], 100) #cr√©ation d'une liste de temps de t2-1 a t2
vx2 = (-1)**1 * np.sqrt(2) * np.sin(np.pi * (t2 - t_values[2])) #formule de vx
vy2 = -(-1)**1 * np.sqrt(2) * np.cos(np.pi * (t2 - t_values[2]))  #formule de vy

# Graphe
plt.figure(figsize=(10, 5))
plt.plot(t2, vx2, label='v_x(t)')
plt.plot(t2, vy2, label='v_y(t)')
plt.title('Composante des vitesses en fonction du temps')
plt.xlabel(f"temps de {min(t2):.2f} √† {max(t2):.2f} (ici, sans unit√©)")
plt.ylabel("Composantes des vitesses")
plt.legend()
plt.grid()
plt.show()
```

```python
t4 = np.linspace(t_values[4]-1, t_values[4], 100) #cr√©ation d'une liste de temps de t4-1 a t4
vx4 = (-1)**3 * np.sqrt(4) * np.sin(np.pi * (t4 - t_values[4])) #formule de vx
vy4 = -(-1)**3 * np.sqrt(4) * np.cos(np.pi * (t4 - t_values[4]))  #formule de vy

#Graphe
plt.figure(figsize=(10, 5))
plt.plot(t4, vx4, label='v_x(t)')
plt.plot(t4, vy4, label='v_y(t)')
plt.title('Composante des vitesses en fonction du temps')
plt.xlabel(f"temps de {min(t4):.2f} √† {max(t4):.2f} (ici, sans unit√©)")
plt.ylabel("Composantes des vitesses")
plt.legend()
plt.grid()
plt.show()
```

```python
print(f'v_x maximale pour t entre 1.71 et 2.71 : {max(vx2):.2f} √† t = {t2[50]}') # v_x est maximale que t est √† √©quidistance du d√©part et de l'arriv√©e
print(f'v_y maximale pour t entre 1.71 et 2.71 : {max(vy2):.2f} √† t = {t2[99]}') # v_y est maximale que t arrive √† t(n+1) (√† son arriv√©e entre les D)
print(f'v_x maximale pour t entre 4.00 et 5.00 : {max(vx4):.2f} √† t = {t4[50]}') # v_x est maximale que t est √† √©quidistance du d√©part et de l'arriv√©e
print(f'v_y maximale pour t entre 4.00 et 5.00 : {max(vy4):.2f} √† t = {t4[99]}')# v_y est maximale que t arrive √† t(n+1) (√† son arriv√©e entre les D)
```

On remarque que les deux admettent les m√™mes maximaux, aussi bien pour la composante de la vitesse selon $x$ que celle selon $y$.
On remarque la vitesse selon $x$ est maximale au milieu de $t_{n+1} - 1$ et $t_{n+1}$, c'est-√†-dire finalement le moment o√π dans le cyclotron, la trajectoire n'est que purement selon $v_x$. Graphiquement cela se confirme avec le fait que la vitesse selon $y$ est de 0 lors du maximum de $v_x$.Pour ce qui est de la vitesse selon $y$, on remarque qu'elle est d'abord n√©gative pour finir positive est maximale au dernier moment. En effet, cela correspond √† D inf√©rieur, o√π la vitesse selon $y$ est dabord n√©gative car la particule descend, puis en remontant sa vitesse devient maximale, pour finir √† l'arriv√©e par n'avoir qu'une composante non nulle selon $y$.

## II/. Partie 2

```python
plt.figure(figsize=(10,5)) # Cr√©ation  du graphe

for n in range(0, 7): # pour n allant de 0 √† 6
    t = np.linspace(t_values[n], t_values[n+1]-1, 100) # liste de temps de t(n) √† t(n+1) - 1
    vx = t * 0 # liste de sa composante selon x dans cette liste de temps
    plt.plot(t, vx)
    t = np.linspace(t_values[n+1]-1, t_values[n+1], 100) # liste de temps de t(n+1)-1 √† t(n+1)
    vx = (-1)**n * np.sqrt(n+1) * np.sin(np.pi * (t - t_values[n+1])) # liste de sa composante selon x dans cette liste de temps
    plt.plot(t, vx)
plt.title('Composante de la vitesse selon x en fonction du temps')
plt.xlabel(f"temps (ici, sans unit√©)")
plt.ylabel("Composante de la vitesse selon x")
plt.grid()
plt.show()

plt.figure(figsize=(10,5)) # Cr√©ation  du graphe

for n in range(0, 7): # pour n allant de 0 √† 6
    t = np.linspace(t_values[n], t_values[n+1]-1, 100) # liste de temps de t(n) √† t(n+1) - 1
    vy = (-1)**n * (np.sqrt(n) + ((t - t_values[n])/tho)) # liste de sa composante selon y dans cette liste de temps
    plt.plot(t, vy)
    t = np.linspace(t_values[n+1]-1, t_values[n+1], 100) # liste de temps de t(n+1)-1 √† t(n+1)
    vy = -(-1)**n * np.sqrt(n+1) * np.cos(np.pi * (t - t_values[n+1])) # liste de sa composante selon y dans cette liste de temps
    plt.plot(t, vy)
plt.title('Composante de la vitesse selon y en fonction du temps')
plt.xlabel(f"temps (ici, sans unit√©)")
plt.ylabel("Composante de la vitesse selon y")
plt.grid()
plt.show()
```

On observe bien ce √† quoi on s'attendait : 
* entre les D, la vitesse n'a qu'une composante non nulle selon y, et parfaitement nul selon x.
* dans D sup√©rieur, la vitesse selon x est n√©gative et minimale le plus loin dans D. En outre, la vitesse selon y passe de positive √† n√©gative, en valant 0 quand la composante selon x est minimale.
* dans D inf√©rieur, la vitesse selon x est positive est maximale le plus loin dans D. En outre, la vitesse selon y passe de n√©gative √† positive, en valant 0 quand la composante selon x est maximale.

```python
plt.figure(figsize=(10,5)) # Cr√©ation  du graphe

for n in range(0, 7): # pour n allant de 0 √† 6
    t = np.linspace(t_values[n], t_values[n+1]-1, 100) # liste de temps de t(n) √† t(n+1) - 1
    vx = t * 0 # liste de sa composante selon x dans cette liste de temps
    vy = (-1)**n * (np.sqrt(n) + ((t - t_values[n])/tho)) # liste de sa composante selon y dans cette liste de temps
    v = np.sqrt(vx**2 + vy**2) # Norme de la vitesse dans cette liste de temps
    plt.plot(t, v)
    t = np.linspace(t_values[n+1]-1, t_values[n+1], 100) # liste de temps de t(n+1)-1 √† t(n+1)
    vx = (-1)**n * np.sqrt(n+1) * np.sin(np.pi * (t - t_values[n+1])) # liste de sa composante selon x dans cette liste de temps
    vy = -(-1)**n * np.sqrt(n+1) * np.cos(np.pi * (t - t_values[n+1])) # liste de sa composante selon y dans cette liste de temps
    v = np.sqrt(vx**2 + vy**2) # Norme de la vitesse dans cette liste de temps
    plt.plot(t, v)

plt.title('Norme de la vitesse en fonction du temps')
plt.xlabel(f"temps (ici, sans unit√©)")
plt.ylabel("Norme de la vitesse")
plt.grid()
plt.show()
```

Ainsi, sur ce graphique, on retrouve bien l'int√©r√™t d'un cyclotron : c'est en fait un acc√©l√©rateur de particule. En effet, en passant dans les D, la particule accumule de l'√©nergie cin√©tique. Sa vitesse augmente donc tr√®s rapidement. Cet instrument a par exemple des utilit√©s dans le domaine de la physique en permettant d'observer des collisions de particules, qui normalement se repoussent. Ainsi, en les acc√©l√©rant, on arrive √† aller au-del√† de leur force de r√©pulsion. Cela a m√™me des applications en m√©decine, car il permet, par des collisions, d'obtenir des particules radioactives, d'excellents traceurs corporelles. 

Pour ce qui est de calculer, pour $n$ variant de $0$ √† $50$, repr√©senter le module de la vitesse $\sqrt{v_x(t_n)^2+v_y(t_n)^2}$ en fonction de $t_n-n$,nous pouvons constater que la norme de la vitesse est continue. De ce fait, en utilisant les formules de la vitesse pour $t$ variant de $t_n$ √† $t_{n+1}-1$, nous devrions retrouver les m√™mes r√©sultats qu'avec les autres formules. Toutefois, celles-ci sont pr√©f√©rables en raison du fait que la composante de la vitesse selon x est nulle. ainsi, on a : $\sqrt{v_y(t_n)^2}$

```python
tn = np.zeros(50) # liste qui contiendra les tn - n

for n in range(0, 50) :
    v_values = np.sqrt(((-1)**n * (np.sqrt(n) + ((t_values - t_values[n])/tho)))**2)
    tn[n] = t_values[n] - n
    
plt.plot(tn, v_values, color='g') 
plt.title('Norme de la vitesse √† l instant t_n en fonction du temps totale pass√© par le proton entre les D')
plt.xlabel(f"temps pass√© par le proton entre les D de {min(tn):.2f} √† {max(tn):.2f} (ici, sans unit√©)")
plt.ylabel("Norme de la vitesse")
plt.grid()
plt.show()
```

On constate enfin que plus le temps total pass√© par le proton entre les D est important, plus sa vitesse √† cet instant d'entr√©e/sortie sera basse. On en d√©duit alors qu'il est tr√®s probable que la vitesse varient donc beaucoup plus d√®s la sortie, de fa√ßon √† rester en accord avec le fait qu'il s'agisse bien d'un acc√©l√©rateur de particule. Cela peut se justifier par le fait qu'√† l'int√©rieur, il n'y a plus de champs. Sachant qu'√† l'exterieur, le module de la vitesse ne fait qu'augmenter, on en d√©duit que les variations de vitesse sont donc de plus en plus intense. Sur la figure  cela peut aussi se comprendre par le fait que plus le temps s'est √©coul√©, plus la particule va loin du point de d√©part. Or la distance entre les D ne changeant pas, et sachant que $v = d / t $, on comprends que plus la temps augmente entre ses plaquues, plus la vitesse baissera, car d est constant.
