### PARTIE I

```python
#1

import numpy as np
L=[] # On crée une liste pour d'abord stocker les valeurs
for i in range(51):
    L.append(i+0.5*((i)**(1/2)))
    
T=np.array(L) #On transformela liste en tableau
print(T)
```

```python
#2

print(T[1]==1.5) #On teste avec == pour vérifier que chaque égalité soit vraie
print(T[4]==5)
print(T[9]==10.5)
print(T[16]==18)
print(T[25]==27.5)
```

```python
#3
#Pour n=1
T2=np.linspace(T[2]-1,T[2],50) #On crée un échantillonage de 50 valeurs de t2 -1 à t2 
print(T2)
Vx=[]
for i in range(len(T2)):  ##Vx, on applique les formules données
    Vx.append((-1)*(np.sqrt(2))*(np.sin((np.pi)*(T2[i]-T[2]))))

Vx1=np.array(Vx)
print(Vx1)
Vy=[]
for i in range(len(T2)):  ##Vy On applique les formules données
    Vy.append((1)*(np.sqrt(2))*(np.cos((np.pi)*(T2[i]-T[2]))))
Vy1=np.array(Vy)   
print(Vy1)

```

```python
#pour n=3
# Même commentaire que pour le cas précédent

T4=np.linspace(T[4]-1,T[4],50)
print(T4)

Vx4=[]
for i in range(len(T4)):  ##Vx
    Vx4.append((-1)*(np.sqrt(4))*(np.sin((np.pi)*(T4[i]-T[4]))))

Vx44=np.array(Vx4)
print(Vx44)
Vy4=[]
for i in range(len(T4)):  ##Vy
    Vy4.append((1)*(np.sqrt(4))*(np.cos((np.pi)*(T4[i]-T[4]))))
Vy44=np.array(Vy4)   
print(Vy44)
```

```python
#4

import matplotlib.pyplot as plt

plt.plot(T2,Vx1)
plt.plot(T2,Vy1)
plt.ylabel('Vitesse en ordonnée')
plt.xlabel('Temps en s de t2 à t2 -1 ')
plt.title('Représentation graphique de la vitesse en fonction du temps')
plt.grid()
plt.show()
```

```python
#6
# On crée des listes vides pour stocker la vitesse
Vx=[]
Vy=[]

for i in range(6):       #On crée une boucle, pour caluler de t0 à t1 -1 puis t1-1 à t1, ensuite t1 à t1-1, t1-1 à t2 etc...
    Tn=np.linspace(T[i],T[i+1]-1,50)   #On crée des valeurs de t pour Ti à Ti+1-1
    Tn1=np.linspace(T[i+1]-1,T[i+1],50)#On crée des valeurs de t pour Ti+1-1 à Ti+1
    for a in range(len(Tn)): # On applique les formules pour Ti à Ti+1 - 1
        Vx.append(0)
        Vy.append(((-1)**i)*(np.sqrt(i)+(Tn[a]-T[i])/0.5))
    for b in range(len(Tn1)): # On applique les formules pour Ti+1 - 1 à Ti+1
        Vx.append(((-1)**i)*(np.sqrt(i+1))*(np.sin((np.pi)*(Tn1[b]-T[i+1]))))
        Vy.append((-(-1)**i)*(np.sqrt(i+1))*(np.cos((np.pi)*(Tn1[b]-T[i+1]))))
            
# On transforme les listes en tableaux            
VX=np.array(Vx) 
VY=np.array(Vy)

# On fait la représentation graphique demandée
Ta=np.linspace(0,T[6],600)
plt.plot(Ta,VX,'b')
plt.plot(Ta,VY,'r')
plt.xlabel('Temps en s')
plt.ylabel('Vitesse en ordonnée')
plt.title('Représentation de vx et vy en fonction de t')
plt.grid()
plt.show()
        
```

```python
#7

M=[]    
for i in range(len(VY)):
    M.append(np.sqrt((VX[i])**2 + VY[i]**2))  #On applique une boucle où on ajoute le calcul du module pour chaque t
    

M1=np.array(M)
plt.plot(Ta,M1)
plt.title('Module de la vitesse en fonction du temps')
plt.xlabel('Temps en s')
plt.ylabel('Module de la vitesse')
plt.grid()
plt.show()
```

On constate que plus la vitesse augmente au cours du temps, on en déduit qu'un cyclotron permet d'accéler la vitesse d'un proton.
