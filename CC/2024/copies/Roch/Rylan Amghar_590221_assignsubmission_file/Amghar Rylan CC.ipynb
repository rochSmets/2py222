```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

## Ensuite on définit nos variables et notre suite de récurrence.
## Puis, on calcule les valeurs de u_n/U :

```python
# Précision il faut faire marcher toutes les cellules 2 fois pour obtenir les résulats (pour une raison que j'ignore).


N = 1000                                # On prend N = 1000 
w = np.empty(N, dtype = 'int64')        # On créé tableau un tableau formé selon N
ns = np.arange(1000)                    # Tableau des 1000 premiers entiers
u = np.empty(N, dtype = 'int64')        # On créé tableau un tableau formé selon N
    
for i in range (1000) :                 # On utlise une boucle for pour calculer 1000 valeurs de w puis de la suite u_n
    w = np.pi / (i+1)
    u[i] = (2-w**2)*u[i-1]-u[i-2]
    
    
plt.plot(ns, u)                         # On dessine le graphe 
plt.xlabel("N")
plt.ylabel("u_n")
```

## Les conditions au limite semblent respectés.

## On représente les valeurs du facteur A :

```python
A = np.empty(N, dtype = 'int64')          # On créé tableau un tableau formé selon N
A[0] = 0
A[1] = 0

for i in range(1, N) :                    # On utilise une boucle for pour calculer les facteurs A
    A[i] = 1/(2-w**2)*u[i]-u[i-1]
    
plt.loglog(ns, A)                         # On trace le graphe avec une double échelle logarithmique car l'échelle logarithmique sur les ordonnées ne donnent pas un bon graphe
plt.show()

#plt.semilogy(ns, A)                      # Vous pouver l'essayer en enlevant le # 
#plt.show()
```
